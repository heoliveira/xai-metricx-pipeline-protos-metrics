```python
import pandas as pd
import numpy as np
from IPython.display import display

# =========================================================
# Funções robustas
# =========================================================

def drop_prediction_columns_inplace(df: pd.DataFrame) -> list[str]:
    """
    Remove colunas de predição ('class_pred', 'proba_pred','y_pred', 'y_pred_proba','classe_predita_model', 'prob_classe_predita_model') se existirem.
    Retorna a lista de colunas removidas.
    """
    cols_to_drop = [c for c in ["class_pred", "proba_pred", "y_pred", "y_pred_proba", "classe_predita_model", "prob_classe_predita_model"] if c in df.columns]
    if cols_to_drop:
        df.drop(columns=cols_to_drop, inplace=True)
    return cols_to_drop

def enforce_instancia_first_column_inplace(df: pd.DataFrame) -> list[str]:
    """Garante que 'instancia' exista e remove colunas antes dela."""
    if "instancia" not in df.columns:
        raise ValueError("Coluna obrigatória 'instancia' não encontrada no DataFrame.")

    cols = list(df.columns)
    idx = cols.index("instancia")
    removed_cols = cols[:idx]
    if removed_cols:
        df.drop(columns=removed_cols, inplace=True)
    return removed_cols


def remove_mean_median_rows_inplace(df: pd.DataFrame) -> int:
    """Remove linhas com MEDIA/MÉDIA/MEDIANA em qualquer coluna."""
    pattern = r"^(MEDIA|MÉDIA|MEDIANA)$"

    mask = df.apply(
        lambda row: row.astype(str).str.upper().str.strip().str.match(pattern).any(),
        axis=1
    )
    removed = int(mask.sum())
    if removed:
        df.drop(index=df[mask].index, inplace=True)
    return removed


def remove_embedded_header_rows_inplace(df: pd.DataFrame) -> int:
    """
    Remove linhas que parecem ser "título" dentro do corpo.
    Regras:
      - instancia == 'instancia' (case-insensitive)
      - ou linha com muitos valores iguais aos nomes das colunas
    """
    removed = 0

    # 1) instancia == 'instancia'
    inst = df["instancia"].astype(str).str.strip().str.lower()
    mask1 = inst.eq("instancia")

    # 2) linha que repete nomes de colunas (heurística)
    colset = set([str(c).strip().lower() for c in df.columns])
    def row_looks_like_header(row):
        vals = [str(v).strip().lower() for v in row.values]
        hits = sum([1 for v in vals if v in colset])
        # se bater em muitas colunas, é bem provável ser header embutido
        return hits >= max(2, int(0.5 * len(df.columns)))

    mask2 = df.apply(row_looks_like_header, axis=1)

    mask = mask1 | mask2
    removed = int(mask.sum())
    if removed:
        df.drop(index=df[mask].index, inplace=True)

    return removed


def coerce_numeric_columns_inplace(df: pd.DataFrame) -> int:
    """
    Converte todas as colunas exceto 'instancia' para numérico.
    Trata vírgula decimal e espaços.
    Retorna quantas colunas foram convertidas.
    """
    converted = 0
    for c in df.columns:
        if c == "instancia":
            continue
        # transforma em string, limpa, troca vírgula por ponto, remove % se houver
        s = df[c].astype(str).str.strip()
        s = s.str.replace("%", "", regex=False)
        s = s.str.replace(",", ".", regex=False)

        df[c] = pd.to_numeric(s, errors="coerce")
        converted += 1
    return converted


def append_mean_median_inplace(df: pd.DataFrame) -> tuple[bool, bool]:
    """
    Adiciona MÉDIA e MEDIANA ao final usando todas as colunas numéricas (agora garantidas).
    """
    num_cols = [c for c in df.columns if c != "instancia"]
    if not num_cols:
        return (False, False)

    # cálculo ignora NaN automaticamente
    mean_vals = df[num_cols].mean(numeric_only=True)
    median_vals = df[num_cols].median(numeric_only=True)

    mean_row = {c: np.nan for c in df.columns}
    median_row = {c: np.nan for c in df.columns}

    mean_row["instancia"] = "MÉDIA"
    median_row["instancia"] = "MEDIANA"

    for c in num_cols:
        mean_row[c] = float(mean_vals[c]) if pd.notna(mean_vals[c]) else np.nan
        median_row[c] = float(median_vals[c]) if pd.notna(median_vals[c]) else np.nan

    # índices únicos
    idx_mean = "MÉDIA"
    idx_median = "MEDIANA"
    if idx_mean in df.index:
        idx_mean = f"MÉDIA_{len(df)}"
    if idx_median in df.index:
        idx_median = f"MEDIANA_{len(df)+1}"

    df.loc[idx_mean] = mean_row
    df.loc[idx_median] = median_row

    return (True, True)


# =========================================================
# Aplicação em TODOS os DataFrames (IN-PLACE) + auditoria
# =========================================================

audit_rows = []

for (dataset, method), df in dfs.items():
    rows_before, cols_before = df.shape

    removed_cols = enforce_instancia_first_column_inplace(df)
   
    removed_pred_cols = drop_prediction_columns_inplace(df)
        
    removed_mm = remove_mean_median_rows_inplace(df)
    removed_embedded_header = remove_embedded_header_rows_inplace(df)

    converted_cols = coerce_numeric_columns_inplace(df)

    added_mean, added_median = append_mean_median_inplace(df)

    rows_after, cols_after = df.shape
    
    
    
    removed_cols = enforce_instancia_first_column_inplace(df)



    # checagem rápida: quantas colunas numéricas ficaram válidas (não tudo NaN)
    num_cols = [c for c in df.columns if c != "instancia"]
    valid_numeric_cols = int((df[num_cols].notna().sum() > 0).sum()) if num_cols else 0

    audit_rows.append({
        "Dataset": dataset,
        "Método": method,
        "Cols antes": cols_before,
        "Rows antes": rows_before,
        "Colunas removidas antes de 'instancia'": ", ".join(removed_cols) if removed_cols else "-",
        "Colunas predição removidas": ", ".join(removed_pred_cols) if removed_pred_cols else "-",
        "Removidas (MEDIA/MEDIANA antigas)": removed_mm,
        "Removidas (linhas título no corpo)": removed_embedded_header,
        "Colunas convertidas p/ numérico": converted_cols,
        "Cols numéricas válidas": valid_numeric_cols,
        "Rows depois": rows_after,
        "Cols depois": cols_after,
        "MÉDIA/MEDIANA adicionadas": (added_mean and added_median),
        "Status": "✅ OK" if (added_mean and added_median) else "⚠️ NÃO ADICIONOU"
    })

audit_df = pd.DataFrame(audit_rows).sort_values(["Dataset", "Método"])
display(audit_df)

print("\nConferência rápida (exemplo): dfs[('UCI-THYROID-DXBIN','SURROGATE')]")

```

```python
import numpy as np
import pandas as pd
from IPython.display import display

LABELS_TO_EXCLUDE = {"MÉDIA", "MEDIANA", "MEDIA"}

def is_aggregated_row(instancia_series: pd.Series) -> pd.Series:
    return instancia_series.astype(str).str.upper().str.strip().isin(LABELS_TO_EXCLUDE)

def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = [str(c).strip() for c in df.columns]
    return df

def robust_numeric(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip()
    s = s.str.replace("%", "", regex=False)
    s = s.str.replace(",", ".", regex=False)
    return pd.to_numeric(s, errors="coerce")

# ---------------------------------------------------------
# Identificação de datasets, métodos e métricas
# ---------------------------------------------------------
datasets = sorted({k[0] for k in dfs.keys()})
methods  = sorted({k[1] for k in dfs.keys()})

metric_set = set()
for (_, _), df in dfs.items():
    df = normalize_columns(df)
    metric_set |= set([c for c in df.columns if c != "instancia"])
metrics = sorted(metric_set)

# ---------------------------------------------------------
# Cálculo: por dataset × método × métrica (com motivo)
# ---------------------------------------------------------
rows_long = []

for dataset in datasets:
    for method in methods:
        key = (dataset, method)
        if key not in dfs:
            continue

        df = normalize_columns(dfs[key])

        # remove linhas agregadas
        if "instancia" in df.columns:
            df = df.loc[~is_aggregated_row(df["instancia"])].copy()

        for metric in metrics:
            # ---- CASO 1: coluna não existe → valor 0.0 + log
            if metric not in df.columns:
                rows_long.append({
                    "Dataset": dataset,
                    "Método": method,
                    "Métrica": metric,
                    "%Δ (Mediana vs Média)": 0.0,
                    "Motivo": "COLUNA NÃO EXISTE"
                })
                continue

            x = robust_numeric(df[metric]).dropna().values.astype(float)

            if len(x) == 0:
                rows_long.append({
                    "Dataset": dataset,
                    "Método": method,
                    "Métrica": metric,
                    "%Δ (Mediana vs Média)": np.nan,
                    "Motivo": "SEM DADOS NUMÉRICOS (tudo NaN após conversão)"
                })
                continue

            mean_v = float(np.mean(x))
            med_v  = float(np.median(x))

            # ---- CASO 2: divisão por zero → valor 0.0 + log
            if mean_v == 0.0:
                rows_long.append({
                    "Dataset": dataset,
                    "Método": method,
                    "Métrica": metric,
                    "%Δ (Mediana vs Média)": 0.0,
                    "Motivo": "MÉDIA = 0 (divisão por zero)"
                })
                continue

            pct = ((med_v - mean_v) / mean_v) * 100.0

            rows_long.append({
                "Dataset": dataset,
                "Método": method,
                "Métrica": metric,
                "%Δ (Mediana vs Média)": pct,
                "Motivo": "OK"
            })

# ---------------------------------------------------------
# Tabela LONGA (auditoria / inspeção)
# ---------------------------------------------------------
df_long = pd.DataFrame(rows_long).sort_values(["Dataset", "Método", "Métrica"])
#display(df_long)

print("\nDiagnóstico (motivos distintos de OK):")
#display(df_long[df_long["Motivo"] != "OK"][["Dataset","Método","Métrica","Motivo"]])

# ---------------------------------------------------------
# Matriz final (Dataset | Método nas linhas, Métricas nas colunas)
# ---------------------------------------------------------
matrix = df_long.pivot_table(
    index=["Dataset", "Método"],
    columns="Métrica",
    values="%Δ (Mediana vs Média)",
    aggfunc="first"
).reset_index()

matrix = matrix[["Dataset", "Método"] + [c for c in matrix.columns if c not in ("Dataset", "Método")]]

display(matrix)

print(
    "\nDefinição usada:\n"
    "  %Δ = ((Mediana - Média) / Média) * 100\n"
    "Interpretação:\n"
    "  %Δ < 0  → Mediana < Média\n"
    "  %Δ > 0  → Mediana > Média\n"
)
```

```python
# =========================================================
# Construção de 1 DataFrame POR MÉTRICA
# Formato:
#   linhas   -> Dataset
#   colunas  -> Métodos XAI
#   valores  -> estatística agregada (MEDIANA por padrão)
#
# Saída:
#   metric_tables = {
#       "fidelity(%)": DataFrame,
#       "infidelity(%)": DataFrame,
#       ...
#   }
# =========================================================

import numpy as np
import pandas as pd
from IPython.display import display

LABELS_TO_EXCLUDE = {"MÉDIA", "MEDIANA", "MEDIA"}

def is_aggregated_row(instancia_series: pd.Series) -> pd.Series:
    return instancia_series.astype(str).str.upper().str.strip().isin(LABELS_TO_EXCLUDE)

# ---------------------------------------------------------
# Configuração
# ---------------------------------------------------------
AGGREGATION = "MEDIANA"   # trocar para "MÉDIA" se quiser

datasets = sorted({k[0] for k in dfs.keys()})
methods  = sorted({k[1] for k in dfs.keys()})

# coleta todas as métricas possíveis
metric_set = set()
for (_, _), df in dfs.items():
    metric_set |= set([c for c in df.columns if c != "instancia"])
metrics = sorted(metric_set)

# ---------------------------------------------------------
# Construção das tabelas por métrica
# ---------------------------------------------------------
metric_tables = {}

for metric in metrics:
    rows = []

    for dataset in datasets:
        row = {"Dataset": dataset}

        for method in methods:
            key = (dataset, method)

            if key not in dfs:
                row[method] = np.nan
                continue

            df = dfs[key].copy()

            # remove linhas agregadas
            if "instancia" in df.columns:
                df = df.loc[~is_aggregated_row(df["instancia"])].copy()

            if metric not in df.columns:
                row[method] = np.nan
                continue

            x = pd.to_numeric(df[metric], errors="coerce").dropna().values.astype(float)

            if len(x) == 0:
                row[method] = np.nan
                continue

            if AGGREGATION == "MEDIANA":
                row[method] = float(np.median(x))
            else:  # MÉDIA
                row[method] = float(np.mean(x))

        rows.append(row)

    metric_df = pd.DataFrame(rows)
    metric_tables[metric] = metric_df

# ---------------------------------------------------------
# Exemplo de visualização
# ---------------------------------------------------------
print(f"Estatística agregada usada: {AGGREGATION}\n")


print('metric = metrics[10]')
print('print(f\"Tabela exemplo para a métrica: {metric}\")')
print('display(metric_tables[metric]')



metric = metrics[0]
print(f"Tabela exemplo para a métrica: {metric}")
display(metric_tables[metric])

```

```python
# =========================================================
# Friedman + Nemenyi para UMA métrica
# Entrada:
#   metric_df: DataFrame no formato
#     Dataset | ANCHOR | LIME | PFI | SHAP | SURROGATE
#
# Saída:
#   - ranks por dataset
#   - rank médio por método
#   - p-valor do Friedman
#   - matriz Nemenyi
# =========================================================

import numpy as np
import pandas as pd
from scipy.stats import friedmanchisquare
import scikit_posthocs as sp
from IPython.display import display

def run_friedman_nemenyi(metric_df: pd.DataFrame, higher_is_better: bool = True):
    """
    Executa ranking + Friedman + Nemenyi para uma métrica.
    
    higher_is_better:
        True  -> maior valor = melhor (rank 1)
        False -> menor valor = melhor (rank 1)  [ex: Infidelity]
    """

    # -----------------------------
    # Preparação
    # -----------------------------
    df = metric_df.copy()
    methods = [c for c in df.columns if c != "Dataset"]

    # remove datasets incompletos (NaN em algum método)
    df_clean = df.dropna(subset=methods)

    if df_clean.shape[0] < 2:
        raise ValueError("Número insuficiente de datasets após remoção de NaN.")

    # -----------------------------
    # Ranking por dataset
    # -----------------------------
    ranks = df_clean[methods].rank(
        axis=1,
        ascending=not higher_is_better,  # False = maior é melhor
        method="average"
    )

    ranks["Dataset"] = df_clean["Dataset"]
    ranks = ranks[["Dataset"] + methods]

    # -----------------------------
    # Friedman test
    # -----------------------------
    stat, p_value = friedmanchisquare(
        *[ranks[m] for m in methods]
    )

    # -----------------------------
    # Rank médio
    # -----------------------------
    mean_ranks = ranks[methods].mean().sort_values()

    # -----------------------------
    # Nemenyi post-hoc
    # -----------------------------
    nemenyi = sp.posthoc_nemenyi_friedman(
        df_clean[methods].values
    )
    nemenyi.columns = methods
    nemenyi.index = methods

    # -----------------------------
    # Exibição
    # -----------------------------
    print("Ranking por dataset:")
    display(ranks)

    print("\nRank médio dos métodos (quanto menor, melhor):")
    display(mean_ranks.to_frame("Rank médio"))

    print(f"\nFriedman test: estatística = {stat:.4f} | p-valor = {p_value:.6f}")

    if p_value < 0.05:
        print("→ Diferença estatisticamente significativa (α = 0.05)")
    else:
        print("→ NÃO foi detectada diferença estatisticamente significativa (α = 0.05)")

    print("\nNemenyi post-hoc (p-valores):")
    display(nemenyi)

    return {
        "ranks": ranks,
        "mean_ranks": mean_ranks,
        "friedman_stat": stat,
        "friedman_p": p_value,
        "nemenyi": nemenyi
    }

print('# Exemplo de uso: Fidelity (maior é melhor)')
print("result_fidelity = run_friedman_nemenyi(metric_tables['fidelity(%)'], higher_is_better=True)")

print('# Exemplo de uso: Infidelity (menor é melhor)')
print("result_infidelity = run_friedman_nemenyi(metric_tables['infidelity(%)'], higher_is_better=False)")
```

```python
# =========================================================
# Diagrama Crítico (Critical Difference Diagram) — versão melhorada
# - evita sobreposição de valores e rótulos
# - alterna posições (zig-zag)
# - conectores em múltiplos níveis
# =========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare
import scikit_posthocs as sp
from IPython.display import display

def plot_critical_diagram_pretty(
    metric_df: pd.DataFrame,
    *,
    higher_is_better: bool = True,
    alpha: float = 0.05,
    title: str | None = None,
    show_table: bool = True
):
    """
    Gera diagrama crítico usando ranks médios e CD de Nemenyi.
    - Usa rank 1 = melhor.
    - Remove datasets com NaN (linha incompleta).
    - Exibe p-valores do Nemenyi e melhora estética do diagrama.
    """

    df = metric_df.copy()
    methods = [c for c in df.columns if c != "Dataset"]

    # remove datasets incompletos
    df_clean = df.dropna(subset=methods).reset_index(drop=True)
    if df_clean.shape[0] < 2:
        raise ValueError("Número insuficiente de datasets após remoção de NaN para gerar o diagrama.")

    X = df_clean[methods].astype(float).values
    N, k = X.shape

    # ranking por dataset
    ranks_df = pd.DataFrame(X, columns=methods).rank(axis=1, ascending=not higher_is_better, method="average")

    # Friedman sobre ranks (correto)
    stat, p_value = friedmanchisquare(*[ranks_df[m].values for m in methods])
    print(f"Friedman (ranks): estatística={stat:.4f} | p-valor={p_value:.6f} | N={N} | k={k}")

    # Nemenyi post-hoc (p-valores)
    nemenyi = sp.posthoc_nemenyi_friedman(X)
    nemenyi.index = methods
    nemenyi.columns = methods
    if show_table:
        print("\nNemenyi post-hoc (p-valores):")
        display(nemenyi)

    # ranks médios
    mean_ranks = ranks_df.mean(axis=0).sort_values()  # menor=melhor
    ordered_methods = mean_ranks.index.tolist()
    ordered_ranks = mean_ranks.values

    # ---------------------------------------------------------
    # CD (Critical Difference) — usando q_alpha tabelado p/ k=5
    # (seu caso típico). Para outros k, mantém valor aproximado.
    # ---------------------------------------------------------
    if abs(alpha - 0.05) < 1e-12 and k == 5:
        q_alpha = 2.728
    else:
        # fallback aproximado (conservador) — se quiser, a gente melhora depois para k diferente
        q_alpha = 2.728

    CD = q_alpha * np.sqrt(k * (k + 1) / (6.0 * N))
    print(f"CD (Nemenyi) ≈ {CD:.4f}")

    # ---------------------------------------------------------
    # Grupos não-significativos (aprox via CD no rank médio):
    # conecta sequências contíguas onde (rank[j]-rank[i]) <= CD
    # ---------------------------------------------------------
    groups = []
    i = 0
    while i < k:
        j = i
        while (j + 1 < k) and ((ordered_ranks[j + 1] - ordered_ranks[i]) <= CD):
            j += 1
        if j > i:
            groups.append((i, j))
        i += 1

    # ---------------------------------------------------------
    # PLOT — estética melhorada
    # ---------------------------------------------------------
    fig_w = max(11, 1.6 * k + 6)
    fig, ax = plt.subplots(figsize=(fig_w, 3.6))

    xmin, xmax = 1, k
    ax.set_xlim(xmin - 0.25, xmax + 0.25)
    #ax.set_ylim(0, 1)
    ax.set_ylim(0, 1.08)
    ax.set_yticks([])
    ax.set_xlabel("Rank médio (menor = melhor)")

    if title is None:
        title = "Diagrama Crítico (Friedman + Nemenyi)"
    ax.set_title(title)

    # linha base (eixo dos ranks)
    y_axis = 0.55
    ax.hlines(y_axis, xmin, xmax, linewidth=1)

    # ticks mais bonitos
    ax.set_xticks(np.arange(1, k + 1, 0.5))
    ax.grid(axis="x", linestyle=":", linewidth=0.6, alpha=0.6)

    # ----------------------------
    # Método + rank (sem sobrepor)
    # Alterna posições em zig-zag
    # ----------------------------
    y_label_low  = 0.33
    y_label_high = 0.41
    y_value_low  = 0.63
    y_value_high = 0.71

    for idx, (m, r) in enumerate(zip(ordered_methods, ordered_ranks)):
        # marca do método na linha
        ax.vlines(r, y_axis - 0.05, y_axis + 0.05, linewidth=1.4)

        # alterna posições (zig-zag)
        if idx % 2 == 0:
            y_lbl = y_label_low
            y_val = y_value_high
        else:
            y_lbl = y_label_high
            y_val = y_value_low

        # nome do método
        ax.text(
            r, y_lbl, m,
            ha="center", va="top",
            fontsize=10
        )

        # valor do rank (com caixa leve para legibilidade)
        ax.text(
            r, y_val, f"{r:.2f}",
            ha="center", va="bottom",
            fontsize=10,
            bbox=dict(boxstyle="round,pad=0.18", facecolor="white", edgecolor="none", alpha=0.85)
        )

        # conector fino entre nome e ponto, para “colar” visualmente
        ax.plot([r, r], [y_lbl + 0.02, y_axis - 0.05], linewidth=0.8, alpha=0.7)

    # ----------------------------
    # Barra do CD (bem acima)
    # ----------------------------
    cd_y = 0.90
    cd_x2 = xmax
    cd_x1 = max(xmin, cd_x2 - CD)
    ax.hlines(cd_y, cd_x1, cd_x2, linewidth=3)
    ax.vlines([cd_x1, cd_x2], cd_y - 0.03, cd_y + 0.03, linewidth=2)
    ax.text((cd_x1 + cd_x2) / 2, cd_y + 0.05, f"CD ≈ {CD:.2f}", ha="center", va="bottom", fontsize=10)

    # ----------------------------
    # Conectores dos grupos não-significativos
    # múltiplos níveis y para evitar sobreposição
    # ----------------------------
    g_y = 0.78
    step = 0.06
    for (a, b) in groups:
        x1 = ordered_ranks[a]
        x2 = ordered_ranks[b]
        ax.hlines(g_y, x1, x2, linewidth=4)
        g_y -= step

    # rodapé com info do teste (opcional e útil)
    ax.text(
        xmin - 0.2, 0.02,
        f"Friedman p={p_value:.4f} | N={N} datasets | k={k} métodos | α={alpha}",
        ha="left", va="bottom", fontsize=9
    )

    plt.tight_layout()
    plt.show()

    return {
        "friedman_stat": stat,
        "friedman_p": p_value,
        "nemenyi_pvals": nemenyi,
        "mean_ranks": mean_ranks,
        "CD": CD,
        "ordered_methods": ordered_methods
    }

print("res_cd_fidelity = plot_critical_diagram_pretty(")
print("    metric_tables['fidelity(%)'],")
print("    higher_is_better=True,")
print("    alpha=0.05,")
print("    title='Fidelity — Diagrama Crítico (5 datasets)'")

```

```python
# === FRIEDMAN: comparar (por métrica) se os DATASETS diferem, usando MÉTODOS como blocos ===
# H0 (por métrica): não há diferença sistemática entre datasets (mesmos ranks) -> pode agregar (inter-datasets)
# H1: pelo menos um dataset difere -> NÃO agregar sem pós-teste (Nemenyi/Wilcoxon+Holm)

import numpy as np
import pandas as pd

# SciPy (Friedman)
try:
    from scipy.stats import friedmanchisquare
except Exception as e:
    raise ImportError("Instale/ative SciPy para usar friedmanchisquare: pip install scipy") from e

ALPHA = 0.05

METHODS  = ["PFI", "SHAP", "LIME", "SURROGATE", "ANCHOR"]
DATASETS = ["FHS", "SEPSIS", "STROKE", "UCI-THYROID-DXBIN", "WDBC"]

# remove coverage(%) se quiser (você pediu remover no plot; aqui deixo opcional)
METRICS = [m for m in metrics if str(m).strip().lower() != "coverage(%)"]

def _mean_metric_from_df(df: pd.DataFrame, metric: str) -> float:
    """Retorna a média da métrica por instância, ignorando linhas 'MÉDIA/MEDIANA' etc."""
    if metric not in df.columns:
        return np.nan

    # 1) remove linhas de resumo por índice
    if isinstance(df.index, pd.Index):
        idx_u = df.index.astype(str).str.strip().str.upper()
        df = df.loc[~idx_u.isin(["MÉDIA", "MEDIA", "MEDIANA", "MEDIAN"])].copy()

    # 2) remove linhas de resumo pela coluna 'instancia' (se existir)
    if "instancia" in df.columns:
        inst_u = df["instancia"].astype(str).str.strip().str.upper()
        df = df.loc[~inst_u.isin(["MÉDIA", "MEDIA", "MEDIANA", "MEDIAN"])].copy()

    # 3) garante numérico
    vals = pd.to_numeric(df[metric], errors="coerce")
    return float(np.nanmean(vals.values))

rows = []

for met in METRICS:
    # monta matriz: linhas = métodos (blocos), colunas = datasets (tratamentos)
    mat = []
    used_methods = []

    for method in METHODS:
        row = []
        for ds in DATASETS:
            key = (ds, method)
            if key not in dfs:
                row.append(np.nan)
                continue
            df = dfs[key]
            row.append(_mean_metric_from_df(df, met))
        row = np.array(row, dtype=float)

        # só usa o método se ele tiver valor em TODOS os datasets (sem NaN)
        if np.isfinite(row).all():
            mat.append(row)
            used_methods.append(method)

    mat = np.array(mat, dtype=float)

    # checagens mínimas
    if mat.shape[0] < 2:
        rows.append({
            "metric": met,
            "n_methods_used": mat.shape[0],
            "methods_used": ", ".join(used_methods),
            "friedman_chi2": np.nan,
            "p_value": np.nan,
            "alpha": ALPHA,
            "reject_H0?": "N/A",
            "decisao": "Dados insuficientes (precisa >=2 métodos completos em todos os datasets)"
        })
        continue

    if mat.shape[1] < 2:
        rows.append({
            "metric": met,
            "n_methods_used": mat.shape[0],
            "methods_used": ", ".join(used_methods),
            "friedman_chi2": np.nan,
            "p_value": np.nan,
            "alpha": ALPHA,
            "reject_H0?": "N/A",
            "decisao": "Dados insuficientes (precisa >=2 datasets)"
        })
        continue

    # friedmanchisquare recebe cada tratamento como um array (1 por dataset), contendo blocos (métodos)
    args = [mat[:, j] for j in range(mat.shape[1])]
    stat, p = friedmanchisquare(*args)

    reject = bool(p < ALPHA)
    rows.append({
        "metric": met,
        "n_methods_used": mat.shape[0],
        "methods_used": ", ".join(used_methods),
        "friedman_chi2": float(stat),
        "p_value": float(p),
        "alpha": ALPHA,
        "reject_H0?": "SIM" if reject else "NÃO",
        "decisao": ("NÃO agregar sem pós-teste (Nemenyi/Wilcoxon+Holm)"
                    if reject else
                    "OK agregar inter-datasets (H0 não rejeitada)")
    })

df_friedman = pd.DataFrame(rows).sort_values(["p_value", "metric"], ascending=[True, True]).reset_index(drop=True)

# formatação opcional (para ficar bonito no notebook)
pd.set_option("display.max_colwidth", 120)
df_friedman
```

```python
# ============================================================
# HEATMAPS (Pearson & Spearman) — COMPLETO (SEM depender de corr_pearson_f)
# - Recria: agg, corr_pearson, corr_spearman
# - Filtra por limiar THRESH (mantém só |corr|>=THRESH; resto vira NaN)
# - Plota 2 heatmaps com cores suaves + fonte menor + valores em negrito
# - Colorbar no lugar certo
# - Gera 2 tabelas: pares correlacionados (Pearson e Spearman)
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# ----------------------------
# CONFIG (AJUSTE AQUI)
# ----------------------------
METHODS  = ["PFI", "SHAP", "LIME", "SURROGATE", "ANCHOR"]
DATASETS = ["FHS", "SEPSIS", "STROKE", "UCI-THYROID-DXBIN", "WDBC"]

THRESH = 0.6  # <<< limiar de |corr| para mostrar
USE_MEDIAN = True  # True=mediana por dataset; False=média

# Grupo B (se df_friedman existir, usa; senão usa fallback)
FALLBACK_GROUPB = [
    "consistency(%)",
    "directional soundness(%)",
    "sufficiency-20(%)",
    "infidelity(%)",
    "soundness(%)",
    "simplicity(%)",
    "selectivity(%)",
    "robustez(%)",
    "stability(%)",
    "completeness(%)",
]

# ----------------------------
# Ajustes visuais
# ----------------------------
CMAP = "RdBu_r"
VMIN, VMAX = -1, 1
CELL_FONT = 7        # menor
ALPHA_IMG = 0.72     # cores mais suaves (0.65..0.80)
TICK_FONT = 8

# ----------------------------
# Seleção de métricas (Grupo B)
# ----------------------------
if "df_friedman" in globals():
    group_metrics = df_friedman.loc[
        df_friedman["reject_H0?"].astype(str).str.upper() == "NÃO",
        "metric"
    ].tolist()
else:
    group_metrics = FALLBACK_GROUPB

# remove coverage se aparecer
group_metrics = [m for m in group_metrics if str(m).strip().lower() != "coverage(%)"]

# ----------------------------
# Helpers
# ----------------------------
def _reduce_metric_from_df(df: pd.DataFrame, metric: str, use_median=True) -> float:
    """Agrega por instância ignorando linhas de resumo."""
    if metric not in df.columns:
        return np.nan

    df2 = df.copy()

    # remove linhas-resumo pelo índice
    idx_u = df2.index.astype(str).str.strip().str.upper()
    df2 = df2.loc[~idx_u.isin(["MÉDIA","MEDIA","MEDIANA","MEDIAN"])]

    # remove linhas-resumo por coluna instancia (se existir)
    if "instancia" in df2.columns:
        inst_u = df2["instancia"].astype(str).str.strip().str.upper()
        df2 = df2.loc[~inst_u.isin(["MÉDIA","MEDIA","MEDIANA","MEDIAN"])]

    vals = pd.to_numeric(df2[metric], errors="coerce").to_numpy(dtype=float)

    if use_median:
        return float(np.nanmedian(vals))
    return float(np.nanmean(vals))

def _build_agg(dfs_dict, metrics, methods, datasets, use_median=True) -> pd.DataFrame:
    """
    agg: linhas=métricas, colunas=métodos
    Cada célula = mediana (ou média) entre datasets da métrica do método.
    """
    agg = pd.DataFrame(index=metrics, columns=methods, dtype=float)
    for met in metrics:
        for method in methods:
            per_ds = []
            for ds in datasets:
                key = (ds, method)
                if key not in dfs_dict:
                    per_ds.append(np.nan)
                    continue
                per_ds.append(_reduce_metric_from_df(dfs_dict[key], met, use_median=use_median))
            per_ds = np.array(per_ds, dtype=float)
            # agrega inter-datasets
            agg.loc[met, method] = float(np.nanmedian(per_ds)) if use_median else float(np.nanmean(per_ds))
    return agg.dropna(axis=0, how="all")

def _filter_corr(C: pd.DataFrame, thresh: float) -> pd.DataFrame:
    """Mantém diagonal e |corr|>=thresh; resto vira NaN."""
    Cf = C.copy()
    for i in range(Cf.shape[0]):
        for j in range(Cf.shape[1]):
            if i == j:
                Cf.iat[i, j] = 1.0
                continue
            v = Cf.iat[i, j]
            if not (np.isfinite(v) and abs(v) >= thresh):
                Cf.iat[i, j] = np.nan
    return Cf

def _pairs_table(M: pd.DataFrame, thresh: float, name: str) -> pd.DataFrame:
    """Extrai pares únicos (triângulo superior) com |corr| >= thresh."""
    pairs = []
    idx = list(M.index)
    for i in range(len(idx)):
        for j in range(i+1, len(idx)):
            v = M.iat[i, j]
            if np.isfinite(v) and abs(v) >= thresh:
                pairs.append((idx[i], idx[j], float(v), abs(float(v))))
    df = pd.DataFrame(pairs, columns=["metric_A", "metric_B", "corr", "abs_corr"])
    if len(df) == 0:
        return pd.DataFrame(columns=["tipo","metric_A","metric_B","corr","strength"])

    def strength(a):
        if a >= 0.90: return "muito forte"
        if a >= 0.70: return "forte"
        if a >= 0.60: return "moderada"
        return "fraca"

    df["strength"] = df["abs_corr"].apply(strength)
    df = df.sort_values(["abs_corr","metric_A","metric_B"], ascending=[False, True, True]).reset_index(drop=True)
    df = df.drop(columns=["abs_corr"])
    df.insert(0, "tipo", name)
    return df

def _heatmap(ax, M, title):
    im = ax.imshow(
        M.values,
        cmap=CMAP,
        vmin=VMIN,
        vmax=VMAX,
        interpolation="nearest",
        aspect="equal",
        alpha=ALPHA_IMG
    )
    ax.set_title(title, fontsize=12, pad=10, fontweight="bold")

    labels = [str(x).replace("(%)", "") for x in M.index]
    ax.set_xticks(np.arange(len(labels)))
    ax.set_yticks(np.arange(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha="right", fontsize=TICK_FONT)
    ax.set_yticklabels(labels, fontsize=TICK_FONT)

    # valores dentro (negrito, menor)
    for i in range(M.shape[0]):
        for j in range(M.shape[1]):
            v = M.iat[i, j]
            if np.isfinite(v):
                color = "black" if abs(v) < 0.75 else "white"
                ax.text(
                    j, i, f"{v:.1f}",
                    ha="center", va="center",
                    fontsize=CELL_FONT,
                    fontweight="bold",
                    color=color
                )

    ax.tick_params(axis="both", which="both", length=0)
    for spine in ax.spines.values():
        spine.set_linewidth(0.8)

    return im

# ============================================================
# 1) Recria agg -> corr -> corr_filtradas
# ============================================================
# OBS: este código assume que "dfs" existe no seu notebook.
# Se "dfs" não existir, vai dar NameError (aí você precisa carregar a estrutura).
agg = _build_agg(dfs, group_metrics, METHODS, DATASETS, use_median=USE_MEDIAN)

corr_pearson  = agg.T.corr(method="pearson",  min_periods=2)
corr_spearman = agg.T.corr(method="spearman", min_periods=2)

corr_pearson_f  = _filter_corr(corr_pearson,  THRESH)
corr_spearman_f = _filter_corr(corr_spearman, THRESH)

print("=== AGG (métrica x método) usada nas correlações ===")
display(agg)

# ============================================================
# 2) PLOT (colorbar no lugar certo)
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(15, 6), constrained_layout=True)

im1 = _heatmap(axes[0], corr_pearson_f,  f"Pearson (|ρ| ≥ {THRESH})")
im2 = _heatmap(axes[1], corr_spearman_f, f"Spearman (|ρ| ≥ {THRESH})")

# colorbar única, à direita, sem sobrepor
cbar = fig.colorbar(im2, ax=axes, location="right", shrink=0.92, pad=0.02)
cbar.set_label("Correlação", fontsize=10)
cbar.ax.tick_params(labelsize=9)

plt.show()

# ============================================================
# 3) Tabelas finais: pares correlacionados
# ============================================================
df_pairs_pearson  = _pairs_table(corr_pearson,  THRESH, "Pearson")
df_pairs_spearman = _pairs_table(corr_spearman, THRESH, "Spearman")

print("=== PARES CORRELACIONADOS (Pearson) ===")
display(df_pairs_pearson)

print("=== PARES CORRELACIONADOS (Spearman) ===")
display(df_pairs_spearman)
```

```python
# ============================================================
# RESUMO SPEARMAN POR MÉTRICA (média das correlações + lista de relacionadas)
# - Reusa o seu fluxo: agg -> corr_spearman
# - Cria df_spearman_summary com:
#   (i) metric
#   (ii) mean_spearman_abs (média de |rho| com as correlacionadas acima do THRESH)
#   (iii) mean_spearman_signed (média de rho (com sinal) com as correlacionadas acima do THRESH)
#   (iv) related_metrics (lista/str das métricas com |rho|>=THRESH, com valor e sinal)
#   (v) n_related
#
# Obs.: Se quiser considerar TODAS as correlações (sem THRESH), troque use_thresh=False
# ============================================================

import numpy as np
import pandas as pd

# --------- GARANTE QUE corr_spearman EXISTA ----------
# (Se você rodou o seu código acima, isso já existe. Se não, recria.)
if "corr_spearman" not in globals():
    # Recria agg e corr_spearman a partir do seu setup
    agg = _build_agg(dfs, group_metrics, METHODS, DATASETS, use_median=USE_MEDIAN)
    corr_spearman = agg.T.corr(method="spearman", min_periods=2)

def _spearman_summary_df(corr_spearman: pd.DataFrame, thresh: float = 0.6, use_thresh: bool = True) -> pd.DataFrame:
    """
    Para cada métrica (linha/coluna em corr_spearman), retorna:
      - média de |rho| (apenas relações acima do limiar, se use_thresh=True)
      - média de rho (com sinal, apenas acima do limiar, se use_thresh=True)
      - lista de métricas relacionadas e seus rhos
    """
    metrics = list(corr_spearman.index)
    rows = []

    for m in metrics:
        # série de correlações da métrica m com todas as outras
        s = corr_spearman.loc[m].copy()

        # remove self-corr
        s = s.drop(index=m, errors="ignore")

        # mantém só valores finitos
        s = s[np.isfinite(s.values)]

        if use_thresh:
            s_rel = s[s.abs() >= thresh]
        else:
            s_rel = s

        if len(s_rel) == 0:
            rows.append({
                "metric": m,
                "mean_spearman_abs": np.nan,
                "mean_spearman_signed": np.nan,
                "n_related": 0,
                "related_metrics": ""
            })
            continue

        # ordena por força (|rho| desc)
        s_rel = s_rel.reindex(s_rel.abs().sort_values(ascending=False).index)

        related_list = [f"{k} ({v:+.3f})" for k, v in s_rel.items()]

        rows.append({
            "metric": m,
            "mean_spearman_abs": float(np.nanmean(np.abs(s_rel.values))),
            "mean_spearman_signed": float(np.nanmean(s_rel.values)),
            "n_related": int(len(s_rel)),
            "related_metrics": "; ".join(related_list)
        })

    df = pd.DataFrame(rows)

    # ordena: primeiro quem tem mais relações, depois maior média |rho|
    df = df.sort_values(["n_related", "mean_spearman_abs", "metric"], ascending=[False, False, True]).reset_index(drop=True)
    return df

# --------- GERA O DATAFRAME FINAL ----------
df_spearman_summary = _spearman_summary_df(
    corr_spearman=corr_spearman,
    thresh=THRESH,
    use_thresh=True  # usa o mesmo THRESH do seu heatmap
)

print("=== RESUMO SPEARMAN POR MÉTRICA (|rho|>=THRESH) ===")
pd.set_option("display.max_colwidth", 180)
display(df_spearman_summary)

# (Opcional) salvar
df_spearman_summary.to_csv("spearman_summary_by_metric_rf.csv", index=False)
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.lines import Line2D

all_metrics = {
    "fidelity",
    "faithfulness",
    "sufficiency-1",
    "sufficiency-5",
    "sufficiency-10",
    "sufficiency-20",
    "soundness",
    "robustez",
    "selectivity",
    "completeness",
    "stability",
    "infidelity",
    "directional soundness",
    "consistency",
    "simplicity",
}

# ============================================================
# Lê os CSVs gerados no mesmo diretório do notebook
# ============================================================
rf_csv  = "spearman_summary_by_metric_rf.csv"
xgb_csv = "spearman_summary_by_metric_xgb.csv"

df_rf  = pd.read_csv(rf_csv)
df_xgb = pd.read_csv(xgb_csv)

# ============================================================
# Helpers
# ============================================================
def normalize_metric_name(name: str) -> str:
    """
    Remove '(%)', espaços extras e padroniza caixa.
    """
    s = str(name).strip()
    s = s.replace("(%)", "").strip()
    s = re.sub(r"\s+", " ", s)
    return s.lower()

def parse_related_metrics_cell(cell: str):
    """
    Converte uma célula do tipo:
      'sufficiency-20(%) (+1.000); completeness(%) (+0.718); ...'
    em lista de tuplas:
      [('sufficiency-20', 1.0), ('completeness', 0.718), ...]
    """
    if pd.isna(cell) or not str(cell).strip():
        return []

    parts = [p.strip() for p in str(cell).split(";") if p.strip()]
    out = []

    for p in parts:
        # captura:
        # nome da métrica + valor entre parênteses
        m = re.match(r"^(.*?)\s*\(([+-]?\d+(?:\.\d+)?)\)\s*$", p)
        if not m:
            # fallback: tenta achar último (...) como rho
            m2 = re.match(r"^(.*)\(([+-]?\d+(?:\.\d+)?)\)\s*$", p)
            if not m2:
                continue
            metric_raw = m2.group(1).strip()
            rho = float(m2.group(2))
        else:
            metric_raw = m.group(1).strip()
            rho = float(m.group(2))

        metric = normalize_metric_name(metric_raw)
        out.append((metric, rho))

    return out

def build_edge_dict(summary_df: pd.DataFrame):
    """
    A partir do CSV de resumo por métrica, monta um dict de arestas:
      {('a','b'): rho}
    Considera coluna 'metric' e 'related_metrics'.
    """
    edge_dict = {}

    for _, row in summary_df.iterrows():
        src = normalize_metric_name(row["metric"])
        related = parse_related_metrics_cell(row["related_metrics"])

        for dst, rho in related:
            a, b = sorted([src, dst])
            edge_dict[(a, b)] = rho  # mantém último valor, suficiente para grafo

    return edge_dict

def pretty_label(metric: str) -> str:
    """
    Rótulo mais bonito para exibição.
    """
    mapping = {
        "directional soundness": "Directional Soundness",
        "selectivity": "Selectivity",
        "sufficiency-20": "Sufficiency-20",
        "sufficiency-10": "Sufficiency-10",
        "sufficiency-5": "Sufficiency-5",
        "sufficiency-1": "Sufficiency-1",
        "robustez": "Robustez",
        "completeness": "Completeness",
        "consistency": "Consistency",
        "stability": "Stability",
        "simplicity": "Simplicity",
        "soundness": "Soundness",
        "infidelity": "Infidelity",
        "faithfulness": "Faithfulness",
        "fidelity": "Fidelity",
    }
    return mapping.get(metric, metric.title())

# ============================================================
# Monta arestas RF / XGB
# ============================================================
rf_edges_dict  = build_edge_dict(df_rf)
xgb_edges_dict = build_edge_dict(df_xgb)

rf_edges  = set(rf_edges_dict.keys())
xgb_edges = set(xgb_edges_dict.keys())

common_edges = rf_edges & xgb_edges
rf_only      = rf_edges - xgb_edges
xgb_only     = xgb_edges - rf_edges

# ============================================================
# Cria grafo
# ============================================================
G = nx.Graph()

# métricas independentes (exemplo)
all_metrics = set(df_rf["metric"].map(normalize_metric_name)).union(
    set(df_xgb["metric"].map(normalize_metric_name))
)

connected_metrics = set()
for a, b in rf_edges.union(xgb_edges):
    connected_metrics.add(a)
    connected_metrics.add(b)

independent_metrics = all_metrics - connected_metrics

for m in independent_metrics:
    G.add_node(m)

# adiciona todas as arestas com grupo
for edge in common_edges:
    G.add_edge(*edge, group="both", rho_rf=rf_edges_dict[edge], rho_xgb=xgb_edges_dict[edge])

for edge in rf_only:
    G.add_edge(*edge, group="rf", rho_rf=rf_edges_dict[edge])

for edge in xgb_only:
    G.add_edge(*edge, group="xgb", rho_xgb=xgb_edges_dict[edge])

# rótulos bonitos
label_map = {node: pretty_label(node) for node in list(G.nodes()) + ["faithfulness", "fidelity", "sufficiency-1"]}

# layout
pos = nx.spring_layout(G, seed=7, k=1.2)

# adiciona manualmente as 3 métricas independentes
pos["faithfulness"] = (0.45, -1.00)
pos["fidelity"] = (0.65, -1.00)
pos["sufficiency-1"] = (0.85, -1.00)

# ============================================================
# Plot
# ============================================================
plt.figure(figsize=(12, 8))

# nós sem cor (somente contorno)
# nós conectados
nx.draw_networkx_nodes(
    G, pos,
    nodelist=list(G.nodes()),
    node_size=2200,
    node_color="white",
    edgecolors="black",
    linewidths=1.8
)

# nós independentes desenhados manualmente
nx.draw_networkx_nodes(
    G,
    pos,
    nodelist=[],
    node_size=2200,
    node_color="white",
    edgecolors="black",
    linewidths=1.8
)

# desenha as 3 bolinhas independentes
for node in ["faithfulness", "fidelity", "sufficiency-1"]:
    nx.draw_networkx_nodes(
        G,
        pos,
        nodelist=[node],
        node_size=2200,
        node_color="white",
        edgecolors="black",
        linewidths=1.8
    )

# arestas
both_edges = [(u, v) for u, v, d in G.edges(data=True) if d["group"] == "both"]
rf_only_edges = [(u, v) for u, v, d in G.edges(data=True) if d["group"] == "rf"]
xgb_only_edges = [(u, v) for u, v, d in G.edges(data=True) if d["group"] == "xgb"]

nx.draw_networkx_edges(
    G,
    pos,
    edgelist=both_edges,
    width=2.8,
    style="solid",
    edge_color="red",
    arrows=True,
    arrowstyle='-',
    connectionstyle="arc3,rad=0.15"
)

nx.draw_networkx_edges(
    G,
    pos,
    edgelist=rf_only_edges,
    width=2.2,
    style="dashed",
    edge_color="blue",
    arrows=True,
    arrowstyle='-',
    connectionstyle="arc3,rad=-0.15"
)

nx.draw_networkx_edges(
    G,
    pos,
    edgelist=xgb_only_edges,
    width=2.2,
    style="dotted",
    edge_color="black",
    arrows=True,
    arrowstyle='-',
    connectionstyle="arc3,rad=0.25"
)
# labels dos nós
nx.draw_networkx_labels(
    G,
    pos,
    labels=label_map,
    font_size=10,
    font_color="black"
)

# legenda
legend_elements = [
    Line2D([0], [0], color="red", lw=2.8, linestyle="solid",
           label="Correlação presente em RF e XGBoost"),
    Line2D([0], [0], color="blue", lw=2.2, linestyle="dashed",
           label="Somente Random Forest"),
    Line2D([0], [0], color="black", lw=2.2, linestyle="dotted",
           label="Somente XGBoost"),
]

plt.legend(handles=legend_elements, loc="upper left", frameon=True)
plt.title("Estrutura de Correlação entre Métricas\n(Spearman |ρ| ≥ 0.6)")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# Grau dos vértices + agrupamento automático
# ============================================================

# garante que todas as métricas do estudo entrem no cálculo,
# mesmo que não estejam desenhadas no grafo principal
all_metrics_full = set(all_metrics)

# grau total no grafo final
degree_total = {}
for m in all_metrics_full:
    degree_total[m] = G.degree(m) if m in G.nodes() else 0

# grau considerando apenas arestas comuns aos dois classificadores
degree_common = {m: 0 for m in all_metrics_full}
for a, b in common_edges:
    degree_common[a] += 1
    degree_common[b] += 1

# tabela de graus
degree_df = pd.DataFrame({
    "metric": sorted(all_metrics_full),
    "degree_total": [degree_total[m] for m in sorted(all_metrics_full)],
    "degree_common": [degree_common[m] for m in sorted(all_metrics_full)],
})

degree_df["metric_pretty"] = degree_df["metric"].apply(pretty_label)
degree_df = degree_df[["metric", "metric_pretty", "degree_total", "degree_common"]]
degree_df = degree_df.sort_values(["degree_total", "degree_common", "metric_pretty"], ascending=[False, False, True])

print("\n=== GRAU DAS MÉTRICAS ===")
print(degree_df.to_string(index=False))

# ------------------------------------------------------------
# Regras de agrupamento
# ------------------------------------------------------------
# 1) Independentes: grau 0
independent_metrics = sorted([
    m for m in all_metrics_full
    if degree_total[m] == 0
], key=pretty_label)

# 2) Principais redundâncias:
#    conectadas em ambos os classificadores (degree_common >= 1)
#    e com alta conectividade total (grau >= 3)
main_redundancies = sorted([
    m for m in all_metrics_full
    if degree_total[m] >= 3 and degree_common[m] >= 1
], key=pretty_label)

# 3) Correlacionadas:
#    conectadas, mas sem alta redundância estrutural
#    aqui usei grau 2 ou 3, excluindo as principais redundâncias
correlated_metrics = sorted([
    m for m in all_metrics_full
    if 2 <= degree_total[m] <= 3 and m not in main_redundancies
], key=pretty_label)

# 4) Específicas:
#    baixa conectividade estrutural
#    aqui usei grau exatamente 1
specific_metrics = sorted([
    m for m in all_metrics_full
    if degree_total[m] == 1
], key=pretty_label)

# ------------------------------------------------------------
# groups no formato que você está usando
# ------------------------------------------------------------
groups = [
    ("Métricas Independentes - |ρ| < 0.6",
     [pretty_label(m) for m in independent_metrics],
     (0.12, 0.62), "#e8f1fb"),

    ("Principais Redundâncias - |ρ| ≥ 0.6 - Independe do Classificador",
     [pretty_label(m) for m in main_redundancies],
     (0.56, 0.62), "#fbeaea"),

    ("Métricas Correlacionadas - |ρ| ≥ 0.6 - Grau <= 3 conexões",
     [pretty_label(m) for m in correlated_metrics],
     (0.12, 0.18), "#eef7e8"),

    ("Métricas Específicas - |ρ| ≥ 0.6 - Baixa conectividade Grau 1",
     [pretty_label(m) for m in specific_metrics],
     (0.56, 0.18), "#f7f2e8")
]

print("\n=== GROUPS ===")
for g in groups:
    print(g[0], "->", g[1])

# opcional: salvar
degree_df.to_csv("metric_degree_table.csv", index=False)
```

```python
# =========================================================
# Diagrama Crítico (Demšar-style) — usando dfs[(dataset, metodo)]
# + TABELA DE RANKS MÉDIOS (interpretável)
# + (opcional) TABELA DE RANKS POR DATASET
# =========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare
import scikit_posthocs as sp
from IPython.display import display

# ----------------------------
# PARAMETRIZAÇÃO VISUAL
# ----------------------------
CD_BAR_COLOR    = "darkorange"   # <<< MUDE AQUI
GROUP_BAR_COLOR = "black"
GRID_ALPHA      = 0.6
GRID_LW         = 0.6

# ----------------------------
# CONFIG DO SEU EXPERIMENTO
# ----------------------------
METHODS  = ["PFI", "SHAP", "LIME", "SURROGATE", "ANCHOR"]
DATASETS = ["FHS", "SEPSIS", "STROKE", "UCI-THYROID-DXBIN", "WDBC"]

ALPHA = 0.05
LOWER_IS_BETTER = {"infidelity(%)"}  # <<< só infidelity é menor=melhor

GROUP_A_METRICS = [
    "sufficiency-1(%)",
    "sufficiency-5(%)",
    "sufficiency-10(%)",
    "fidelity(%)",
    "faithfulness(%)",
]

# ----------------------------
# Helpers
# ----------------------------
def metric_value(df, metric):
    """Mediana por instância, removendo linhas-resumo e forçando numérico."""
    if metric not in df.columns:
        return np.nan
    df2 = df.copy()
    if "instancia" in df2.columns:
        inst = df2["instancia"].astype(str).str.strip().str.upper()
        df2 = df2.loc[~inst.isin(["MÉDIA","MEDIA","MEDIANA","MEDIAN"])].copy()
    idx_u = df2.index.astype(str).str.strip().str.upper()
    df2 = df2.loc[~idx_u.isin(["MÉDIA","MEDIA","MEDIANA","MEDIAN"])].copy()

    vals = pd.to_numeric(df2[metric], errors="coerce").to_numpy(dtype=float)
    return float(np.nanmedian(vals))

def is_lower_better(metric: str) -> bool:
    return metric in LOWER_IS_BETTER

def critical_difference(k, N, alpha=0.05):
    # Seu caso típico: k=5, alpha=0.05 -> q=2.728
    if abs(alpha - 0.05) < 1e-12 and k == 5:
        q = 2.728
    else:
        q = 2.728  # fallback conservador
    return q * np.sqrt(k * (k + 1) / (6.0 * N))

def _build_metric_table_from_dfs(metric: str) -> pd.DataFrame:
    """
    DataFrame: index=datasets, columns=methods, valor=mediana da métrica.
    Mantém apenas datasets completos (sem NaN).
    """
    rows = []
    used_ds = []
    for ds in DATASETS:
        row = {}
        ok = True
        for m in METHODS:
            key = (ds, m)
            if key not in dfs:
                ok = False
                break
            v = metric_value(dfs[key], metric)
            if not np.isfinite(v):
                ok = False
                break
            row[m] = float(v)
        if ok:
            rows.append(row)
            used_ds.append(ds)

    if len(rows) == 0:
        return pd.DataFrame(columns=METHODS)

    return pd.DataFrame(rows, index=used_ds, columns=METHODS).astype(float)

def _maximal_cliques_bruteforce(nodes, edges_set):
    """Cliques maximais (k=5 ok)."""
    nodes = list(nodes)
    n = len(nodes)

    def is_clique(sub):
        for i in range(len(sub)):
            for j in range(i+1, len(sub)):
                if tuple(sorted((sub[i], sub[j]))) not in edges_set:
                    return False
        return True

    cliques = []
    for mask in range(1, 1 << n):
        sub = [nodes[i] for i in range(n) if (mask >> i) & 1]
        if len(sub) < 2:
            continue
        if is_clique(sub):
            cliques.append(sub)

    maximal = []
    for c in cliques:
        s = set(c)
        if any(s < set(d) for d in cliques):
            continue
        maximal.append(c)

    maximal.sort(key=lambda x: (-len(x), x))
    return maximal

# ----------------------------
# Função principal
# ----------------------------
def plot_critical_diagram_from_dfs(
    metric: str,
    *,
    alpha: float = 0.05,
    title: str | None = None,
    show_nemenyi_table: bool = True,
    show_rank_tables: bool = True,        # <<< NOVO
    show_rank_by_dataset: bool = False    # <<< NOVO (opcional)
):
    metric_df = _build_metric_table_from_dfs(metric)

    if metric_df.shape[0] < 2:
        raise ValueError(f"Datasets insuficientes para {metric} (>=2 completos).")

    higher_is_better = not is_lower_better(metric)

    # se menor é melhor: inverte sinal (para sempre rankear com maior=melhor)
    X = metric_df.copy().astype(float)
    if not higher_is_better:
        X = -X

    # ranks por dataset (1 = melhor)  -> como "maior=melhor", usamos ascending=False
    ranks_df = X.rank(axis=1, ascending=False, method="average")

    # ranks médios (menor=melhor)
    mean_ranks = ranks_df.mean(axis=0).sort_values()
    df_mean_ranks = mean_ranks.reset_index()
    df_mean_ranks.columns = ["method", "mean_rank"]
    df_mean_ranks["posicao"] = np.arange(1, len(df_mean_ranks) + 1)
    df_mean_ranks = df_mean_ranks[["posicao", "method", "mean_rank"]]

    # Friedman em ranks
    stat, p_value = friedmanchisquare(*[ranks_df[m].values for m in METHODS])
    print(f"\nMÉTRICA: {metric}")
    print(f"Menor é melhor? {'SIM' if (not higher_is_better) else 'NÃO'}")
    print(f"Friedman χ² = {stat:.4f} | p = {p_value:.6f} | N={X.shape[0]} | k={X.shape[1]}")

    # Tabelas de ranks (para interpretação)
    if show_rank_tables:
        print("\n=== RANKS MÉDIOS (1 = melhor) ===")
        display(df_mean_ranks)

        if show_rank_by_dataset:
            print("\n=== RANKS POR DATASET (1 = melhor) ===")
            display(ranks_df.loc[:, METHODS])

    # Nemenyi (wide)
    nemenyi = sp.posthoc_nemenyi_friedman(X)  # PASSA O DF WIDE
    if not isinstance(nemenyi, pd.DataFrame):
        nemenyi = pd.DataFrame(nemenyi, index=METHODS, columns=METHODS)
    else:
        nemenyi = nemenyi.reindex(index=METHODS, columns=METHODS)
    np.fill_diagonal(nemenyi.values, 1.0)

    if show_nemenyi_table:
        print("\nNemenyi post-hoc (p-valores):")
        display(nemenyi)

    # ordenação do plot (por rank médio)
    ordered_methods = mean_ranks.index.tolist()
    ordered_ranks   = mean_ranks.values
    N, k = X.shape

    # CD
    CD = critical_difference(k, N, alpha=alpha)

    # grafo de "não diferença" (p>=alpha)
    edges = set()
    for i in range(len(ordered_methods)):
        for j in range(i+1, len(ordered_methods)):
            a, b = ordered_methods[i], ordered_methods[j]
            pv = nemenyi.loc[a, b]
            if pd.notna(pv) and float(pv) >= alpha:
                edges.add(tuple(sorted((a, b))))

    # cliques maximais (grupos)
    cliques = _maximal_cliques_bruteforce(ordered_methods, edges)

    intervals = []
    for c in cliques:
        rs = [mean_ranks[m] for m in c]
        a, b = float(np.min(rs)), float(np.max(rs))
        if abs(b - a) < 1e-12:
            continue
        intervals.append((a, b, c))
    intervals.sort(key=lambda t: (-(t[1]-t[0]), -len(t[2])))

    # ----------------------------
    # PLOT
    # ----------------------------
    fig_w = max(11, 1.6 * k + 6)
    fig, ax = plt.subplots(figsize=(fig_w, 3.6))

    xmin, xmax = 1, k
    ax.set_xlim(xmin - 0.25, xmax + 0.25)
    ax.set_ylim(0, 1.08)
    ax.set_yticks([])
    ax.set_xlabel("Rank médio (menor = melhor)")

    if title is None:
        title = f"{metric} — Diagrama Crítico ({N} datasets)"
    ax.set_title(title, fontweight="bold")

    # linha base
    y_axis = 0.55
    ax.hlines(y_axis, xmin, xmax, linewidth=1)

    # grid
    ax.set_xticks(np.arange(1, k + 1, 0.5))
    ax.grid(axis="x", linestyle=":", linewidth=GRID_LW, alpha=GRID_ALPHA)

    # marcações + labels (zig-zag)
    y_label_low  = 0.33
    y_label_high = 0.41
    y_value_low  = 0.63
    y_value_high = 0.71

    for idx, (m, r) in enumerate(zip(ordered_methods, ordered_ranks)):
        ax.vlines(r, y_axis - 0.05, y_axis + 0.05, linewidth=1.4)

        if idx % 2 == 0:
            y_lbl = y_label_low
            y_val = y_value_high
        else:
            y_lbl = y_label_high
            y_val = y_value_low

        ax.text(r, y_lbl, m, ha="center", va="top", fontsize=9)
        ax.text(
            r, y_val, f"{r:.2f}",
            ha="center", va="bottom",
            fontsize=9,
            bbox=dict(boxstyle="round,pad=0.18", facecolor="white", edgecolor="none", alpha=0.85)
        )
        ax.plot([r, r], [y_lbl + 0.02, y_axis - 0.05], linewidth=0.8, alpha=0.7)

    # CD bar
    cd_y = 0.90
    cd_x2 = xmax
    cd_x1 = max(xmin, cd_x2 - CD)
    ax.hlines(cd_y, cd_x1, cd_x2, linewidth=3, color=CD_BAR_COLOR)
    ax.vlines([cd_x1, cd_x2], cd_y - 0.03, cd_y + 0.03, linewidth=2, color=CD_BAR_COLOR)
    ax.text((cd_x1 + cd_x2)/2, cd_y + 0.05, f"CD ≈ {CD:.2f}", ha="center", va="bottom", fontsize=9)

    # conectores (grupos não-significativos)
    g_y = 0.78
    step = 0.06
    drawn = set()
    for (a, b, c) in intervals:
        key = (round(a, 3), round(b, 3))
        if key in drawn:
            continue
        drawn.add(key)
        ax.hlines(g_y, a, b, linewidth=4, color=GROUP_BAR_COLOR)
        g_y -= step

    ax.text(
        xmin - 0.2, 0.02,
        f"Friedman p={p_value:.4f} | N={N} datasets | k={k} métodos | α={alpha} | {'maior=melhor' if higher_is_better else 'menor=melhor'}",
        ha="left", va="bottom", fontsize=9
    )

    plt.tight_layout()
    plt.show()

    return {
        "friedman_stat": stat,
        "friedman_p": p_value,
        "nemenyi_pvals": nemenyi,
        "mean_ranks": mean_ranks,
        "mean_ranks_table": df_mean_ranks,
        "ranks_by_dataset": ranks_df,
        "CD": CD,
        "ordered_methods": ordered_methods,
        "metric_table": metric_df
    }

# =========================================================
# EXEMPLOS
# =========================================================
# Um plot:
# res = plot_critical_diagram_from_dfs("sufficiency-5(%)", alpha=ALPHA, show_rank_by_dataset=True)

# Rodar Grupo A:
results_groupA = {}
for met in GROUP_A_METRICS:
    try:
        results_groupA[met] = plot_critical_diagram_from_dfs(
            met,
            alpha=ALPHA,
            title=f"{met} — Diagrama Crítico ({len(DATASETS)} datasets)",
            show_nemenyi_table=True,
            show_rank_tables=True,
            show_rank_by_dataset=False  # mude p/ True se quiser
        )
    except Exception as e:
        print(f"[{met}] ERRO: {e}")
```

```python
# ============================================================
# UPSET PLOT (Random Forest) — Sobreposição entre métricas
# baseada em correlação Spearman |rho| >= THRESH
# Dados vêm de: dfs[(dataset, method)] com métricas como colunas
#
# Requisitos:
#   pip install upsetplot
# ============================================================

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from upsetplot import UpSet, from_indicators

# ----------------------------
# CONFIG
# ----------------------------
METHODS  = ["PFI", "SHAP", "LIME", "SURROGATE", "ANCHOR"]
DATASETS = ["FHS", "SEPSIS", "STROKE", "UCI-THYROID-DXBIN", "WDBC"]

THRESH = 0.6
USE_MEDIAN = True  # agrega por dataset com mediana (recomendado)

# métricas que você quer marcar em laranja (sem sobreposição no seu critério)
HIGHLIGHT_ORANGE = {
    "sufficiency-5(%)",
    "fidelity(%)",
    "sufficiency-10(%)",
    "faithfulness(%)",
}

# ----------------------------
# Helpers
# ----------------------------
def _drop_summary_rows(df: pd.DataFrame) -> pd.DataFrame:
    df2 = df.copy()
    # remove linhas-resumo pelo índice
    idx_u = df2.index.astype(str).str.strip().str.upper()
    df2 = df2.loc[~idx_u.isin(["MÉDIA","MEDIA","MEDIANA","MEDIAN"])]
    # remove linhas-resumo por coluna instancia (se existir)
    if "instancia" in df2.columns:
        inst_u = df2["instancia"].astype(str).str.strip().str.upper()
        df2 = df2.loc[~inst_u.isin(["MÉDIA","MEDIA","MEDIANA","MEDIAN"])]
    return df2

def _reduce_metric_from_df(df: pd.DataFrame, metric: str, use_median=True) -> float:
    if metric not in df.columns:
        return np.nan
    df2 = _drop_summary_rows(df)
    vals = pd.to_numeric(df2[metric], errors="coerce").to_numpy(dtype=float)
    if vals.size == 0:
        return np.nan
    return float(np.nanmedian(vals)) if use_median else float(np.nanmean(vals))

def _infer_metrics(dfs_dict, datasets, methods) -> list:
    # pega a união de colunas de métricas (exclui 'instancia')
    cols = set()
    for ds in datasets:
        for m in methods:
            key = (ds, m)
            if key in dfs_dict:
                df = dfs_dict[key]
                cols |= set(df.columns)
    cols = [c for c in cols if str(c).strip().lower() != "instancia"]
    # ordena com uma ordem "humana"
    cols = sorted(cols, key=lambda x: str(x))
    return cols

def build_agg(dfs_dict, metrics, methods, datasets, use_median=True) -> pd.DataFrame:
    """
    agg: linhas=métricas, colunas=métodos
    Cada célula = mediana (ou média) inter-datasets da métrica do método.
    """
    agg = pd.DataFrame(index=metrics, columns=methods, dtype=float)
    for met in metrics:
        for method in methods:
            per_ds = []
            for ds in datasets:
                key = (ds, method)
                if key not in dfs_dict:
                    per_ds.append(np.nan)
                    continue
                per_ds.append(_reduce_metric_from_df(dfs_dict[key], met, use_median=use_median))
            per_ds = np.array(per_ds, dtype=float)
            agg.loc[met, method] = float(np.nanmedian(per_ds)) if use_median else float(np.nanmean(per_ds))
    # remove métricas totalmente vazias
    return agg.dropna(axis=0, how="all")

def build_related_sets_from_corr(corr: pd.DataFrame, thresh: float) -> dict:
    """
    Para cada métrica i: set(i) = {j | |corr(i,j)| >= thresh} U {i}
    """
    sets = {}
    for i in corr.index:
        rel = set([i])
        for j in corr.columns:
            if i == j:
                continue
            v = corr.loc[i, j]
            if np.isfinite(v) and abs(v) >= thresh:
                rel.add(j)
        sets[i] = rel
    return sets

def sets_to_indicator_df(sets: dict) -> pd.DataFrame:
    all_items = sorted(set().union(*sets.values()))
    metrics = list(sets.keys())
    mat = pd.DataFrame(False, index=all_items, columns=metrics)
    for m, elems in sets.items():
        for e in elems:
            mat.loc[e, m] = True
    return mat

# ----------------------------
# 0) Checagem: dfs existe?
# ----------------------------
if "dfs" not in globals():
    raise NameError("Não encontrei o dicionário 'dfs' no ambiente. Carregue sua estrutura dfs[(dataset,method)] primeiro.")

# ----------------------------
# 1) Descobre métricas disponíveis e monta agg
# ----------------------------
METRICS = _infer_metrics(dfs, DATASETS, METHODS)

# (opcional) remove coverage(%) se existir
METRICS = [m for m in METRICS if str(m).strip().lower() != "coverage(%)"]

agg = build_agg(dfs, METRICS, METHODS, DATASETS, use_median=USE_MEDIAN)

print("=== AGG (métrica x método) ===")
display(agg)

# ----------------------------
# 2) Spearman entre métricas (métodos como amostras)
# ----------------------------
# correlação entre linhas -> usar agg.T.corr()
corr_spearman = agg.T.corr(method="spearman", min_periods=2)

# ----------------------------
# 3) Conjuntos de sobreposição via limiar
# ----------------------------
sets = build_related_sets_from_corr(corr_spearman, THRESH)
ind = sets_to_indicator_df(sets)

# ----------------------------
# 4) UpSet plot
# ----------------------------
data = from_indicators(ind.columns.tolist(), ind.reset_index(drop=False).rename(columns={"index": "item"}))

fig = plt.figure(figsize=(15, 6))
up = UpSet(
    data,
    subset_size="count",
    show_counts=True,
    sort_by="degree",
    sort_categories_by="input",
    intersection_plot_elements=18,
)
axes = up.plot(fig=fig)

# ----------------------------
# 5) Pintar de laranja os "set sizes" das métricas destacadas
#    (sem usar up._categories — compatível com versões antigas)
# ----------------------------
ORANGE = "tab:orange"

ax_totals = axes.get("totals", None)
if ax_totals is not None:
    # A ordem das categorias está nos ticklabels do eixo Y
    ylabels = [t.get_text() for t in ax_totals.get_yticklabels()]

    # patches: 1 barra por categoria, na MESMA ordem dos ylabels (normalmente)
    patches = list(ax_totals.patches)

    # Se por alguma razão os patches vierem invertidos, tentamos alinhar:
    if len(patches) == len(ylabels):
        for lab, patch in zip(ylabels, patches):
            if lab in HIGHLIGHT_ORANGE:
                patch.set_facecolor(ORANGE)
                patch.set_edgecolor("black")
                patch.set_linewidth(0.6)

ax_matrix = axes.get("matrix", None)
if ax_matrix is not None:
    # também colore os nomes no painel da matriz
    for lab in ax_matrix.get_yticklabels():
        if lab.get_text() in HIGHLIGHT_ORANGE:
            lab.set_color(ORANGE)
            lab.set_fontweight("bold")


# ----------------------------
# 6) Diagnóstico: quais são realmente "sem sobreposição" dado o THRESH?
#    (conjunto unitário: só ela mesma)
# ----------------------------
no_overlap_true = [m for m, s in sets.items() if s == {m}]
print("\n[Métricas realmente sem sobreposição para |ρ| >= %.2f]:" % THRESH)
print(no_overlap_true)

# (opcional) imprima as métricas destacadas que de fato não sobrepõem
print("\n[Destacadas (laranja) que também estão sem sobreposição]:")
print([m for m in HIGHLIGHT_ORANGE if m in no_overlap_true])
```

```python
# ============================================================
# UPSET PLOT — a partir da tabela "Métrica" x "Relacionadas"
# (já unificada) — Random Forest
#
# Entrada: df_rel com colunas: ["Métrica", "Relacionadas"]
# - "Relacionadas" pode ter \n, múltiplos espaços, aspas, etc.
#
# Requisitos:
#   pip install upsetplot
# ============================================================

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from upsetplot import UpSet, from_indicators

# ----------------------------
# 0) COLE AQUI seu recorte (ou use df_rel já existente)
# ----------------------------
rows = [
    ("sufficiency-10(%)", "sufficiency-5(%)  sufficiency-20(%)  soundness(%)  completeness(%)  consistency(%)"),
    ("sufficiency-5(%)",  "sufficiency-10(%)  sufficiency-20(%)  soundness(%)\nstability(%) completeness(%)"),
    ("simplicity(%)",     "stability(%)"),
    ("sufficiency-20(%)", "sufficiency-10(%) \nsufficiency-5(%) \nconsistency(%)"),
    ("directional soundness(%)", "selectivity(%) robustez(%)  completeness(%)"),
    ("selectivity(%)",    "robustez(%)  directional soundness(%)  infidelity(%) soundness(%)"),
    ("infidelity(%)",     "selectivity(%)  soundness(%)  robustez(%)"),
    ("soundness(%)",      "sufficiency-5(%)  selectivity(%)  infidelity(%)  robustez(%) sufficiency-10(%)"),
    ("robustez(%)",       "selectivity(%)  \ndirectional soundness(%) \ninfidelity(%) \n soundness(%)"),
    ("consistency(%)",    "sufficiency-20(%)  sufficiency-10(%)"),
    ("stability(%)",      "simplicity(%)  completeness(%)  sufficiency-5(%)"),
    ("completeness(%)",   "stability(%) sufficiency-5(%) sufficiency-10(%)  directional soundness(%)"),
]

df_rel = pd.DataFrame(rows, columns=["Métrica", "Relacionadas"])

# (opcional) remove duplicatas exatas
df_rel = df_rel.drop_duplicates(subset=["Métrica", "Relacionadas"]).reset_index(drop=True)

# ----------------------------
# 1) Parser de "Relacionadas" -> lista limpa
# ----------------------------
def _parse_related(s: str):
    if s is None or (isinstance(s, float) and np.isnan(s)):
        return []
    s = str(s).replace('"', " ").replace("'", " ")
    s = s.replace("\n", " ").replace("\t", " ")
    # separa por 2+ espaços OU por vírgula OU por ponto e vírgula
    parts = re.split(r"(?:\s{2,}|,|;)", s)
    out = []
    for p in parts:
        p = p.strip()
        if not p:
            continue
        out.append(p)
    return out

# ----------------------------
# 2) Constrói conjuntos (cada métrica é um "set")
#    Set(m) = {m} U Relacionadas(m)
# ----------------------------
sets = {}
for _, r in df_rel.iterrows():
    m = str(r["Métrica"]).strip()
    rel = set(_parse_related(r["Relacionadas"]))
    rel.add(m)  # garante o próprio elemento
    sets[m] = rel

# universo de itens (todas as métricas citadas)
all_items = sorted(set().union(*sets.values()))

# ----------------------------
# 3) Matriz indicadora para UpSet
#    Linhas = itens; Colunas = sets (métricas)
# ----------------------------
ind = pd.DataFrame(False, index=all_items, columns=list(sets.keys()))
for m, elems in sets.items():
    for e in elems:
        if e in ind.index:
            ind.loc[e, m] = True

# ----------------------------
# 4) Constrói objeto UpSet e plota
# ----------------------------
data = from_indicators(ind.columns.tolist(), ind.reset_index(drop=False).rename(columns={"index": "item"}))

fig = plt.figure(figsize=(14, 7))
up = UpSet(
    data,
    subset_size="count",
    show_counts=True,
    sort_by="degree",
    sort_categories_by="input",
    intersection_plot_elements=18
)
axes = up.plot(fig=fig)

plt.suptitle("UpSet — Sobreposição entre métricas (Random Forest) | relações unificadas", y=1.02, fontsize=14, fontweight="bold")
plt.show()

# ----------------------------
# 5) (Opcional) Diagnóstico rápido: métricas com pouco compartilhamento
#    Aqui: conta quantos itens existem em cada set
# ----------------------------
print("\n[Tamanho de cada set = |{m} ∪ Relacionadas(m)| ]")
for m in sets:
    print(f"{m:25s} -> {len(sets[m])}")
```

```python
# ============================================================
# UPSET MULTI-PLOTS (5 por linha) + bins clínicos + coberturas
# - Mostra na tela (plt.show)
# - Também salva um PDF paginado (5 colunas por linha)
#
# Requisitos:
#   pip install pandas openpyxl matplotlib upsetplot
# ============================================================

import math
import io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from upsetplot import UpSet

# -----------------------------
# 1) CONFIG
# -----------------------------
EXCEL_PATH   = r"Path('../databases/SEPSIS/data/exams para upset SP.xlsx')"
COL_IS_PROTO = "is_proto"  # 1 = protótipo, 0 = restante

# Suas variáveis (exatamente como você listou)
VARS = [
    "Admission Weight (Kg)",
    "Anion gap",
    "BUN",
    "Calcium non-ionized",
    "Chloride (serum)",
    "Creatinine (serum)",
    "Glucose (serum)",
    "HCO3 (serum)",
    "Heart Rate",
    "Hemoglobin",
    "INR",
    "Magnesium",
    "Non Invasive Blood Pressure diastolic",
    "Non Invasive Blood Pressure systolic",
    "O2 saturation pulseoxymetry",
    "PTT",
    "Phosphorous",
    "Platelet Count",
    "Potassium (serum)",
    "Respiratory Rate",
    "Sodium (serum)",
    "SpO2 Desat Limit",
    "Temperature Fahrenheit",
    "anchor_age",
    "septic",
]

# Layout do PDF
PLOTS_PER_ROW = 4
ROWS_PER_PAGE = 7                 # 5x2 = 10 plots por página
PDF_OUT       = "upset_sepsis_5porlinha.pdf"

# -----------------------------
# 2) BINS CLÍNICOS (editáveis)
# OBS:
# - Para alguns campos (peso, idade), melhor usar quartis
# - Para "septic" é binário (0/1) e não precisa de cut; mas vamos tratar como bins "0" e "1"
# -----------------------------
CLINICAL_BINS = {
    "Anion gap": [(-np.inf, 8, "baixo"), (8, 16, "normal"), (16, np.inf, "alto")],
    "BUN": [(-np.inf, 20, "baixo/normal"), (20, 40, "moderado"), (40, np.inf, "alto")],
    "Creatinine (serum)": [(-np.inf, 1.2, "normal"), (1.2, 2.0, "moderado"), (2.0, np.inf, "alto")],
    "Glucose (serum)": [(-np.inf, 70, "hipo"), (70, 180, "alvo"), (180, np.inf, "hiper")],
    "HCO3 (serum)": [(-np.inf, 22, "baixo"), (22, 26, "normal"), (26, np.inf, "alto")],
    "Sodium (serum)": [(-np.inf, 135, "hipo"), (135, 145, "normal"), (145, np.inf, "hiper")],
    "Potassium (serum)": [(-np.inf, 3.5, "hipo"), (3.5, 5.0, "normal"), (5.0, np.inf, "hiper")],
    "Platelet Count": [(-np.inf, 150, "baixo"), (150, 450, "normal"), (450, np.inf, "alto")],
    "Heart Rate": [(-np.inf, 60, "baixo"), (60, 100, "normal"), (100, np.inf, "alto")],
    "Respiratory Rate": [(-np.inf, 12, "baixo"), (12, 20, "normal"), (20, np.inf, "alto")],
    "Temperature Fahrenheit": [(-np.inf, 96.8, "hipo"), (96.8, 100.4, "normal"), (100.4, np.inf, "febre")],
    "O2 saturation pulseoxymetry": [(-np.inf, 90, "baixo"), (90, 95, "limite"), (95, np.inf, "bom")],
    "SpO2 Desat Limit": [(-np.inf, 88, "critico"), (88, 92, "baixo"), (92, np.inf, "ok")],
    "Chloride (serum)": [(-np.inf, 98, "baixo"), (98, 106, "normal"), (106, np.inf, "alto")],
    "Hemoglobin": [(-np.inf, 10, "baixo"), (10, 17, "normal"), (17, np.inf, "alto")],
    "INR": [(-np.inf, 1.2, "normal"), (1.2, 2.0, "moderado"), (2.0, np.inf, "alto")],
    "Magnesium": [(-np.inf, 1.7, "baixo"), (1.7, 2.2, "normal"), (2.2, np.inf, "alto")],
    "PTT": [(-np.inf, 25, "baixo"), (25, 35, "normal"), (35, np.inf, "alto")],
    "Phosphorous": [(-np.inf, 2.5, "baixo"), (2.5, 4.5, "normal"), (4.5, np.inf, "alto")],
    "Non Invasive Blood Pressure systolic": [(-np.inf, 90, "hipotensao"), (90, 120, "normal"), (120, np.inf, "alto")],
    "Non Invasive Blood Pressure diastolic": [(-np.inf, 60, "baixo"), (60, 80, "normal"), (80, np.inf, "alto")],
    # Calcium non-ionized: pode variar por unidade. Se seus valores estiverem em outra escala,
    # ajuste aqui. Mantive bins genéricos com fallback por quartis se ficar estranho.
    "Calcium non-ionized": [(-np.inf, 4.5, "baixo"), (4.5, 5.6, "normal"), (5.6, np.inf, "alto")],
}

# Para variáveis sem bins clínicos definidos, usar quartis (recomendado para peso, idade, etc.)
USE_QUARTILES_IF_MISSING = True

# -----------------------------
# 3) Helpers
# -----------------------------
def make_bins(series: pd.Series, varname: str) -> pd.Series:
    """Bins clínicos (se definidos) ou quartis."""
    # septic: trata como categoria 0/1
    if varname == "septic":
        s = pd.to_numeric(series, errors="coerce")
        return s.round().astype("Int64").astype(str)

    s = pd.to_numeric(series, errors="coerce")

    if varname in CLINICAL_BINS:
        intervals = CLINICAL_BINS[varname]
        edges = [intervals[0][0]] + [b for (_, b, _) in intervals]
        labels = [lab for (_, _, lab) in intervals]
        binned = pd.cut(s, bins=edges, labels=labels, right=True, include_lowest=True)
        return binned.astype("object")

    if USE_QUARTILES_IF_MISSING:
        try:
            b = pd.qcut(s, q=4, duplicates="drop")

            # <<< AJUSTE: arredonda os limites do intervalo (ex.: 1 casa)
            DEC = 1  # mude aqui (0, 1, 2...)

            # b é Categorical com Intervalos; formatamos os limites
            cats = b.cat.categories
            new_labels = [f"({iv.left:.{DEC}f}, {iv.right:.{DEC}f}]" for iv in cats]

            return b.cat.rename_categories(new_labels).astype(str)
        except Exception:
            return pd.Series(["unbinned"] * len(s), index=s.index)

    return pd.Series(["unbinned"] * len(s), index=s.index)


def coverage_internal_formula(x_global: pd.Series, x_proto: pd.Series) -> float:
    g = pd.to_numeric(x_global, errors="coerce").dropna()
    p = pd.to_numeric(x_proto, errors="coerce").dropna()
    if len(g) == 0 or len(p) == 0:
        return np.nan
    denom = (g.max() - g.min())
    if denom == 0:
        return np.nan
    return float((p.max() - p.min()) / denom)


def build_upset_series_for_var(df: pd.DataFrame, var: str, col_is_proto: str):
    """Cria série para UpSet + métricas auxiliares."""
    bins = make_bins(df[var], var)
    tmp = df[[var, col_is_proto]].copy()
    tmp["_bin"] = bins
    tmp = tmp.dropna(subset=["_bin", col_is_proto])

    # contagens
    global_counts = tmp.groupby("_bin").size()
    proto_counts  = tmp[tmp[col_is_proto] == 1].groupby("_bin").size()

    all_bins = sorted(set(global_counts.index.astype(str)))
    global_counts = global_counts.reindex(all_bins, fill_value=0)
    proto_counts  = proto_counts.reindex(all_bins,  fill_value=0)

    # cobertura de bins por presença
    bins_global_present = (global_counts > 0).sum()
    bins_proto_present  = ((proto_counts > 0) & (global_counts > 0)).sum()
    cov_bins_presence = (bins_proto_present / bins_global_present) if bins_global_present else np.nan

    # monta interseções (limpo: apenas Global-only e Global&Proto)
    rows = []
    for b in all_bins:
        g = int(global_counts.loc[b])
        p = int(proto_counts.loc[b])

        if p > 0:
            rows.append({
                f"Global_{b}": True,
                f"Proto_{b}": True,
                "count": p
            })
        if (g - p) > 0:
            rows.append({
                f"Global_{b}": True,
                f"Proto_{b}": False,
                "count": g - p
            })

    if not rows:
        return None, all_bins, global_counts, proto_counts, cov_bins_presence

    df_rows = pd.DataFrame(rows).fillna(False)
    set_cols = [c for c in df_rows.columns if c != "count"]
    df_rows[set_cols] = df_rows[set_cols].astype(bool)
    series = df_rows.set_index(set_cols)["count"]
    return series, all_bins, global_counts, proto_counts, cov_bins_presence


def render_upset_to_image(df: pd.DataFrame, var: str, col_is_proto: str):
    """Renderiza um UpSet em imagem (PNG em memória) e retorna bytes PNG."""
    series, all_bins, gcounts, pcounts, cov_bins_presence = build_upset_series_for_var(df, var, col_is_proto)
    if series is None:
        return None

    # coberturas
    cov_internal = coverage_internal_formula(df[var], df.loc[df[col_is_proto] == 1, var])

    # figura do UpSet (uma por variável)
    fig = plt.figure(figsize=(6.2, 4.8))
    
    # ===============================
    # AJUSTE MÍNIMO DOS NÚMEROS DAS BARRAS
    # ===============================
    COUNT_FONTSIZE = 8        # <<< diminui/aumenta fonte
    COUNT_FMT = "{:,}"        # <<< formatação (milhar com vírgula)

    up = UpSet(series, sort_by="cardinality", show_counts=True, min_subset_size=1)
    axes = up.plot(fig=fig)  # retorna dict de eixos
    
    # eixo das barras superiores
    ax_bar = None
    if "axes" in locals():
        ax_bar = axes.get("intersection_size", None) or axes.get("intersections", None)

    if ax_bar is not None:
        for txt in ax_bar.texts:
            try:
                val = int(float(txt.get_text()))
                txt.set_text(COUNT_FMT.format(val))
                txt.set_fontsize(COUNT_FONTSIZE)
            except Exception:
                pass





    fig.suptitle(f"UpSet — {var} (Global vs Protótipos)", y=1.02)

    # escolhe um eixo "bom" para anotação: o da matriz normalmente tem espaço livre
    # fallback: pega o primeiro eixo se não existir
    ax_anno = axes.get("matrix", None)
    if ax_anno is None:
        ax_anno = fig.axes[0]

    txt = []
    if not np.isnan(cov_bins_presence):
        txt.append(f"Cobertura_bins (presença): {cov_bins_presence*100:.1f}%")
    else:
        txt.append("Cobertura_bins (presença): n/a")

    if not np.isnan(cov_internal):
        txt.append(f"Cobertura_interna (min/max): {cov_internal*100:.1f}%")
    else:
        txt.append("Cobertura_interna (min/max): n/a")

    # anotação pequena, no canto superior esquerdo, sem ocupar “faixa”
    ax_anno.text(
        1.15, -0.09, 
        "\n".join(txt),
        transform=ax_anno.transAxes,
        va="top", ha="left",
        fontsize=8,
        bbox=dict(boxstyle="round,pad=0.15", facecolor="white", alpha=0.75, edgecolor="none")
    )

    fig.tight_layout()

    # salva em memória
    buf = io.BytesIO()
    fig.savefig(buf, format="png", dpi=180, bbox_inches="tight")
    plt.close(fig)
    buf.seek(0)
    return buf.getvalue()


# -----------------------------
# 4) Carrega dados
# -----------------------------
df = pd.read_excel(EXCEL_PATH, engine="openpyxl")

if COL_IS_PROTO not in df.columns:
    raise ValueError(f"Coluna '{COL_IS_PROTO}' não encontrada. Colunas: {list(df.columns)}")

df[COL_IS_PROTO] = pd.to_numeric(df[COL_IS_PROTO], errors="coerce").fillna(0).astype(int)

# mantém só variáveis existentes
vars_ok = [v for v in VARS if v in df.columns]
missing = [v for v in VARS if v not in df.columns]
print(f"[OK] Variáveis encontradas: {len(vars_ok)}")
if missing:
    print("[ATENÇÃO] Variáveis NÃO encontradas no Excel:")
    for m in missing:
        print(" -", m)

if not vars_ok:
    raise ValueError("Nenhuma variável da lista foi encontrada. Ajuste os nomes conforme o Excel.")

# -----------------------------
# 5) MOSTRA NA TELA (plt.show) — uma por variável
# -----------------------------
for var in vars_ok:
    series, all_bins, gcounts, pcounts, cov_bins_presence = build_upset_series_for_var(df, var, COL_IS_PROTO)
    if series is None:
        print(f"[SKIP] {var}: sem dados.")
        continue

    cov_internal = coverage_internal_formula(df[var], df.loc[df[COL_IS_PROTO] == 1, var])

    fig = plt.figure(figsize=(6.2, 4.8))
    up = UpSet(series, sort_by="cardinality", show_counts=True, min_subset_size=1)
    axes = up.plot(fig=fig)
    # ===============================
    # AJUSTE DOS NÚMEROS DAS BARRAS (topo)
    # ===============================
    COUNT_FONTSIZE = 8
    COUNT_ROT = 90
    COUNT_FMT = "{:,}"

    ax_bar = axes.get("intersection_size", None) or axes.get("intersections", None)

    if ax_bar is not None:
        for txt in ax_bar.texts:
            try:
                val = int(float(txt.get_text()))
                txt.set_text(COUNT_FMT.format(val))
                txt.set_fontsize(COUNT_FONTSIZE)
                txt.set_rotation(COUNT_ROT)
                txt.set_ha("left")
                txt.set_va("bottom")

                # <<< AFASTAMENTO FINO
                x, y = txt.get_position()
                txt.set_position((x, y + 250))

            except Exception:
                pass




    # ===============================
    # AJUSTE MÍNIMO DOS NÚMEROS DAS BARRAS
    # ===============================
    COUNT_FONTSIZE = 8        # <<< diminui/aumenta fonte
    COUNT_FMT = "{:,}"        # <<< formatação (milhar com vírgula)

    # eixo das barras superiores
    ax_bar = axes.get("intersection_size", None) or axes.get("intersections", None)

    if ax_bar is not None:
        for txt in ax_bar.texts:
            try:
                val = int(float(txt.get_text()))
                txt.set_text(COUNT_FMT.format(val))
                txt.set_fontsize(COUNT_FONTSIZE)
            except Exception:
                pass
            
    fig.suptitle(f"UpSet — {var} (Global vs Protótipos)", y=1.02)

    # <<< USE O EIXO DE BARRAS (TOPO) PRA TER ESPAÇO LIVRE
    ax_anno = axes.get("intersections", None) or axes.get("intersection_size", None) or fig.axes[0]
    txt = []
    txt.append(f"Cobertura_bins (presença): {cov_bins_presence*100:.1f}%" if not np.isnan(cov_bins_presence) else "Cobertura_bins (presença): n/a")
    txt.append(f"Cobertura_interna (min/max): {cov_internal*100:.1f}%" if not np.isnan(cov_internal) else "Cobertura_interna (min/max): n/a")

    # ===============================
    # ANOTAÇÃO DE COBERTURA (canto inferior direito)
    # ===============================
    ax_anno = axes.get("matrix", None) or fig.axes[0]

    ANN_FONTSIZE = 8
    X_ANN, Y_ANN = 1.15, -0.09       # <<< canto inferior direito
    HA, VA = "right", "bottom"      # <<< ancora

    ax_anno.text(
        X_ANN, Y_ANN,
        "\n".join(txt),
        transform=ax_anno.transAxes,
        va=VA, ha=HA,
        fontsize=ANN_FONTSIZE,
        # <<< se você NÃO quiser caixinha, apague a linha bbox abaixo
        bbox=dict(boxstyle="round,pad=0.15", facecolor="white", alpha=0.85, edgecolor="none")
    )


    fig.tight_layout()
    plt.show()

# -----------------------------
# 6) PDF paginado (5 por linha)
# -----------------------------
images = []
for var in vars_ok:
    img = render_upset_to_image(df, var, COL_IS_PROTO)
    if img is not None:
        images.append((var, img))

plots_per_page = PLOTS_PER_ROW * ROWS_PER_PAGE
n_pages = math.ceil(len(images) / plots_per_page)

with PdfPages(PDF_OUT) as pdf:
    for p in range(n_pages):
        start = p * plots_per_page
        end = min((p+1) * plots_per_page, len(images))
        chunk = images[start:end]

        # >>>>>> AQUI ESTAVA FALTANDO <<<<<<
        fig, axes = plt.subplots(
            ROWS_PER_PAGE, PLOTS_PER_ROW,
            figsize=(PLOTS_PER_ROW*3.0, ROWS_PER_PAGE*2.6)
        )

        # <<< AJUSTE DE ESPAÇAMENTO NO PDF (agora no lugar certo)
        HSPACE = 0.06
        WSPACE = 0.03
        TOP    = 0.90
        BOTTOM = 0.03
        LEFT   = 0.02
        RIGHT  = 0.98

        fig.subplots_adjust(
            left=LEFT, right=RIGHT,
            top=TOP, bottom=BOTTOM,
            wspace=WSPACE, hspace=HSPACE
        )

        axes = np.array(axes).reshape(-1)

        # limpa
        for ax in axes:
            ax.axis("off")

        # preenche com imagens
        for ax, (var, img_bytes) in zip(axes, chunk):
            im = plt.imread(io.BytesIO(img_bytes))
            ax.imshow(im)
            ax.axis("off")

        fig.suptitle(
            f"UpSet — Cobertura (Global vs Protótipos) — Página {p+1}/{n_pages}",
            y=0.99, fontsize=12
        )

        pdf.savefig(fig)
        plt.close(fig)

print(f"[OK] PDF salvo em: {PDF_OUT}")
```

```python
# ============================================================
# UPSET MULTI-PLOTS (5 por linha) + bins clínicos + coberturas
# - Mostra na tela (plt.show)
# - Também salva um PDF paginado (5 colunas por linha)
#
# Requisitos:
#   pip install pandas openpyxl matplotlib upsetplot
# ============================================================

import math
import io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from upsetplot import UpSet

# -----------------------------
# 1) CONFIG
# -----------------------------
EXCEL_PATH   = r"Path('../databases/SEPSIS/data/exams para upset.xlsx')"
COL_IS_PROTO = "is_proto"  # 1 = protótipo, 0 = restante

# Suas variáveis (exatamente como você listou)
VARS = [
    "Admission Weight (Kg)",
    "Anion gap",
    "BUN",
    "Calcium non-ionized",
    "Chloride (serum)",
    "Creatinine (serum)",
    "Glucose (serum)",
    "HCO3 (serum)",
    "Heart Rate",
    "Hemoglobin",
    "INR",
    "Magnesium",
    "Non Invasive Blood Pressure diastolic",
    "Non Invasive Blood Pressure systolic",
    "O2 saturation pulseoxymetry",
    "PTT",
    "Phosphorous",
    "Platelet Count",
    "Potassium (serum)",
    "Respiratory Rate",
    "Sodium (serum)",
    "SpO2 Desat Limit",
    "Temperature Fahrenheit",
    "anchor_age",
    "septic",
]

# Layout do PDF
PLOTS_PER_ROW = 4
ROWS_PER_PAGE = 7                 # 5x2 = 10 plots por página
PDF_OUT       = "upset_sepsis_5porlinha.pdf"

# -----------------------------
# 2) BINS CLÍNICOS (editáveis)
# OBS:
# - Para alguns campos (peso, idade), melhor usar quartis
# - Para "septic" é binário (0/1) e não precisa de cut; mas vamos tratar como bins "0" e "1"
# -----------------------------
CLINICAL_BINS = {
    "Anion gap": [(-np.inf, 8, "baixo"), (8, 16, "normal"), (16, np.inf, "alto")],
    "BUN": [(-np.inf, 20, "baixo/normal"), (20, 40, "moderado"), (40, np.inf, "alto")],
    "Creatinine (serum)": [(-np.inf, 1.2, "normal"), (1.2, 2.0, "moderado"), (2.0, np.inf, "alto")],
    "Glucose (serum)": [(-np.inf, 70, "hipo"), (70, 180, "alvo"), (180, np.inf, "hiper")],
    "HCO3 (serum)": [(-np.inf, 22, "baixo"), (22, 26, "normal"), (26, np.inf, "alto")],
    "Sodium (serum)": [(-np.inf, 135, "hipo"), (135, 145, "normal"), (145, np.inf, "hiper")],
    "Potassium (serum)": [(-np.inf, 3.5, "hipo"), (3.5, 5.0, "normal"), (5.0, np.inf, "hiper")],
    "Platelet Count": [(-np.inf, 150, "baixo"), (150, 450, "normal"), (450, np.inf, "alto")],
    "Heart Rate": [(-np.inf, 60, "baixo"), (60, 100, "normal"), (100, np.inf, "alto")],
    "Respiratory Rate": [(-np.inf, 12, "baixo"), (12, 20, "normal"), (20, np.inf, "alto")],
    "Temperature Fahrenheit": [(-np.inf, 96.8, "hipo"), (96.8, 100.4, "normal"), (100.4, np.inf, "febre")],
    "O2 saturation pulseoxymetry": [(-np.inf, 90, "baixo"), (90, 95, "limite"), (95, np.inf, "bom")],
    "SpO2 Desat Limit": [(-np.inf, 88, "critico"), (88, 92, "baixo"), (92, np.inf, "ok")],
    "Chloride (serum)": [(-np.inf, 98, "baixo"), (98, 106, "normal"), (106, np.inf, "alto")],
    "Hemoglobin": [(-np.inf, 10, "baixo"), (10, 17, "normal"), (17, np.inf, "alto")],
    "INR": [(-np.inf, 1.2, "normal"), (1.2, 2.0, "moderado"), (2.0, np.inf, "alto")],
    "Magnesium": [(-np.inf, 1.7, "baixo"), (1.7, 2.2, "normal"), (2.2, np.inf, "alto")],
    "PTT": [(-np.inf, 25, "baixo"), (25, 35, "normal"), (35, np.inf, "alto")],
    "Phosphorous": [(-np.inf, 2.5, "baixo"), (2.5, 4.5, "normal"), (4.5, np.inf, "alto")],
    "Non Invasive Blood Pressure systolic": [(-np.inf, 90, "hipotensao"), (90, 120, "normal"), (120, np.inf, "alto")],
    "Non Invasive Blood Pressure diastolic": [(-np.inf, 60, "baixo"), (60, 80, "normal"), (80, np.inf, "alto")],
    # Calcium non-ionized: pode variar por unidade. Se seus valores estiverem em outra escala,
    # ajuste aqui. Mantive bins genéricos com fallback por quartis se ficar estranho.
    "Calcium non-ionized": [(-np.inf, 4.5, "baixo"), (4.5, 5.6, "normal"), (5.6, np.inf, "alto")],
}

# Para variáveis sem bins clínicos definidos, usar quartis (recomendado para peso, idade, etc.)
USE_QUARTILES_IF_MISSING = True

# -----------------------------
# 3) Helpers
# -----------------------------
def make_bins(series: pd.Series, varname: str) -> pd.Series:
    """Bins clínicos (se definidos) ou quartis."""
    # septic: trata como categoria 0/1
    if varname == "septic":
        s = pd.to_numeric(series, errors="coerce")
        return s.round().astype("Int64").astype(str)

    s = pd.to_numeric(series, errors="coerce")

    if varname in CLINICAL_BINS:
        intervals = CLINICAL_BINS[varname]
        edges = [intervals[0][0]] + [b for (_, b, _) in intervals]
        labels = [lab for (_, _, lab) in intervals]
        binned = pd.cut(s, bins=edges, labels=labels, right=True, include_lowest=True)
        return binned.astype("object")

    if USE_QUARTILES_IF_MISSING:
        try:
            b = pd.qcut(s, q=4, duplicates="drop")

            # <<< AJUSTE: arredonda os limites do intervalo (ex.: 1 casa)
            DEC = 1  # mude aqui (0, 1, 2...)

            # b é Categorical com Intervalos; formatamos os limites
            cats = b.cat.categories
            new_labels = [f"({iv.left:.{DEC}f}, {iv.right:.{DEC}f}]" for iv in cats]

            return b.cat.rename_categories(new_labels).astype(str)
        except Exception:
            return pd.Series(["unbinned"] * len(s), index=s.index)

    return pd.Series(["unbinned"] * len(s), index=s.index)


def coverage_internal_formula(x_global: pd.Series, x_proto: pd.Series) -> float:
    g = pd.to_numeric(x_global, errors="coerce").dropna()
    p = pd.to_numeric(x_proto, errors="coerce").dropna()
    if len(g) == 0 or len(p) == 0:
        return np.nan
    denom = (g.max() - g.min())
    if denom == 0:
        return np.nan
    return float((p.max() - p.min()) / denom)


def build_upset_series_for_var(df: pd.DataFrame, var: str, col_is_proto: str):
    """Cria série para UpSet + métricas auxiliares."""
    bins = make_bins(df[var], var)
    tmp = df[[var, col_is_proto]].copy()
    tmp["_bin"] = bins
    tmp = tmp.dropna(subset=["_bin", col_is_proto])

    # contagens
    global_counts = tmp.groupby("_bin").size()
    proto_counts  = tmp[tmp[col_is_proto] == 1].groupby("_bin").size()

    all_bins = sorted(set(global_counts.index.astype(str)))
    global_counts = global_counts.reindex(all_bins, fill_value=0)
    proto_counts  = proto_counts.reindex(all_bins,  fill_value=0)

    # cobertura de bins por presença
    bins_global_present = (global_counts > 0).sum()
    bins_proto_present  = ((proto_counts > 0) & (global_counts > 0)).sum()
    cov_bins_presence = (bins_proto_present / bins_global_present) if bins_global_present else np.nan

    # monta interseções (limpo: apenas Global-only e Global&Proto)
    rows = []
    for b in all_bins:
        g = int(global_counts.loc[b])
        p = int(proto_counts.loc[b])

        if p > 0:
            rows.append({
                f"Global_{b}": True,
                f"Proto_{b}": True,
                "count": p
            })
        if (g - p) > 0:
            rows.append({
                f"Global_{b}": True,
                f"Proto_{b}": False,
                "count": g - p
            })

    if not rows:
        return None, all_bins, global_counts, proto_counts, cov_bins_presence

    df_rows = pd.DataFrame(rows).fillna(False)
    set_cols = [c for c in df_rows.columns if c != "count"]
    df_rows[set_cols] = df_rows[set_cols].astype(bool)
    series = df_rows.set_index(set_cols)["count"]
    return series, all_bins, global_counts, proto_counts, cov_bins_presence


def render_upset_to_image(df: pd.DataFrame, var: str, col_is_proto: str):
    """Renderiza um UpSet em imagem (PNG em memória) e retorna bytes PNG."""
    series, all_bins, gcounts, pcounts, cov_bins_presence = build_upset_series_for_var(df, var, col_is_proto)
    if series is None:
        return None

    # coberturas
    cov_internal = coverage_internal_formula(df[var], df.loc[df[col_is_proto] == 1, var])

    # figura do UpSet (uma por variável)
    fig = plt.figure(figsize=(6.2, 4.8))
    
    # ===============================
    # AJUSTE MÍNIMO DOS NÚMEROS DAS BARRAS
    # ===============================
    COUNT_FONTSIZE = 8        # <<< diminui/aumenta fonte
    COUNT_FMT = "{:,}"        # <<< formatação (milhar com vírgula)

    up = UpSet(series, sort_by="cardinality", show_counts=True, min_subset_size=1)
    axes = up.plot(fig=fig)  # retorna dict de eixos
    
    # eixo das barras superiores
    ax_bar = None
    if "axes" in locals():
        ax_bar = axes.get("intersection_size", None) or axes.get("intersections", None)

    if ax_bar is not None:
        for txt in ax_bar.texts:
            try:
                val = int(float(txt.get_text()))
                txt.set_text(COUNT_FMT.format(val))
                txt.set_fontsize(COUNT_FONTSIZE)
            except Exception:
                pass





    fig.suptitle(f"UpSet — {var} (Global vs Protótipos)", y=1.02)

    # escolhe um eixo "bom" para anotação: o da matriz normalmente tem espaço livre
    # fallback: pega o primeiro eixo se não existir
    ax_anno = axes.get("matrix", None)
    if ax_anno is None:
        ax_anno = fig.axes[0]

    txt = []
    if not np.isnan(cov_bins_presence):
        txt.append(f"Cobertura_bins (presença): {cov_bins_presence*100:.1f}%")
    else:
        txt.append("Cobertura_bins (presença): n/a")

    if not np.isnan(cov_internal):
        txt.append(f"Cobertura_interna (min/max): {cov_internal*100:.1f}%")
    else:
        txt.append("Cobertura_interna (min/max): n/a")

    # anotação pequena, no canto superior esquerdo, sem ocupar “faixa”
    ax_anno.text(
        1.15, -0.09, 
        "\n".join(txt),
        transform=ax_anno.transAxes,
        va="top", ha="left",
        fontsize=8,
        bbox=dict(boxstyle="round,pad=0.15", facecolor="white", alpha=0.75, edgecolor="none")
    )

    fig.tight_layout()

    # salva em memória
    buf = io.BytesIO()
    fig.savefig(buf, format="png", dpi=180, bbox_inches="tight")
    plt.close(fig)
    buf.seek(0)
    return buf.getvalue()


# -----------------------------
# 4) Carrega dados
# -----------------------------
df = pd.read_excel(EXCEL_PATH, engine="openpyxl")

if COL_IS_PROTO not in df.columns:
    raise ValueError(f"Coluna '{COL_IS_PROTO}' não encontrada. Colunas: {list(df.columns)}")

df[COL_IS_PROTO] = pd.to_numeric(df[COL_IS_PROTO], errors="coerce").fillna(0).astype(int)

# mantém só variáveis existentes
vars_ok = [v for v in VARS if v in df.columns]
missing = [v for v in VARS if v not in df.columns]
print(f"[OK] Variáveis encontradas: {len(vars_ok)}")
if missing:
    print("[ATENÇÃO] Variáveis NÃO encontradas no Excel:")
    for m in missing:
        print(" -", m)

if not vars_ok:
    raise ValueError("Nenhuma variável da lista foi encontrada. Ajuste os nomes conforme o Excel.")

# -----------------------------
# 5) MOSTRA NA TELA (plt.show) — uma por variável
# -----------------------------
for var in vars_ok:
    series, all_bins, gcounts, pcounts, cov_bins_presence = build_upset_series_for_var(df, var, COL_IS_PROTO)
    if series is None:
        print(f"[SKIP] {var}: sem dados.")
        continue

    cov_internal = coverage_internal_formula(df[var], df.loc[df[COL_IS_PROTO] == 1, var])

    fig = plt.figure(figsize=(6.2, 4.8))
    up = UpSet(series, sort_by="cardinality", show_counts=True, min_subset_size=1)
    axes = up.plot(fig=fig)
    # ===============================
    # AJUSTE DOS NÚMEROS DAS BARRAS (topo)
    # ===============================
    COUNT_FONTSIZE = 8
    COUNT_ROT = 90
    COUNT_FMT = "{:,}"

    ax_bar = axes.get("intersection_size", None) or axes.get("intersections", None)

    if ax_bar is not None:
        for txt in ax_bar.texts:
            try:
                val = int(float(txt.get_text()))
                txt.set_text(COUNT_FMT.format(val))
                txt.set_fontsize(COUNT_FONTSIZE)
                txt.set_rotation(COUNT_ROT)
                txt.set_ha("left")
                txt.set_va("bottom")

                # <<< AFASTAMENTO FINO
                x, y = txt.get_position()
                txt.set_position((x, y + 250))

            except Exception:
                pass




    # ===============================
    # AJUSTE MÍNIMO DOS NÚMEROS DAS BARRAS
    # ===============================
    COUNT_FONTSIZE = 8        # <<< diminui/aumenta fonte
    COUNT_FMT = "{:,}"        # <<< formatação (milhar com vírgula)

    # eixo das barras superiores
    ax_bar = axes.get("intersection_size", None) or axes.get("intersections", None)

    if ax_bar is not None:
        for txt in ax_bar.texts:
            try:
                val = int(float(txt.get_text()))
                txt.set_text(COUNT_FMT.format(val))
                txt.set_fontsize(COUNT_FONTSIZE)
            except Exception:
                pass
            
    fig.suptitle(f"UpSet — {var} (Global vs Protótipos)", y=1.02)

    # <<< USE O EIXO DE BARRAS (TOPO) PRA TER ESPAÇO LIVRE
    ax_anno = axes.get("intersections", None) or axes.get("intersection_size", None) or fig.axes[0]
    txt = []
    txt.append(f"Cobertura_bins (presença): {cov_bins_presence*100:.1f}%" if not np.isnan(cov_bins_presence) else "Cobertura_bins (presença): n/a")
    txt.append(f"Cobertura_interna (min/max): {cov_internal*100:.1f}%" if not np.isnan(cov_internal) else "Cobertura_interna (min/max): n/a")

    # ===============================
    # ANOTAÇÃO DE COBERTURA (canto inferior direito)
    # ===============================
    ax_anno = axes.get("matrix", None) or fig.axes[0]

    ANN_FONTSIZE = 8
    X_ANN, Y_ANN = 1.15, -0.09       # <<< canto inferior direito
    HA, VA = "right", "bottom"      # <<< ancora

    ax_anno.text(
        X_ANN, Y_ANN,
        "\n".join(txt),
        transform=ax_anno.transAxes,
        va=VA, ha=HA,
        fontsize=ANN_FONTSIZE,
        # <<< se você NÃO quiser caixinha, apague a linha bbox abaixo
        bbox=dict(boxstyle="round,pad=0.15", facecolor="white", alpha=0.85, edgecolor="none")
    )


    fig.tight_layout()
    plt.show()

# -----------------------------
# 6) PDF paginado (5 por linha)
# -----------------------------
images = []
for var in vars_ok:
    img = render_upset_to_image(df, var, COL_IS_PROTO)
    if img is not None:
        images.append((var, img))

plots_per_page = PLOTS_PER_ROW * ROWS_PER_PAGE
n_pages = math.ceil(len(images) / plots_per_page)

with PdfPages(PDF_OUT) as pdf:
    for p in range(n_pages):
        start = p * plots_per_page
        end = min((p+1) * plots_per_page, len(images))
        chunk = images[start:end]

        # >>>>>> AQUI ESTAVA FALTANDO <<<<<<
        fig, axes = plt.subplots(
            ROWS_PER_PAGE, PLOTS_PER_ROW,
            figsize=(PLOTS_PER_ROW*3.0, ROWS_PER_PAGE*2.6)
        )

        # <<< AJUSTE DE ESPAÇAMENTO NO PDF (agora no lugar certo)
        HSPACE = 0.06
        WSPACE = 0.03
        TOP    = 0.90
        BOTTOM = 0.03
        LEFT   = 0.02
        RIGHT  = 0.98

        fig.subplots_adjust(
            left=LEFT, right=RIGHT,
            top=TOP, bottom=BOTTOM,
            wspace=WSPACE, hspace=HSPACE
        )

        axes = np.array(axes).reshape(-1)

        # limpa
        for ax in axes:
            ax.axis("off")

        # preenche com imagens
        for ax, (var, img_bytes) in zip(axes, chunk):
            im = plt.imread(io.BytesIO(img_bytes))
            ax.imshow(im)
            ax.axis("off")

        fig.suptitle(
            f"UpSet — Cobertura (Global vs Protótipos) — Página {p+1}/{n_pages}",
            y=0.99, fontsize=12
        )

        pdf.savefig(fig)
        plt.close(fig)

print(f"[OK] PDF salvo em: {PDF_OUT}")
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# ============================================================
# TABELA (LAYOUT “BLOCO POR MÉTODO” IGUAL AO ANEXO)
# - Linhas: métricas
# - Colunas: NOME REAL DOS DATASETS (inferido do dict dfs) + Média + Mediana + IC 95%
# - Cada célula contém:
#     An: 0.xx
#     Li: 0.xx
#     PF: 0.xx
#     Sh: 0.xx
#     Su: 0.xx
# - Dados vêm de: dfs[(dataset, metodo)]  (como você já usa)
# ============================================================

# ---------- AJUSTE SE NECESSÁRIO ----------
# Se no seu df as métricas estão em % (0..100), deixe False (vai dividir por 100).
# Se já estão em [0,1], coloque True.
VALUES_ARE_0_1 = False

# Métodos e abreviações (iguais ao seu padrão do anexo)
METHODS = ["ANCHOR", "LIME", "PFI", "SHAP", "SURROGATE"]
METHOD_ABBR = {
    "ANCHOR": "An",
    "LIME": "Li",
    "PFI": "PF",
    "SHAP": "Sh",
    "SURROGATE": "Su",
}

# Se seus dfs usam nomes “bonitos” no índice/label, mapeie aqui (opcional)
DISPLAY_NAME = {
    "ANCHOR": "ANCHOR",
    "LIME": "LIME",
    "PFI": "PFI",
    "SHAP": "SHAP (Tree)",     # se no seu projeto SHAP vira "SHAP (Tree)"
    "SURROGATE": "Surrogate",
}

# Se a chave do dfs for ("SEPSIS", "SHAP") mas você quer exibir "SHAP (Tree)",
# isso NÃO muda a chave; só muda o texto na célula.
# -----------------------------------------


# ============================================================
# Helpers
# ============================================================

def _coerce_num(s: pd.Series) -> pd.Series:
    return pd.to_numeric(s, errors="coerce")

def _to_0_1(x):
    x = pd.to_numeric(x, errors="coerce")
    if VALUES_ARE_0_1:
        y = x
    else:
        y = x / 100.0
    return np.clip(y, 0, 1)

def get_median_row_from_df(df: pd.DataFrame, metrics: list[str]) -> pd.Series:
    """
    Extrai a linha de MEDIANA do df, seguindo sua lógica:
    1) coluna "estatistica"/"stat"/"statistic" com valor median/mediana
    2) índice com median/mediana
    3) fallback: mediana por coluna (numérica)
    """
    cols_ok = [m for m in metrics if m in df.columns]
    if not cols_ok:
        raise ValueError(f"Nenhuma das métricas {metrics} foi encontrada. Colunas: {list(df.columns)[:60]}")

    stat_cols = [c for c in df.columns if str(c).strip().lower() in {"estatistica", "stat", "statistic"}]
    if stat_cols:
        c = stat_cols[0]
        mask = df[c].astype(str).str.strip().str.lower().isin({"median", "mediana"})
        if mask.any():
            row = df.loc[mask, cols_ok].iloc[0]
            return row.reindex(metrics)

    idx = df.index.astype(str).str.strip().str.lower()
    hit = np.where(np.isin(idx, ["median", "mediana"]))[0]
    if len(hit) > 0:
        row = df.iloc[int(hit[0])][cols_ok]
        return row.reindex(metrics)

    row = df[cols_ok].apply(_coerce_num).median(axis=0, skipna=True)
    return row.reindex(metrics)

def format_block(method_to_value: dict, decimals=2) -> str:
    """
    Formata célula no estilo do anexo:
      An: 0.71
      Li: 0.68
      ...
    """
    lines = []
    for m in METHODS:
        ab = METHOD_ABBR[m]
        v = method_to_value.get(m, np.nan)
        if pd.isna(v):
            lines.append(f"{ab}: -")
        else:
            lines.append(f"{ab}: {v:.{decimals}f}")
    return "\n".join(lines)

def compute_ci95(values: np.ndarray, conf=0.95):
    """IC normal approx (igual ao seu uso simples)."""
    values = np.asarray(values, dtype=float)
    values = values[np.isfinite(values)]
    if values.size == 0:
        return (np.nan, np.nan)
    z = stats.norm.ppf(1 - (1 - conf) / 2)
    mu = values.mean()
    sd = values.std(ddof=1) if values.size > 1 else 0.0
    half = z * sd / np.sqrt(values.size) if values.size > 1 else 0.0
    return (mu - half, mu + half)

# ============================================================
# 1) Inferir DATASETS reais a partir das chaves do dict dfs
# ============================================================
# Esperado: dfs[(dataset, metodo)] = DataFrame
datasets = sorted({k[0] for k in dfs.keys() if isinstance(k, tuple) and len(k) == 2})
if not datasets:
    raise ValueError("Não consegui inferir datasets. Verifique se dfs tem chaves no formato (dataset, metodo).")

# ============================================================
# 2) Inferir lista de MÉTRICAS a partir do primeiro df encontrado
#    (ou defina manualmente se preferir)
# ============================================================
# pega um df qualquer existente
_any_key = next(iter(dfs.keys()))
_any_df = dfs[_any_key]

# tenta filtrar colunas que parecem métricas (terminam com (%) ou são numéricas)
# se você já tem lista fixa, substitua METRICS abaixo pela sua lista.
candidate_cols = []
for c in _any_df.columns:
    cs = str(c)
    if cs.endswith("(%)"):
        candidate_cols.append(cs)

# fallback: se não achar "(%)", pega colunas numéricas (exclui colunas de texto comuns)
if not candidate_cols:
    numeric_cols = _any_df.select_dtypes(include=[np.number]).columns.tolist()
    # remove colunas de controle comuns
    drop_like = {"fold", "etapa", "stat", "statistic", "estatistica", "dataset", "metodo", "method"}
    candidate_cols = [c for c in numeric_cols if str(c).strip().lower() not in drop_like]

METRICS = candidate_cols
if not METRICS:
    raise ValueError("Não consegui inferir métricas. Defina manualmente METRICS.")

print("Datasets inferidos:", datasets)
print("Métricas inferidas (amostra):", METRICS[:12], ("..." if len(METRICS) > 12 else ""))

# ============================================================
# 3) Construir matriz base: por (dataset, method) -> vetor de medianas [0,1]
# ============================================================
# med_01[(dataset, method)] = Series(metrics->value_01)
med_01 = {}

missing_pairs = []
for d in datasets:
    for m in METHODS:
        key = (d, m)
        if key not in dfs:
            missing_pairs.append(key)
            continue
        dfm = dfs[key]
        med_row = get_median_row_from_df(dfm, METRICS)
        med_row_01 = med_row.apply(_to_0_1)  # normaliza [0,1]
        med_01[(d, m)] = med_row_01

if not med_01:
    raise ValueError("Nenhum par (dataset, método) encontrado em dfs para METHODS informados.")

if missing_pairs:
    print("\n[ATENÇÃO] Pares (dataset, método) ausentes em dfs (exibindo até 20):")
    print(missing_pairs[:20])

# ============================================================
# 4) Gerar DataFrame final no LAYOUT do anexo
# ============================================================
rows = []

for metric in METRICS:
    row = {"Métrica": metric.replace("(%)", "")}  # no anexo aparece sem "(%)" às vezes

    # ---- Colunas por dataset: bloco com An/Li/PF/Sh/Su ----
    for d in datasets:
        block_vals = {}
        for m in METHODS:
            s = med_01.get((d, m))
            block_vals[m] = float(s.get(metric, np.nan)) if s is not None else np.nan
        row[d] = format_block(block_vals, decimals=2)

    # ---- Agregados por coluna extra:
    # Média / Mediana: por método, ao longo dos datasets
    mean_by_method = {}
    median_by_method = {}
    ci_by_method = {}

    for m in METHODS:
        vals = []
        for d in datasets:
            s = med_01.get((d, m))
            if s is not None:
                vals.append(float(s.get(metric, np.nan)))
        vals = np.array(vals, dtype=float)

        mean_by_method[m] = np.nanmean(vals) if np.isfinite(vals).any() else np.nan
        median_by_method[m] = np.nanmedian(vals) if np.isfinite(vals).any() else np.nan
        lo, hi = compute_ci95(vals, conf=0.95)
        ci_by_method[m] = f"[{lo:.2f},{hi:.2f}]" if np.isfinite(lo) and np.isfinite(hi) else "[-,-]"

    row["Média"] = format_block(mean_by_method, decimals=2)
    row["Mediana"] = format_block(median_by_method, decimals=2)

    # IC 95% é texto, então formatamos diferente mantendo a ordem dos métodos
    ci_lines = []
    for m in METHODS:
        ci_lines.append(f"{METHOD_ABBR[m]}: {ci_by_method[m]}")
    row["IC 95%"] = "\n".join(ci_lines)

    rows.append(row)

df_layout = pd.DataFrame(rows)

# ============================================================
# 5) Exportar CSV
# ============================================================
out_csv = "tabela_layout_multidataset_por_metricas.csv"
df_layout.to_csv(out_csv, index=False, encoding="utf-8-sig")

print(f"\nOK! DataFrame gerado com {df_layout.shape[0]} métricas x {df_layout.shape[1]} colunas.")
print("CSV salvo em:", out_csv)

# Se quiser ver no console (pode ser grande):
# print(df_layout.to_string(index=False))
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# ============================================================
# CONFIGURAÇÕES
# ============================================================

DECIMALS = 4
VALUES_ARE_0_1 = False   # métricas em %

METHODS = ["LIME", "SHAP", "PFI", "ANCHOR", "SURROGATE"]

DISPLAY_METHOD = {
    "LIME": "LIME",
    "SHAP": "SHAP (Tree)",
    "PFI": "PFI",
    "ANCHOR": "ANCHOR",
    "SURROGATE": "Surrogate",
}

# >>> LISTA FIXA DAS SUAS MÉTRICAS <<<
METRICS = [
    "fidelity(%)",
    "infidelity(%)",
    "faithfulness(%)",
    "simplicity(%)",
    "consistency(%)",
    "robustez(%)",
    "completeness(%)",
    "selectivity(%)",
    "soundness(%)",
    "directional soundness(%)",
    "stability(%)",
    "sufficiency-1(%)",
    "sufficiency-5(%)",
    "sufficiency-10(%)",
    "sufficiency-20(%)",
    "coverage(%)",
]

CONF_LEVEL = 0.95
Z = stats.norm.ppf(1 - (1 - CONF_LEVEL) / 2)

# ============================================================
# HELPERS
# ============================================================

def _to_0_1(x):
    x = pd.to_numeric(x, errors="coerce")
    return np.clip(x / 100.0, 0, 1)

def get_median_row(df, metrics):
    cols = [m for m in metrics if m in df.columns]
    if not cols:
        raise ValueError(f"Métricas não encontradas no df. Colunas: {list(df.columns)[:40]}")

    stat_cols = [c for c in df.columns if str(c).lower() in {"estatistica","stat","statistic"}]
    if stat_cols:
        sc = stat_cols[0]
        mask = df[sc].astype(str).str.lower().isin({"median","mediana"})
        if mask.any():
            return df.loc[mask, cols].iloc[0]

    idx = df.index.astype(str).str.lower()
    if idx.isin(["median","mediana"]).any():
        return df.loc[idx.isin(["median","mediana"]), cols].iloc[0]

    return df[cols].apply(pd.to_numeric, errors="coerce").median(axis=0)

def ci95(values):
    values = np.asarray(values, dtype=float)
    values = values[np.isfinite(values)]
    if values.size < 2:
        return np.nan, np.nan
    mean = values.mean()
    std = values.std(ddof=1)
    half = Z * std / np.sqrt(values.size)
    return mean - half, mean + half

# ============================================================
# 1) DATASETS
# ============================================================

datasets = sorted({k[0] for k in dfs.keys() if isinstance(k, tuple)})
print("Datasets:", datasets)

# ============================================================
# 2) EXTRAIR MEDIANAS NORMALIZADAS
# ============================================================

data = {}

for d in datasets:
    for m in METHODS:
        key = (d, m)
        if key not in dfs:
            continue
        dfm = dfs[key]
        med = get_median_row(dfm, METRICS)
        data[key] = med.map(_to_0_1)

# ============================================================
# 3) DATAFRAME FINAL
# ============================================================

rows = []

for metric in METRICS:
    metric_name = metric.replace("(%)", "")

    for m in METHODS:
        row = {"Métrica": metric_name, "Método": DISPLAY_METHOD[m]}
        vals = []

        for d in datasets:
            v = data.get((d, m), {}).get(metric, np.nan)
            row[d] = round(float(v), DECIMALS) if np.isfinite(v) else np.nan
            vals.append(v)

        vals = np.array(vals, dtype=float)

        row["Média"]   = round(np.nanmean(vals), DECIMALS) if np.isfinite(vals).any() else np.nan
        row["Mediana"] = round(np.nanmedian(vals), DECIMALS) if np.isfinite(vals).any() else np.nan

        lo, hi = ci95(vals)
        row["IC 95%"] = f"[{lo:.{DECIMALS}f},{hi:.{DECIMALS}f}]" if np.isfinite(lo) else ""

        rows.append(row)

df_excel = pd.DataFrame(rows)

# ordenação estética
df_excel["__ord"] = df_excel["Método"].map({
    "ANCHOR": 0, "LIME": 1, "PFI": 2, "SHAP (Tree)": 3, "Surrogate": 4
})
df_excel = df_excel.sort_values(["Métrica","__ord"]).drop(columns="__ord")

# ============================================================
# 4) EXPORTAÇÃO
# ============================================================

df_excel.to_excel("metricas_multidataset_layout_limpo.xlsx", index=False)
df_excel.to_csv("metricas_multidataset_layout_limpo.csv", index=False, encoding="utf-8-sig")

print("✔ Excel e CSV gerados com sucesso")
print("Shape final:", df_excel.shape)
```
