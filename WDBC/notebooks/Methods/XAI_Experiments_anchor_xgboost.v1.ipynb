```python
# === Anchor (Interpretabilidade) — usando X (train+test) e Y, com paralelização === #teste para reaproveitar anchor
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from alibi.explainers import AnchorTabular
from joblib import Parallel, delayed
import psutil, multiprocessing
import time   # <<< NOVO: para medir tempo

# -------------------------------------------------
# Premissas do ambiente:
# - X: DataFrame completo (train+test), ordenado por "id"
# - Y: Series/coluna binária (0/1) alinhada a X
# - model: classificador já treinado
# - ENTRADAS_SELECIONADAS: índices (posições) em X para explicar
# - BASE_PATH: diretório base para salvar o Excel
# -------------------------------------------------

# Nomes de classes (0 -> 'benigno', 1 -> 'maligno')
class_names = ['B', 'M']

# ==== (0) PARAM CONFIG IGUAL AO CÓDIGO DE MÉTRICAS ====
ANCHOR_THRESHOLD  = 0.95   # mesmo do outro código
ANCHOR_BEAM_SIZE  = 1      # mesmo do outro código
ANCHOR_MAX_ANCHOR = 5      # mesmo do outro código
ANCHOR_SEED       = 13     # mesmo do outro código

# cache opcional em memória das explicações completas
ANCHOR_EXPLANATIONS = {}   # será preenchido DEPOIS do Parallel

# ==== (1) GARANTIR MESMAS FEATURES DO TREINO ====
feat_cols = list(model.feature_names_in_)


feature_names = feat_cols  # o Anchor usará estes nomes exatamente nesta ordem

# ==== (2) PREDICTOR "ADAPTADOR" ====
def predict_fn(arr: np.ndarray):
    arr = np.asarray(arr)
    if arr.ndim == 1:
        arr = arr.reshape(1, -1)
    return model.predict(arr)

# ==== (3) CRIAR EXPLICADOR ====
anchor_explainer = AnchorTabular(predict_fn, feature_names)

X_fit = X[feature_names].to_numpy()
anchor_explainer.fit(X_fit, disc_perc=(25, 50, 75))

# Preparar y e X alinhados (garantir índices 0..n-1)
y = pd.Series(Y)

# ============================
# (5) Função para UMA instância
# ============================
def _explain_one_anchor(pos: int):
    """
    Calcula a explicação Anchor para a instância na posição `pos`
    (índice inteiro em X) e retorna um dicionário com os resultados.
    """
    t0 = time.time()
    print(f"[ANCHOR] Iniciando explicação da instância pos={pos}...")

    x_row = X[X['id'] == pos][feature_names].to_numpy()

    # explicação — passar 2D para o Anchor (1, n_features)
    explanation = anchor_explainer.explain(
        x_row.reshape(1, -1),
        seed=ANCHOR_SEED,
        threshold=ANCHOR_THRESHOLD,
        beam_size=ANCHOR_BEAM_SIZE,
        max_anchor_size=ANCHOR_MAX_ANCHOR,
        verbose=False
    )

    y_pred_label = int(predict_fn(x_row)[0])

    if explanation.anchor:
        regra = ' AND '.join(explanation.anchor)
    else:
        regra = '<sem condição>'

    dt = time.time() - t0
    print(f"[METRIC-END] Anchor para instância pos={pos} levou {dt:.3f}s")

    return {
        "idx_instancia": pos,                      # chave para linkar com X/Y
        "predicao": class_names[y_pred_label],
        "precision_anchor": float(explanation.precision),
        "cobertura_anchor": float(explanation.coverage),
        "regra_anchor": regra,
        "diagnostico": X[X['id']==pos]['diagnosis'].values[0],
        "explanation": explanation                 # <<<<<< guarda a explicação aqui
    }

# ============================
# (6) Configuração de paralelização
# ============================
RAM_GB  = psutil.virtual_memory().total / 1024**3
MAX_CPU = multiprocessing.cpu_count()

NOTEBOOKS_RODANDO = 1  # se estiver com 2 notebooks em paralelo, mude para 2

N_JOBS_ANCHOR = max(
    1,
    min(
        MAX_CPU,
        int((RAM_GB // 4) // NOTEBOOKS_RODANDO) or 1
    )
)
print(f"[ANCHOR] Rodando explicações para {len(ENTRADAS_SELECIONADAS)} instâncias com n_jobs={N_JOBS_ANCHOR} (CPUs={MAX_CPU}, RAM≈{RAM_GB:.1f} GB)")

# ============================
# (7) Executar em paralelo
# ============================
anchor_results = Parallel(n_jobs=N_JOBS_ANCHOR, backend="loky")(
    delayed(_explain_one_anchor)(pos) for pos in ENTRADAS_SELECIONADAS
)

# ---------- MONTA CACHE PARA O CÓDIGO DE MÉTRICAS ----------
ANCHOR_EXPLANATIONS = {
    r["idx_instancia"]: r["explanation"]
    for r in anchor_results
}

# monta df_anchor SEM a coluna 'explanation'
df_anchor = pd.DataFrame(
    [
        {k: v for k, v in r.items() if k != "explanation"}
        for r in anchor_results
    ]
)

# Salvar em Excel no diretório BASE_PATH
out_xlsx = "interpretacao_anchor.xlsx"
df_anchor.to_excel(f"{BASE_PATH}/{out_xlsx}", index=False)
print(f"✅ Arquivo {out_xlsx} gerado com sucesso em {BASE_PATH}.")

# ================== EXIBIÇÕES GRÁFICAS ==================

# Distribuição da Precisão
plt.figure(figsize=(10, 5))
plt.hist(df_anchor["precision_anchor"], bins=10, edgecolor='black')
plt.title("Distribuição da Precisão das Regras Anchor")
plt.xlabel("Precisão")
plt.ylabel("Frequência")
plt.grid(True)
plt.tight_layout()
plt.show()

# Distribuição da Cobertura
plt.figure(figsize=(10, 5))
plt.hist(df_anchor["cobertura_anchor"], bins=10, edgecolor='black')
plt.title("Distribuição da Cobertura das Regras Anchor")
plt.xlabel("Cobertura")
plt.ylabel("Frequência")
plt.grid(True)
plt.tight_layout()
plt.show()

# Top 10 regras mais frequentes
(df_anchor["regra_anchor"]
 .value_counts()
 .head(10)
 .plot(kind='barh', figsize=(10, 6)))
plt.title("Top 10 Regras Anchor mais frequentes")
plt.xlabel("Frequência")
plt.ylabel("Regra")
plt.tight_layout()
plt.show()

# ====== Versão com nós circulares ======
def draw_decision_tree(ax, conditions, prediction, precisao=None, cobertura=None):
    n = len(conditions)
    h_spacing = 1.5
    v_spacing = -1.5
    x = 0.0
    y = 0.0

    for i, cond in enumerate(conditions):
        circ = plt.Circle((x, y), 0.3, ec='black', fc="lightblue")
        ax.add_patch(circ)
        ax.text(x, y, cond, fontsize=9, ha='center', va='center', wrap=True)
        if i < n - 1:
            x_left = x - h_spacing / (2 ** i)
            y_next = y + v_spacing
            ax.plot([x, x_left], [y, y_next], 'k-')
            x, y = x_left, y_next

    y_final = y + v_spacing
    ax.plot([x, x], [y, y_final], 'k-')

    circ_leaf = plt.Circle((x, y_final), 0.3, ec='black', fc="lightgreen")
    ax.add_patch(circ_leaf)
    ax.text(x, y_final, prediction.upper(), fontsize=10, ha='center', va='center')

    if (precisao is not None) and (cobertura is not None):
        ax.set_title(f"Prec: {precisao:.2f} | Cob: {cobertura:.2f}", fontsize=10)

    ax.set_xlim(-3, 3)
    ax.set_ylim(y_final - 1, 1.5)
    ax.axis("off")


n_arvores = len(df_anchor)  # min(9, len(df_anchor))s
n_cols = 3
n_rows = math.ceil(n_arvores / n_cols)

fig, axs = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axs = np.atleast_1d(axs).flatten()

for i in range(n_arvores):
    regra = df_anchor.iloc[i]["regra_anchor"]
    pred = df_anchor.iloc[i]["predicao"]
    prec = df_anchor.iloc[i]["precision_anchor"]
    cob = df_anchor.iloc[i]["cobertura_anchor"]
    conditions = regra.split(" AND ") if isinstance(regra, str) and len(regra) else ["<sem condição>"]
    draw_decision_tree(axs[i], conditions, pred, prec, cob)

for j in range(n_arvores, len(axs)):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.show()
# ================== FIM DO CÓDIGO ANCHOR ==================
```

```python
# === Anchor Interpretabilidade [Funções de Cálculo das Métricas] ===

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics.pairwise import cosine_similarity
import re
import numpy as np
from sklearn.metrics import mean_squared_error
from itertools import combinations
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import spearmanr
from typing import Dict, List, Optional, Tuple
import time


# =====================================================
# Helpers
# =====================================================

_lit_clean = re.compile(r"\s+")
_feat_name_re = re.compile(r"^([^\s<>=]+)")

def literal_to_feature(literal: str) -> str:
    """Extrai apenas o nome da feature (para 'jaccard por feature')."""
    s = literal.strip()
    m = _feat_name_re.match(s)
    return m.group(1) if m else s


def normalize_literal_token(literal: str) -> str:
    """Normaliza a string do literal para comparação (jaccard por 'literal')."""
    s = literal.replace('≤', '<=').replace('≥', '>=').strip()
    s = _lit_clean.sub(" ", s)
    return s

def anchor_to_set(anchor_list: List[str], representation: str = "literal") -> set:
    """
    representation:
      - 'literal' (default): cada condição completa é um elemento do set
      - 'feature': colapsa por nome de feature
    """
    if not anchor_list:
        return set()
    if representation == "feature":
        return {literal_to_feature(z) for z in anchor_list}
    # default: 'literal'
    return {normalize_literal_token(z) for z in anchor_list}

def _get_row_and_index(X_df: pd.DataFrame, pos_id):
    """
    pos_id pode ser:
      - um rótulo do índice de X_df (X_df.index)
      - um valor real da coluna 'id'
      - uma posição inteira (iloc)
    Retorna: (row_df, idx_label)
    """
    # 1) tentar como rótulo de índice
    if pos_id in X_df.index:
        row = X_df.loc[[pos_id]]
        return row, row.index[0]

    # 2) tentar na coluna 'id' como valor real
    if 'id' in X_df.columns:
        mask = X_df['id'] == pos_id
        if mask.any():
            row = X_df.loc[mask].iloc[[0]]
            return row, row.index[0]

    # 3) fallback: tratar como posição
    if isinstance(pos_id, (int, np.integer)) and 0 <= pos_id < len(X_df):
        row = X_df.iloc[[pos_id]]
        return row, row.index[0]

    raise ValueError(
        f"pos_id={pos_id} não encontrado nem no index nem na coluna 'id'. "
        f"Verifique se você está passando o ID real correto."
    )

def calcular_ic95_percentual(valores):
    media = np.mean(valores)
    std = np.std(valores, ddof=1)
    n = len(valores)
    erro_padrao = std / np.sqrt(n)
    ic_low = media - 1.96 * erro_padrao
    ic_high = media + 1.96 * erro_padrao
    return ic_low, ic_high

def calcular_ic95_percentual_median(valores):
    mediana = np.median(valores)
    std = np.std(valores, ddof=1)
    n = len(valores)
    erro_padrao = std / np.sqrt(n)
    ic_low = mediana - 1.96 * erro_padrao
    ic_high = mediana + 1.96 * erro_padrao
    return ic_low, ic_high

def formatar_ic_percentual(valor_central, limite_inferior, limite_superior):
    return f"{valor_central:.1f} [{limite_inferior:.1f}, {limite_superior:.1f}]"

# ========= Funções: fidelity e completeness (usando df_anchor_all) =========

def get_anchor_fidelity_from_global(df_anchor_all: pd.DataFrame, idx_instancia: int):
    """
    Retorna dicionário com fidelity para a instância idx_instancia,
    usando 'precision_anchor' da rodada GLOBAL (df_anchor_all).
    """
    hit = df_anchor_all.loc[df_anchor_all["idx_instancia"] == int(idx_instancia)]
    if hit.empty:
        return {
            "idx_instancia": int(idx_instancia),
            "fidelity_real": np.nan,
            "fidelity_percent": np.nan,
            "pred_name": None,
            "regra_anchor": None,
        }
    r = hit.iloc[0]
    fid = float(r["precision_anchor"]) if pd.notna(r["precision_anchor"]) else np.nan
    pred_val = r.get("pred_name", r.get("predicao", None))
    return {
        "idx_instancia": int(idx_instancia),
        "idx": r.get("idx", None),
        "fidelity_real": fid,
        "fidelity_percent": (fid * 100.0) if np.isfinite(fid) else np.nan,
        "pred_name": pred_val,
        "regra_anchor": r.get("regra_anchor", None),
    }


def get_anchor_completeness_from_global(df_anchor_all: pd.DataFrame, idx_instancia: int):
    cov_col = "cobertura_anchor"

    hit = df_anchor_all.loc[df_anchor_all["idx_instancia"] == int(idx_instancia)]
    if hit.empty:
        return {
            "idx_instancia": int(idx_instancia),
            "idx": None,
            "completeness_real": np.nan,
            "completeness_percent": np.nan,
            "coverage_source_col": cov_col,
        }

    r = hit.iloc[0]
    cov = float(r[cov_col]) if pd.notna(r[cov_col]) else np.nan

    return {
        "idx_instancia": int(idx_instancia),
        "idx": r.get("idx", None),
        "completeness_real": cov,
        "completeness_percent": (cov * 100.0) if np.isfinite(cov) else np.nan,
        "coverage_source_col": cov_col,
    }

def compute_anchor_infidelity_const_for_instance(
    df_anchor_all: pd.DataFrame,
    X_df: pd.DataFrame,
    model,
    feature_names: list[str],
    pos_id: int,
    pos_class: int = 1,
):
    hit = df_anchor_all.loc[df_anchor_all["idx_instancia"] == int(pos_id)]
    if hit.empty:
        return {
            "pos_id": int(pos_id),
            "idx": None,
            "regra_anchor": None,
            "coverage_used": 0.0,
            "n_covered": 0,
            "gA_mean_prob": np.nan,
            "infidelity_abs": np.nan,
            "infidelity_percent": np.nan,
        }
    r = hit.iloc[0]
    regra_str = (r.get("regra_anchor") or "").strip()
    anchor_list = [] if (regra_str == "" or regra_str == "<regra vazia>") else [
        c.strip() for c in regra_str.split(" AND ")
    ]

    coverage_used, mask = support_on_df(anchor_list, X_df)
    idx = r.get("idx", None)

    n_cov = int(mask.sum())
    if n_cov == 0:
        return {
            "pos_id": int(pos_id),
            "idx": idx,
            "regra_anchor": regra_str if regra_str else "<regra vazia>",
            "coverage_used": 0.0,
            "n_covered": 0,
            "gA_mean_prob": np.nan,
            "infidelity_abs": np.nan,
            "infidelity_percent": np.nan,
        }

    if hasattr(model, "feature_names_in_"):
        feat_model = list(model.feature_names_in_)
    else:
        feat_model = list(feature_names)

    feat_model = [c for c in feat_model if c not in EXCLUDE_COLS]
    feat_ok = [c for c in feat_model if c in X_df.columns]

    X_cov = X_df.loc[mask, feat_ok]

    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(X_cov)
        fz = probs[:, pos_class]
    else:
        preds = model.predict(X_cov)
        fz = preds.astype(float)

    gA = float(np.mean(fz))
    infid_abs = float(np.mean((fz - gA) ** 2))
    infid_pct = infid_abs * 100.0

    return {
        "pos_id": int(pos_id),
        "idx": idx,
        "regra_anchor": regra_str if regra_str else "<regra vazia>",
        "coverage_used": float(coverage_used),
        "n_covered": n_cov,
        "gA_mean_prob": gA,
        "infidelity_abs": infid_abs,
        "infidelity_percent": infid_pct,
    }

# ======================================================
# 1) FAITHFULNESS / SUFFICIENCY - faithfulness_probabilistica 
# “Quão parecida é a confiança do modelo na instância explicada em relação à confiança média do modelo em todos os pontos que satisfazem a regra Anchor?”
#
#Então, na prática, está medindo homogeneidade de probabilidade da classe-alvo dentro da região do Anchor.
#Se a região ancorada é “coerente” com a instância (mesmo nível de confiança), faithfulness fica perto de 1 (ou 100%).
#
# =====================================================
def compute_anchor_faithfulness_suff_for_instance(
    df_anchor_all: pd.DataFrame,
    X_df: pd.DataFrame,
    model,
    feature_names: list[str],
    pos_id,   # label ou posição
):
    pos_int = int(pos_id)

    hit = df_anchor_all.loc[df_anchor_all["idx_instancia"] == pos_int]
    if hit.empty:
        return {
            "pos_id": pos_int,
            "idx": None,
            "regra_anchor": None,
            "y_pred_class": None,
            "p_x": np.nan,
            "p_A": np.nan,
            "coverage_used": 0.0,
            "n_covered": 0,
            "faithfulness_abs": np.nan,
            "faithfulness_percent": np.nan,
        }
    r = hit.iloc[0]
    regra_str = (r.get("regra_anchor") or "").strip()
    anchor_list = [] if (regra_str == "" or regra_str == "<regra vazia>") else [
        c.strip() for c in regra_str.split(" AND ")
    ]
    idx_label = r.get("idx", None)

    x_row_ser, idx_label_X = _get_row_and_index(X_df, pos_id)

    if hasattr(model, "feature_names_in_"):
        feat_model = list(model.feature_names_in_)
    else:
        feat_model = list(feature_names)
    feat_model = [c for c in feat_model if c not in EXCLUDE_COLS]
    feat_ok = [c for c in feat_model if c in X_df.columns]

    x_row = x_row_ser[feat_ok].to_frame().T

    if hasattr(model, "predict_proba"):
        y_pred = int(model.predict(x_row)[0])
        class_order = list(model.classes_) if hasattr(model, "classes_") else [0, 1]
        c_idx = class_order.index(y_pred) if y_pred in class_order else y_pred
        p_x = float(model.predict_proba(x_row)[0, c_idx])
    else:
        y_pred = int(model.predict(x_row)[0])
        c_idx = y_pred
        p_x = float(y_pred)

    coverage_used, mask_A = support_on_df(anchor_list, X_df)
    n_cov = int(mask_A.sum())
    if n_cov == 0:
        return {
            "pos_id": pos_int,
            "idx": idx_label if idx_label is not None else idx_label_X,
            "regra_anchor": regra_str if regra_str else "<regra vazia>",
            "y_pred_class": y_pred,
            "p_x": p_x,
            "p_A": np.nan,
            "coverage_used": 0.0,
            "n_covered": 0,
            "faithfulness_abs": np.nan,
            "faithfulness_percent": np.nan,
        }

    X_cov = X_df.loc[mask_A, feat_ok]

    if hasattr(model, "predict_proba"):
        p_cov = model.predict_proba(X_cov)[:, c_idx]
    else:
        p_cov = model.predict(X_cov).astype(float)

    p_A = float(np.mean(p_cov))

    faith_abs = float(1.0 - abs(p_A - p_x))
    faith_pct = 100.0 * faith_abs

    return {
        "pos_id": pos_int,
        "idx": idx_label if idx_label is not None else idx_label_X,
        "regra_anchor": regra_str if regra_str else "<regra vazia>",
        "y_pred_class": y_pred,
        "p_x": p_x,
        "p_A": p_A,
        "coverage_used": float(coverage_used),
        "n_covered": n_cov,
        "faithfulness_abs": faith_abs,
        "faithfulness_percent": faith_pct,
    }

def compute_anchor_simplicity_for_instance(
    df_anchor_all: pd.DataFrame,
    pos_id: int,
    max_len: int | None = None,
):
    hit = df_anchor_all.loc[df_anchor_all["idx_instancia"] == int(pos_id)]
    if hit.empty:
        return {
            "pos_id": int(pos_id),
            "idx": None,
            "regra_anchor": None,
            "n_conditions": np.nan,
            "simplicity_abs": np.nan,
            "simplicity_percent": np.nan,
        }

    r = hit.iloc[0]
    regra_str = (r.get("regra_anchor") or "").strip()
    if regra_str == "" or regra_str == "<regra vazia>":
        n_cond = 0
    else:
        n_cond = len([c for c in regra_str.split(" AND ") if c.strip() != ""])

    # LÓGICA REVISADA: Limita max_len a um máximo de 5
    if max_len is None:
        if "n_conditions" in df_anchor_all.columns and df_anchor_all["n_conditions"].notna().any():
            max_len_data = int(df_anchor_all["n_conditions"].max())
            # Garante que o normalizador (max_len) não passe de 5, mas seja no mínimo 1
            max_len = max(1, min(5, max_len_data))
        else:
            # Fallback padrão sensato, limitado a 3 (que é <= 5)
            max_len = 3

    simplicity_abs = float(1.0 - min(n_cond, max_len) / max_len)
    simplicity_percent = 100.0 * simplicity_abs

    return {
        "pos_id": int(pos_id),
        "idx": r.get("idx", None),
        "regra_anchor": regra_str if regra_str else "<regra vazia>",
        "n_conditions": int(n_cond),
        "simplicity_abs": simplicity_abs,
        "simplicity_percent": simplicity_percent,
    }

def _cond_to_feature(cond: str, feature_names):
    for name in sorted(feature_names, key=len, reverse=True):
        if re.search(rf"\\b{re.escape(name)}\\b", cond):
            return name
    return None


# =====================================================
# 2) CONSISTENCY
# =====================================================
def compute_anchor_consistency_for_instance(
    explainer,
    X_df: pd.DataFrame,
    feature_names: list,
    pos_id,   # label ou posição
    *,
    R: int = 3,
    base_seed: int = 13,
    metric: str = "jaccard",
    representation: str = "literal",
    threshold: float = 0.95,
    beam_size: int = 1,
    max_anchor_size: int = 5,
):
    assert representation in {"literal", "feature"}
    assert metric in {"jaccard", "cosine", "spearman"}

    start_time = time.time()
    print(f"[METRIC-START] Consistency for pos={pos_id} R={R} metric={metric} repr={representation}")

    x_row_ser, idx_label = _get_row_and_index(X_df, pos_id)
    x_vec = x_row_ser[feature_names].to_numpy()

    seeds = [base_seed + 9973 * i for i in range(R)]

    run_sets = []
    run_rules = []

    for run_i, s in enumerate(seeds, start=1):
        print(f"[CONSISTENCY] pos={pos_id} run {run_i}/{R} seed={s}")
        exp = explainer.explain(
            x_vec,
            threshold=threshold,
            beam_size=beam_size,
            max_anchor_size=max_anchor_size,
            seed=s,
            verbose=False,
        )
        anchor_list = exp.anchor or []
        run_rules.append(" AND ".join(anchor_list) if anchor_list else "<regra vazia>")

        if representation == "literal":
            items = set([c.strip() for c in anchor_list])
        else:
            items = set()
            for c in anchor_list:
                f = _cond_to_feature(c, feature_names)
                if f is not None:
                    items.add(f)
        run_sets.append(items)

    pair_sims = []

    if metric == "jaccard":
        for a, b in combinations(run_sets, 2):
            union = len(a | b)
            inter = len(a & b)
            sim = 1.0 if union == 0 else inter / union
            pair_sims.append(sim)
    else:
        vocab = sorted(set().union(*run_sets))
        if len(vocab) == 0:
            pair_sims = [1.0]
        else:
            M = np.zeros((R, len(vocab)), dtype=float)
            tok2idx = {t: j for j, t in enumerate(vocab)}
            for i, items in enumerate(run_sets):
                for t in items:
                    M[i, tok2idx[t]] = 1.0

            for i, j in combinations(range(R), 2):
                v1, v2 = M[i], M[j]
                if metric == "cosine":
                    n1, n2 = np.linalg.norm(v1), np.linalg.norm(v2)
                    sim = 1.0 if (n1 == 0 or n2 == 0) else float(np.dot(v1, v2) / (n1 * n2))
                else:  # spearman
                    rho, _ = spearmanr(v1, v2)
                    if np.isnan(rho):
                        rho = 1.0 if np.all(v1 == v2) else 0.0
                    sim = (rho + 1.0) / 2.0
                pair_sims.append(sim)

    cons_abs = float(np.mean(pair_sims)) if len(pair_sims) > 0 else 1.0
    elapsed = time.time() - start_time
    print(f"[METRIC-END] Consistency pos={pos_id} took {elapsed:.3f}s -> {cons_abs:.4f}")
    return {
        "pos_id": pos_id,
        "idx": idx_label,
        "R": R,
        "metric": metric,
        "representation": representation,
        "consistency_abs": cons_abs,
        "consistency_percent": cons_abs * 100.0,
        "runs_anchors": run_rules,
    }


def _set_similarity(a: set, b: set, metric: str = "jaccard"):
    if metric == "jaccard":
        u = len(a | b)
        i = len(a & b)
        return 1.0 if u == 0 else i / u
    elif metric == "cosine":
        vocab = sorted((a | b))
        if not vocab:
            return 1.0
        v1 = np.array([1.0 if t in a else 0.0 for t in vocab])
        v2 = np.array([1.0 if t in b else 0.0 for t in vocab])
        n1, n2 = np.linalg.norm(v1), np.linalg.norm(v2)
        return 1.0 if (n1 == 0 or n2 == 0) else float(np.dot(v1, v2) / (n1 * n2))
    elif metric == "spearman":
        vocab = sorted((a | b))
        if not vocab:
            return 1.0
        v1 = np.array([1.0 if t in a else 0.0 for t in vocab])
        v2 = np.array([1.0 if t in b else 0.0 for t in vocab])
        rho, _ = spearmanr(v1, v2)
        if np.isnan(rho):
            rho = 1.0 if np.all(v1 == v2) else 0.0
        return (rho + 1.0) / 2.0
    else:
        raise ValueError(f"Métrica desconhecida: {metric}")

def _sample_neighborhood_row(
    x_row,
    X_df: pd.DataFrame,
    eps_rel: float = 0.05,
    binary_cols: list[str] | None = None,
    flip_prob: float = 0.05,
    max_flips: int | None = None,
    rng: np.random.Generator | None = None,
):
    """
    Gera um vizinho x' de x_row no espaço de features.

    - x_row: linha alvo (pode ser Series ou DataFrame 1xD)
    - X_df: DataFrame de referência (SOMENTE features do modelo, sem id, y, etc.)
    """

    if rng is None:
        rng = np.random.default_rng()

    # --- garantir que trabalhamos como Series 1D ---
    if isinstance(x_row, pd.DataFrame):
        assert len(x_row) == 1, "x_row DataFrame deve ter exatamente 1 linha"
        x_ser = x_row.iloc[0].copy()
    else:
        # garante Series com índice das colunas
        x_ser = pd.Series(x_row).copy()

    # cópia que vamos perturbar
    x_prime = x_ser.copy()

    # colunas numéricas do espaço de referência
    num_cols = [
        c for c in X_df.columns
        if pd.api.types.is_numeric_dtype(X_df[c])
    ]

    if binary_cols is None:
        binary_cols = []

    # --- 1) perturbar colunas numéricas com janela relativa eps_rel ---
    for col in num_cols:
        if col not in x_prime.index:
            continue

        v = x_prime[col]

        # se já veio NaN por algum motivo, caímos para a mediana global
        if pd.isna(v):
            base = float(X_df[col].median())
        else:
            # <<< AJUSTE para o FutureWarning: Series -> escalar
            if isinstance(v, pd.Series):
                v = v.iloc[0]
            base = float(v)

        col_min = float(X_df[col].min())
        col_max = float(X_df[col].max())

        if not np.isfinite(col_min) or not np.isfinite(col_max):
            # se a coluna toda for NaN, apenas mantém o valor base
            x_prime[col] = base
            continue

        delta = eps_rel * (col_max - col_min)
        low  = max(col_min, base - delta)
        high = min(col_max, base + delta)

        if low >= high or not np.isfinite(low) or not np.isfinite(high):
            new_val = base
        else:
            new_val = rng.uniform(low, high)

        x_prime[col] = new_val

    # --- 2) flip em colunas binárias ---
    flips = 0
    for col in binary_cols:
        if col not in x_prime.index:
            continue

        v = x_prime[col]
        if pd.isna(v):
            continue

        # de novo, garantir escalar
        if isinstance(v, pd.Series):
            v = v.iloc[0]

        if rng.random() < flip_prob:
            v_int = int(round(float(v)))
            x_prime[col] = 1 - v_int
            flips += 1
            if max_flips is not None and flips >= max_flips:
                break

    # --- retorna como DataFrame 1xD com MESMAS colunas ---
    x_prime_df = x_prime.to_frame().T
    x_prime_df = x_prime_df[X_df.columns]  # garante mesma ordem

    return x_prime_df

# =====================================================
# 3) ROBUSTNESS
# =====================================================
def compute_anchor_robustness_for_instance(     
    explainer,
    model,
    X_df: pd.DataFrame,
    feature_names: list[str],   # <- DEVE SER o espaço de features do modelo (30 colunas)
    pos_id,                     # id real ou índice
    *,
    N: int = 3,
    eps_rel: float = 0.05,
    binary_cols: list[str] | None = None,
    flip_prob: float = 0.05,
    max_flips: int | None = None,
    metric: str = "jaccard",
    representation: str = "literal",
    mode: str = "worst",
    seed_explain: int = 13,
    seed_perturb: int = 123,
    exp_x: Optional[object] = None,
    anchor_kwargs: Optional[dict] = None,
):
    assert representation in {"literal", "feature"}
    assert metric in {"jaccard", "cosine", "spearman"}
    assert mode in {"worst", "mean"}

    start_time = time.time()
    print(f"[METRIC-START] Robustness for pos={pos_id} N={N} metric={metric} mode={mode}")

    rng = np.random.default_rng(seed_perturb)

    # === (1) espaço de features usado pelo Anchor e pelo modelo ===
    # <<< AJUSTE: garantir que só usamos as features do modelo (sem id, y, etc.)
    X_feat = X_df[feature_names].copy()

    # === (2) recuperar linha alvo a partir do ID real ou índice ===
    x_row_df, idx_label = _get_row_and_index(X_df, pos_id)   # <<< USA função ajustada
    # garantir que é DataFrame 1xD
    if isinstance(x_row_df, pd.Series):
        x_row_df = x_row_df.to_frame().T

    # só as features do modelo para x
    x_row_feat = x_row_df[feature_names].copy()               # <<< AJUSTE
    #print(f"x_row_feat: {x_row_feat.shape}")
    #print(f"x_row_feat:\n{x_row_feat}")

    # parâmetros default do Anchor
    if anchor_kwargs is None:
        anchor_kwargs = dict(
            threshold=getattr(explainer, "threshold", 0.95),
            beam_size=getattr(explainer, "beam_size", 4),
            max_anchor_size=getattr(explainer, "max_anchor_size", 6),
            verbose=False
        )

    # === (3) explicação da instância original (x) ===
    if exp_x is None:
        exp_x = explainer.explain(
            x_row_feat.values[0],   # <<< AJUSTE: passa só as features, 1D
            seed=seed_explain,
            **anchor_kwargs
        )
    anchor_x = exp_x.anchor or []
    rule_x = " AND ".join(anchor_x) if anchor_x else "<regra vazia>"

    # representação da âncora de x
    if representation == "literal":
        items_x = set([c.strip() for c in anchor_x])
    else:
        items_x = set()
        for c in anchor_x:
            f = _cond_to_feature(c, feature_names)
            if f is not None:
                items_x.add(f)

    # predição original do modelo em x (usando só as features)
    y_hat_x = int(model.predict(x_row_feat)[0])

    sims = []
    inv_count = 0
    pred_same = 0
    pred_same_cond = 0
    cond_den = 0

    for k in range(int(N)):
        iter_start = time.time()
        print(f"[ROBUSTNESS] pos={pos_id} iter {k+1}/{int(N)}")

        # === (4) amostrar vizinho de x no espaço de features ===
        x_prime = _sample_neighborhood_row(
            x_row=x_row_feat,   # <<< passa só as features
            X_df=X_feat,        # <<< referência só com features
            eps_rel=eps_rel,
            binary_cols=binary_cols,
            flip_prob=flip_prob,
            max_flips=max_flips,
            rng=rng,
        )
        # garantir DataFrame 1xD
        if isinstance(x_prime, pd.Series):
            x_prime = x_prime.to_frame().T

        # só features para x'
        x_prime_feat = x_prime[feature_names].copy()  # <<< AJUSTE

        #print(f"x_prime: {x_prime.shape}")
        #print(f"x_prime_feat: {x_prime_feat.shape}")
        #print(f"x_prime_feat:\n{x_prime_feat}")

        # === (5) explicação em x' (Anchor) ===
        exp_p = explainer.explain(
            x_prime_feat.values[0],
            seed=seed_explain,
            **anchor_kwargs
        )
        anchor_p = exp_p.anchor or []

        # conjunto de itens em x'
        if representation == "literal":
            items_p = set([c.strip() for c in anchor_p])
        else:
            items_p = set()
            for c in anchor_p:
                f = _cond_to_feature(c, feature_names)
                if f is not None:
                    items_p.add(f)

        # similaridade das âncoras
        sims.append(_set_similarity(items_x, items_p, metric=metric))

        # === (6) verificar se x' satisfaz a regra A(x) ===
        if len(anchor_x) == 0:
            inv = True
        else:
            # usamos o próprio x_prime_feat (1 linha, colunas = feature_names)
            df_one = x_prime_feat.copy()
            try:
                query = " and ".join(anchor_x).replace('≤', '<=').replace('≥', '>=')
                for name in sorted(feature_names, key=len, reverse=True):
                    query = re.sub(rf'\b{re.escape(name)}\b', f'`{name}`', query)
                mask = df_one.eval(query, engine="python").astype(bool)
                inv = bool(mask.iloc[0])
            except Exception as e:
                print(f"[WARN] falha ao avaliar regra em x': {e}")
                inv = False

        inv_count += int(inv)

        # === (7) predição do modelo em x' (só features) ===
        y_hat_p = int(model.predict(x_prime_feat)[0])  # <<< AJUSTE

        if y_hat_p == y_hat_x:
            pred_same += 1

        if inv:
            cond_den += 1
            if y_hat_p == y_hat_x:
                pred_same_cond += 1

        iter_elapsed = time.time() - iter_start
        # print(f"[ROBUSTNESS] iter {k+1} took {iter_elapsed:.3f}s")

    # === (8) Estabilidade da explicação ===
    sims = np.array(sims, dtype=float)
    if sims.size == 0:
        sim_min = 1.0
        sim_mean = 1.0
        stability_abs = 1.0
    else:
        sim_min = float(np.min(sims))
        sim_mean = float(np.mean(sims))
        stability_abs = sim_min if mode == "worst" else sim_mean

    stability_percent = 100.0 * stability_abs

    # === (9) Robustez da predição ===
    inv_rate = inv_count / float(N) if N > 0 else np.nan
    pred_stab = pred_same / float(N) if N > 0 else np.nan
    pred_stab_cond = (pred_same_cond / float(cond_den)) if cond_den > 0 else np.nan

    if cond_den > 0:
        robustness_abs = float(pred_stab_cond)
    else:
        robustness_abs = float(pred_stab)

    if np.isnan(robustness_abs):
        robustness_abs = np.nan
        robustness_percent = np.nan
    else:
        robustness_percent = 100.0 * robustness_abs

    elapsed = time.time() - start_time
    print(
        f"[METRIC-END] Robustness pos={pos_id} took {elapsed:.3f}s -> "
        f"{robustness_percent:.3f}% (prediction invariance)"
    )

    return {
        "pos_id": pos_id,
        "idx": idx_label,
        "representation": representation,
        "metric": metric,
        "mode": mode,
        "N": int(N),
        "eps_rel": float(eps_rel),

        "robustness_abs": robustness_abs,
        "robustness_percent": robustness_percent,

        "stability_abs": stability_abs,
        "stability_percent": stability_percent,
        "sim_min": sim_min,
        "sim_mean": sim_mean,

        "anchor_x": rule_x,
        "inv_rate": inv_rate,
        "pred_stability": pred_stab,
        "pred_stability_condA": pred_stab_cond,
    }

def _get_class_index(model, pos_class):
    classes = getattr(model, "classes_", None)
    if classes is None:
        return int(pos_class)
    idx = np.where(classes == pos_class)[0]
    return int(idx[0]) if len(idx) else int(np.argmax(classes))

def _score_proba(model, X1row_df, class_index):
    if hasattr(model, "predict_proba"):
        proba = model.predict_proba(X1row_df)[0]
        return float(proba[class_index])
    pred = int(model.predict(X1row_df)[0])
    return 1.0 if pred == model.classes_[class_index] else 0.0

def _parse_literal(lit):
    s = lit.strip().replace('`', '')
    s = s.replace('≤', '<=').replace('≥', '>=')
    s = re.sub(r'\\s+', ' ', s)

    m = re.match(r'^(-?\\d+\\.?\\d*)\\s*(<=|<)\\s*([A-Za-z0-9_]+)\\s*(<=|<)\\s*(-?\\d+\\.?\\d*)$', s)
    if m:
        low, op1, feat, op2, high = m.groups()
        return {'type':'interval', 'feat':feat, 'low':float(low), 'op_low':op1, 'high':float(high), 'op_high':op2}

    m = re.match(r'^([A-Za-z0-9_]+)\\s*(<=|>=|<|>)\\s*(-?\\d+\\.?\\d*)$', s)
    if m:
        feat, op, val = m.groups()
        return {'type':'ineq', 'feat':feat, 'op':op, 'val':float(val)}

    m = re.match(r'^([A-Za-z0-9_]+)\\s*(==|=)\\s*(-?\\d+\\.?\\d*)$', s)
    if m:
        feat, op, val = m.groups()
        return {'type':'eq', 'feat':feat, 'op':op, 'val':float(val)}

    return {'type':'unknown', 'raw':lit}

def _break_literal_minimal(x0_ser, literal_parsed, feature_ranges=None, eps=1e-3):
    x1 = x0_ser.copy()

    if literal_parsed.get('type') == 'eq':
        f, v = literal_parsed['feat'], literal_parsed['val']
        cur = float(x1[f])
        if v in (0.0, 1.0) and cur in (0.0, 1.0):
            x1[f] = 1.0 - cur
        else:
            lo, hi = (feature_ranges.get(f, (cur-1.0, cur+1.0)) if feature_ranges else (cur-1.0, cur+1.0))
            x1[f] = lo if abs(cur - lo) <= abs(hi - cur) else hi
        return x1

    if literal_parsed.get('type') == 'ineq':
        f, op, t = literal_parsed['feat'], literal_parsed['op'], literal_parsed['val']
        cur = float(x1[f])
        if feature_ranges and f in feature_ranges:
            lo, hi = feature_ranges[f]
            step = max(eps, 0.01 * max(1.0, hi - lo))
        else:
            step = max(eps, 0.01 * max(1.0, abs(cur)))
        if op in ('<=','<'):
            x1[f] = max(cur, t + step)
        elif op in ('>=','>'):
            x1[f] = min(cur, t - step)
        return x1

    if literal_parsed.get('type') == 'interval':
        f = literal_parsed['feat']
        low, high = literal_parsed['low'], literal_parsed['high']
        cur = float(x1[f])
        if feature_ranges and f in feature_ranges:
            lo, hi = feature_ranges[f]
            step = max(eps, 0.01 * max(1.0, hi - lo))
        else:
            step = max(eps, 0.01 * max(1.0, abs(cur)))
        to_low  = abs(cur - (low - step))
        to_high = abs((high + step) - cur)
        if to_low <= to_high:
            x1[f] = low - step
        else:
            x1[f] = high + step
        return x1

    for f in x1.index:
        cur = float(x1[f])
        x1[f] = cur + eps
        break
    return x1

def compute_anchor_selectivity(
    model,
    x0,                          # 1D: pandas Series (recomendado) ou ndarray com ordem feature_names
    feature_names,
    anchor_literals,             # list[str] como vem do Anchor (ex.: ["genero = 1", "40 <= idade < 60"]),
    target_class=None,           # se None, usa a classe prevista pelo modelo,
    feature_ranges=None,         # dict opcional {feat: (min,max)} para calibrar passos,
    eps=1e-3,                    # passo mínimo para sair do limite/intervalo,
    agg="mean",                  # 'mean' ou 'max' (agregador das quedas por literal),
    normalize=True               # se True, retorna também % (queda/base),
):
    if isinstance(x0, np.ndarray):
        x0_ser = pd.Series(x0, index=feature_names)
    else:
        x0_ser = x0.reindex(feature_names)
    if target_class is None:
        pred = int(model.predict(pd.DataFrame([x0_ser.values], columns=feature_names))[0])
        target_class = pred
    class_index = _get_class_index(model, target_class)
    base_score = _score_proba(model, pd.DataFrame([x0_ser.values], columns=feature_names), class_index)
    if not anchor_literals:
        out = {
            'base_score': base_score,
            'drops_per_literal': [],
            'broken_score_mean': base_score,
            'broken_score_max': base_score,
            'sel_abs': 0.0,
            'sel_percent': 0.0 if normalize else None,
            'n_literals': 0
        }
        return out

    drops = []
    for lit in anchor_literals:
        p = _parse_literal(lit)
        x_break = _break_literal_minimal(x0_ser, p, feature_ranges=feature_ranges, eps=eps)
        sc_b = _score_proba(model, pd.DataFrame([x_break.values], columns=feature_names), class_index)
        drops.append({'literal': lit, 'score_break': sc_b, 'drop': max(0.0, base_score - sc_b)})

    broken_mean = float(np.mean([d['score_break'] for d in drops])) if drops else base_score
    broken_max  = float(np.min([d['score_break'] for d in drops])) if drops else base_score

    sel_abs = float(np.mean([d['drop'] for d in drops])) if agg == "mean" else float(np.max([d['drop'] for d in drops]))
    sel_pct = (100.0 * sel_abs / base_score) if (normalize and base_score > 0) else (0.0 if normalize else None)

    return {
        'base_score': base_score,
        'drops_per_literal': drops,
        'broken_score_mean': broken_mean,
        'broken_score_max': broken_max,
        'sel_abs': sel_abs,
        'sel_percent': sel_pct,
        'n_literals': len(drops)
    }
    
def _parse_literal_to_sign(lit: str):
    s = lit.strip().replace("≤", "<=").replace("≥", ">")

    m = re.match(r'^\\s*([\\-0-9\\.]+)\\s*(?:<=|<)\\s*([A-Za-z0-9_]+)\\s*(?:<=|<)\\s*([\\-0-9\\.]+)\\s*$', s)
    if m:
        feat = m.group(2)
        return feat, 0

    m = re.match(r'^\\s*([A-Za-z0-9_]+)\\s*(<=|<|>=|>)\\s*([\\-0-9\\.]+)\\s*$', s)
    if m:
        feat, op, thr = m.group(1), m.group(2), float(m.group(3))
        if op in (">=", ">"):
            return feat, +1
        else:
            return feat, -1

    m = re.match(r'^\\s*([A-Za-z0-9_]+)\\s*=\\s*([^\\s]+)\\s*$', s)
    if m:
        feat, v = m.group(1), m.group(2)
        v_norm = str(v).strip().lower()
        if v_norm in {"1", "true", "yes", "sim"}:
            return feat, +1
        if v_norm in {"0", "false", "no", "nao", "não"}:
            return feat, -1
        return feat, 0

    return None, 0

def _anchor_rule_to_feature_signs(anchor_rule_str: str):
    if not anchor_rule_str or anchor_rule_str == "<regra vazia>":
        return {}

    feats = {}
    for lit in anchor_rule_str.split(" AND "):
        feat, sgn = _parse_literal_to_sign(lit)
        if feat is None:
            continue
        feats[feat] = np.sign(feats.get(feat, 0) + sgn).astype(int)
    return feats

def compute_directional_soundness_anchor(anchor_rule_str: str,
                                         pred_class: int,
                                         direcao_esperada: dict,
                                         *,
                                         ignore_neutral=True):
    anchor_signs = _anchor_rule_to_feature_signs(anchor_rule_str)
    n_literals = len(anchor_signs)

    rows = []
    hits = 0
    denom = 0

    for feat, s_anchor in anchor_signs.items():
        gt = direcao_esperada.get(feat, 0)
        gt_eff = gt if pred_class == 1 else -gt

        if ignore_neutral:
            if gt_eff == 0 or s_anchor == 0:
                rows.append({
                    "feature": feat, "sign_anchor": s_anchor,
                    "gt": gt, "gt_eff": gt_eff, "counted": False, "match": np.nan
                })
                continue

        denom += 1
        match = int(np.sign(s_anchor) == np.sign(gt_eff))
        hits += match

        rows.append({
            "feature": feat, "sign_anchor": s_anchor,
            "gt": gt, "gt_eff": gt_eff, "counted": True, "match": match
        })

    dirsound_abs = int(hits)
    dirsound_percent = (100.0 * hits / denom) if denom > 0 else np.nan

    debug_df = pd.DataFrame(rows)
    return {
        "dirsound_abs": dirsound_abs,
        "dirsound_percent": dirsound_percent,
        "n_compared": int(denom),
        "n_literals": int(n_literals),
        "debug": debug_df
    }
# -----------------------------------------------------------
# Perturbação leve da instância
# -----------------------------------------------------------
def perturb_instance(
    x0: pd.Series,
    *,
    binary_cols: List[str],
    cont_cols: List[str],
    flip_prob: float = 0.05,
    sigma: float = 0.05,
    ranges: Optional[Dict[str, Tuple[float, float]]] = None,
    rng: Optional[np.random.RandomState] = None,
) -> pd.Series:
    """
    Flipa binárias com probabilidade flip_prob; em contínuas aplica jitter ~ N(0, sigma*range).
    Clipa para o intervalo observado (ranges).
    """
    if rng is None:
        rng = np.random.RandomState(123)

    x = x0.copy()

    # binárias 0/1
    for c in binary_cols:
        if rng.rand() < flip_prob:
            v = x[c]
            if pd.isna(v):
                continue
            x[c] = 1 - int(v)

    # contínuas
    if ranges is None:
        ranges = {c: (float(x0[c]), float(x0[c])) for c in cont_cols}
    for c in cont_cols:
        lo, hi = ranges.get(c, (float(x[c]), float(x[c])))
        r = max(hi - lo, 1e-6)
        jitter = rng.normal(0.0, sigma * r)
        val = float(x[c].iloc[0]) + jitter
        # clipe para o intervalo observado
        val = min(max(val, lo), hi)
        x[c] = val

    return x

def eval_anchor_on_row(anchor_conditions: List[str], row_df: pd.DataFrame) -> bool:
    """Retorna True se a linha satisfaz TODAS as condições da âncora."""
    if not anchor_conditions:
        return True
    q = anchor_conditions_to_query(anchor_conditions, row_df.columns.tolist())
    try:
        mask = row_df.eval(q, engine="python").astype(bool)
        return bool(mask.iloc[0])
    except Exception:
        return False
    
# -----------------------------------------------------------
# Métrica de estabilidade para UMA instância
# -----------------------------------------------------------
from joblib import Parallel, delayed
import numpy as np
import pandas as pd
from typing import *
import warnings


# ============================================================
# 1) Compila âncora A(x) em predicado rápido (sem pandas / eval)
# ============================================================

def compile_anchor_predicate(anchor_literals: List[str], feature_names: List[str]):
    """
    Compila a âncora A(x) em uma função Python:
         f(vec_row: np.ndarray) -> bool
    Usando apenas comparações numéricas, sem pandas.eval.
    """
    if not anchor_literals:
        # âncora vazia => sempre verdadeiro
        return lambda v: True

    checks = []

    for lit in anchor_literals:
        s = (
            lit.replace("≤", "<=").replace("≥", ">=")
               .replace("==", "=").strip()
        )

        # intervalos e desigualdades simples (forma mais comum das regras)
        if "<=" in s:
            f, v = s.split("<=")
            f = f.strip()
            v = float(v.strip())
            idx = feature_names.index(f)
            checks.append((idx, "<=", v))
        elif ">=" in s:
            f, v = s.split(">=")
            f = f.strip()
            v = float(v.strip())
            idx = feature_names.index(f)
            checks.append((idx, ">=", v))
        elif "<" in s:
            f, v = s.split("<")
            f = f.strip()
            v = float(v.strip())
            idx = feature_names.index(f)
            checks.append((idx, "<", v))
        elif ">" in s:
            f, v = s.split(">")
            f = f.strip()
            v = float(v.strip())
            idx = feature_names.index(f)
            checks.append((idx, ">", v))
        elif "=" in s:
            f, v = s.split("=")
            f = f.strip()
            v = float(v.strip())
            idx = feature_names.index(f)
            checks.append((idx, "==", v))
        else:
            warnings.warn(f"[anchor compile] literal desconhecido: {lit}")

    def fast_predicate(vec_row: np.ndarray) -> bool:
        for idx, op, th in checks:
            x = vec_row[idx]
            if op == "<=" and not (x <= th): return False
            if op == ">=" and not (x >= th): return False
            if op == "<"  and not (x <  th): return False
            if op == ">"  and not (x >  th): return False
            if op == "==" and not (x == th): return False
        return True

    return fast_predicate


# ============================================================
# 2) Worker paralelo para uma única perturbação
# ============================================================

def _anchor_stability_worker(
    x0_vec: np.ndarray,
    explainer,
    anchor_A0: List[str],
    S0: set,
    prec0: float,
    cov0: float,
    feature_names: List[str],
    ranges: Dict[int, Tuple[float,float]],
    binary_idx: np.ndarray,
    cont_idx: np.ndarray,
    flip_prob: float,
    sigma: float,
    seed: int,
    anchor_kwargs: dict,
    representation: str,
    pred_A0,  # predicado compilado para membership
):
    rng = np.random.default_rng(seed)

    # ---- perturbação vetorizada
    xk = x0_vec.copy()

    # contínuas: ruído normal
    if len(cont_idx) > 0:
        noise = rng.normal(0.0, sigma, size=len(cont_idx))
        xk[cont_idx] += noise
        # clipe usando ranges
        for j, cind in enumerate(cont_idx):
            lo, hi = ranges[cind]
            xk[cind] = np.clip(xk[cind], lo, hi)

    # binárias: flips com prob flip_prob
    if len(binary_idx) > 0:
        flips = rng.random(len(binary_idx)) < flip_prob
        xk[binary_idx] = np.where(flips, 1 - xk[binary_idx], xk[binary_idx])

    # ---- nova âncora em xk
    exp_k = explainer.explain(xk, **anchor_kwargs)
    A_k = exp_k.anchor or []

    # representação em conjunto
    if representation == "literal":
        S_k = set(A_k)
    else:
        # mapeia literal -> nome da feature
        S_k = set(l.split()[0] for l in A_k)

    # Jaccard / add / del
    inter = len(S0 & S_k)
    union = len(S0 | S_k)
    jacc = 1.0 if union == 0 else inter / union

    add = 0.0 if union == 0 else len(S_k - S0) / union
    delete = 0.0 if union == 0 else len(S0 - S_k) / union
    exact = 1.0 if S0 == S_k else 0.0

    # membership: xk satisfaz A(x0)?
    memb = 1.0 if pred_A0(xk) else 0.0

    # drifts de precisão/cobertura (se disponíveis)
    pk = float(exp_k.precision) if (exp_k.precision is not None) else np.nan
    ck = float(exp_k.coverage)  if (exp_k.coverage  is not None) else np.nan

    dprec = abs(pk - prec0) if (not np.isnan(prec0) and not np.isnan(pk)) else np.nan
    dcov  = abs(ck - cov0)  if (not np.isnan(cov0)  and not np.isnan(ck)) else np.nan

    return jacc, add, delete, exact, memb, dprec, dcov


# ============================================================
# 3) Função principal OTIMIZADA (com binary_cols / cont_cols)
# ============================================================

def compute_anchor_stability_for_instance(
    *,
    explainer,                  # AnchorTabular já fitado
    X_df: pd.DataFrame,         # espaço das regras (ex.: original)
    feature_names: List[str],
    pos_id: int,              # pode ser posição OU label agora
    N: int = 5,                # número de perturbações
    binary_cols: Optional[List[str]] = None,
    cont_cols: Optional[List[str]] = None,
    flip_prob: float = 0.05,
    sigma: float = 0.05,
    representation: str = "literal",   # 'literal' (default) ou 'feature'
    seed_explain: Optional[int] = 13,  # fixa a seed do Anchor (reduz variação adicional)
    seed_perturb: int = 123,
    anchor_kwargs: Optional[dict] = None,  # ex.: {'threshold':0.95,'beam_size':4,'max_anchor_size':6}
    exp_x: Optional[object] = None,       # explicação já precomputada
) -> Dict[str, float]:

    if anchor_kwargs is None:
        anchor_kwargs = dict(threshold=0.95, beam_size=1, max_anchor_size=5, verbose=False)
    # garante seed fixo no Anchor (variação vem só da perturbação)
    anchor_kwargs = dict(anchor_kwargs)
    if seed_explain is not None:
        anchor_kwargs.setdefault("seed", seed_explain)

    # ------------------------------------------------
    # 1) Extrai matriz de features e linha alvo (x0)
    # ------------------------------------------------
    X_feat_df = X_df[feature_names]
    X_feat = X_feat_df.to_numpy()
    n_feat = X_feat.shape[1]

    # >>> AJUSTE: usar ID real (pos_id) com o helper _get_row_and_index,
    # em vez de tratar pos_id como posição iloc
    x_row_df, idx_label = _get_row_and_index(X_df, pos_id)
    if isinstance(x_row_df, pd.Series):
        x_row_df = x_row_df.to_frame().T

    # série 1D com as MESMAS features usadas no Anchor/modelo
    x0_ser = x_row_df[feature_names].iloc[0]
    x0_vec = x0_ser.to_numpy()

    # ------------------------------------------------
    # 2) Âncora original em x0 (reaproveitando exp_x se vier)
    # ------------------------------------------------
    if exp_x is not None:
        exp0 = exp_x
    else:
        exp0 = explainer.explain(x0_vec, **anchor_kwargs)

    A0 = exp0.anchor or []

    if representation == "literal":
        S0 = set(A0)
    else:
        S0 = set(l.split()[0] for l in A0)

    prec0 = float(exp0.precision) if (exp0.precision is not None) else np.nan
    cov0  = float(exp0.coverage)  if (exp0.coverage  is not None) else np.nan

    # predicado rápido para membership A(x0)
    pred_A0 = compile_anchor_predicate(A0, feature_names)

    # ------------------------------------------------
    # 3) Mapeamento de binary_cols / cont_cols -> índices
    # ------------------------------------------------
    all_idx = np.arange(n_feat)

    if binary_cols is not None:
        binary_idx = np.array(
            [feature_names.index(c) for c in binary_cols if c in feature_names],
            dtype=int
        )
    else:
        # inferir binárias por cardinalidade
        uniq = [np.unique(X_feat[:, i]).size for i in range(n_feat)]
        binary_idx = np.array([i for i, u in enumerate(uniq) if u <= 2], dtype=int)

    if cont_cols is not None:
        cont_idx = np.array(
            [feature_names.index(c) for c in cont_cols if c in feature_names],
            dtype=int
        )
    else:
        cont_idx = np.array([i for i in all_idx if i not in set(binary_idx)], dtype=int)

    # ------------------------------------------------
    # 4) Ranges das contínuas (para clipe)
    # ------------------------------------------------
    ranges = {
        int(i): (float(X_feat[:, i].min()), float(X_feat[:, i].max()))
        for i in cont_idx
    }

    # ------------------------------------------------
    # 5) Loop sequencial das N perturbações (sem paralelismo)
    # ------------------------------------------------
    seeds = [seed_perturb + 9973 * k for k in range(N)]

    results_list = []
    for s in seeds:
        xstart = time.strftime('%Y-%m-%d %H:%M:%S')
        res = _anchor_stability_worker(
            x0_vec,
            explainer,
            A0,
            S0,
            prec0,
            cov0,
            feature_names,
            ranges,
            binary_idx,
            cont_idx,
            flip_prob,
            sigma,
            s,
            anchor_kwargs,
            representation,
            pred_A0,    
        )
        print(f"[STABILITY_anchor_stability_worker] pos={pos_id} perturbation with seed={s}:Tempo : {time.strftime('%Y-%m-%d %H:%M:%S')} - {xstart}")
        results_list.append(res)

    results = np.array(results_list, dtype=object)

    jaccs = results[:, 0].astype(float)
    adds  = results[:, 1].astype(float)
    dels  = results[:, 2].astype(float)
    exact = results[:, 3].astype(float)
    membs = results[:, 4].astype(float)
    dprec = results[:, 5].astype(float)
    dcov  = results[:, 6].astype(float)

    jacc_mean = float(np.nanmean(jaccs)) if len(jaccs) else np.nan
    add_mean  = float(np.nanmean(adds))  if len(adds)  else np.nan
    del_mean  = float(np.nanmean(dels))  if len(dels) else np.nan
    exact_rt  = float(np.nanmean(exact)) if len(exact) else np.nan
    memb_rt   = float(np.nanmean(membs)) if len(membs) else np.nan
    dprec_m   = float(np.nanmean(dprec)) if len(dprec) else np.nan
    dcov_m    = float(np.nanmean(dcov))  if len(dcov)  else np.nan

    return {
        "pos_id": int(pos_id),
        "idx": idx_label,
        "anchor_orig": " AND ".join(A0) if A0 else "<regra vazia>",
        "n_literals_orig": int(len(S0)),
        "precision_orig": prec0,
        "coverage_orig": cov0,
        # principais:
        "stability_abs": jacc_mean,
        "stability_percent": 100.0 * jacc_mean if np.isfinite(jacc_mean) else np.nan,
        # diagnósticos:
        "add_rate_mean": add_mean,
        "del_rate_mean": del_mean,
        "exact_match_rate": exact_rt,
        "membership_rate": memb_rt,
        "d_precision_mean": dprec_m,
        "d_coverage_mean": dcov_m,
        # parâmetros usados:
        "N": int(N),
        "representation": representation,
        "flip_prob": float(flip_prob),
        "sigma": float(sigma),
    }


def support_on_df(anchor_conditions, df: pd.DataFrame):
    """
    Coverage global: % de linhas de df que satisfazem TODAS as condições.
    """
    if not anchor_conditions:
        return 1.0, pd.Series(True, index=df.index)
    query = anchor_conditions_to_query(anchor_conditions, df.columns.tolist())
    try:
        mask = df.eval(query, engine="python").astype(bool)
    except Exception as e:
        print(f"[WARN] Falha ao avaliar regra no DataFrame: {e}\nRegra: {anchor_conditions}\nQuery: {query}")
        mask = pd.Series(False, index=df.index)
    return float(mask.mean()), mask

def _wrap_feat_names(expr: str, feature_names: List[str]) -> str:
    expr = expr.replace('≤', '<=').replace('≥', '>=')
    expr = re.sub(r'(?<![<>!=])=(?!=)', '==', expr)  # "=" -> "=="
    expr = expr.replace(' AND ', ' and ')
    for name in sorted(feature_names, key=len, reverse=True):
        expr = re.sub(rf'\b{re.escape(name)}\b', f'`{name}`', expr)
    return expr


def anchor_conditions_to_query(anchor_conditions: List[str], feature_names: List[str]) -> str:
    if not anchor_conditions:
        return ""
    parts = []
    for cond in anchor_conditions:
        parts.append(_wrap_feat_names(cond.strip(), feature_names))
    return " and ".join(parts)
```

```python
# ===================== contribuições Anchor + Tabela de métricas (instância / média / mediana) =====================
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")

# --------------------------------- utils básicos ---------------------------------
def _coerce_binary_series(s):
    s = pd.Series(s).copy()
    map_true  = {"1", 1, True, "true", "True", "sim", "Sim", "yes", "Yes", "y", "Y", "positivo", "Positivo"}
    map_false = {"0", 0, False, "false", "False", "nao", "não", "Nao", "Não", "no", "No", "n", "N", "negativo", "Negativo"}
    def _m(v):
        sv = str(v).strip()
        if sv in map_true:  return 1
        if sv in map_false: return 0
        return v
    s = s.map(_m)
    s = pd.to_numeric(s, errors="coerce")
    return s

THRESHOLD = 0.5  # caso derive y a partir de probabilidade

TOP_K_LIME = 10  # número de features a exibir no plot

def _infer_y_from_X(Xdf, threshold=0.5):
    """
    Extrai y de colunas em X:
      (1) alvo explícito (y/target/label/classe/...),
      (2) probabilidade única (y_proba/proba_1/score_pos/...),
      (3) fallback: coluna ~[0,1] vira proba->y (>= threshold).
    Retorna (y_series, cols_to_drop).
    """
    TARGET_CANDIDATES = ['y','classe','target','label','y_true','diagnosis','outcome']
    PROBA_CANDIDATES  = ['y_proba','proba','proba_1','p1','prob_1','prob_pos','score','score_pos','pred_proba_1']

    # 1) alvo explícito
    for col in TARGET_CANDIDATES:
        if col in Xdf.columns:
            y_raw = Xdf[col]
            y_bin = _coerce_binary_series(y_raw).rename('y')
            return y_bin, [col]

    # 2) probabilidade em única coluna
    for col in PROBA_CANDIDATES:
        if col in Xdf.columns:
            y_bin = (pd.to_numeric(Xdf[col], errors='coerce') >= threshold).astype('Int64').rename('y')
            return y_bin, [col]

    # 3) fallback: heurística p/ coluna [0,1]
    for col in Xdf.columns:
        s = pd.to_numeric(Xdf[col], errors='coerce')
        if s.notna().mean() > 0.9 and (s.between(0, 1, inclusive='both').mean() > 0.9):
            y_bin = (s >= threshold).astype('Int64').rename('y')
            return y_bin, [col]

    raise ValueError("Não encontrei y nem y_proba dentro de X para alinhar o alvo.")

# --------------------------------- preparo do X (remove alvos e probas) ---------------------------------
try:
    EXCLUDE_COLUMNS
except NameError:
    EXCLUDE_COLUMNS = []

y_inferido, y_cols_to_drop = _infer_y_from_X(X)

TARGET_LIKE = ['y','classe','target','label','y_true','diagnosis','outcome']
PROBA_LIKE  = ['y_proba','proba','proba_0','proba_1','prob_0','prob_1','p0','p1',
               'score','score_pos','score_neg','pred_proba_0','pred_proba_1']

cols_drop   = [c for c in (EXCLUDE_COLUMNS + TARGET_LIKE + PROBA_LIKE + y_cols_to_drop) if c in X.columns]
X_feat_full = X.drop(columns=list(set(cols_drop)), errors='ignore').copy()

# --- alinhamento com o modelo: mesmas colunas e ordem que o modelo usou no treino ---
if hasattr(model, "feature_names_in_"):
    feat_cols_eff = [c for c in model.feature_names_in_.tolist() if c in X_feat_full.columns]
    missing = [c for c in model.feature_names_in_.tolist() if c not in X_feat_full.columns]
    if len(missing) > 0:
        raise ValueError(f"As seguintes colunas esperadas pelo modelo não estão em X: {missing}")
    X_feat = X_feat_full.loc[:, feat_cols_eff].copy()
else:
    if hasattr(model, "n_features_in_"):
        n_expected = int(model.n_features_in_)
        if X_feat_full.shape[1] > n_expected:
            X_feat = X_feat_full.iloc[:, :n_expected].copy()
        elif X_feat_full.shape[1] == n_expected:
            X_feat = X_feat_full.copy()
        else:
            raise ValueError(
                f"X tem {X_feat_full.shape[1]} features, mas o modelo espera {n_expected}. "
                f"Adicione as colunas faltantes ou informe feature_names_in_."            )
    else:
        X_feat = X_feat_full.copy()

# y alinhado
y_alinhado = pd.Series(y_inferido).reindex(X_feat.index).astype('Int64').rename('y')

# --------------------------------- split 8/1/1 consistente ---------------------------------
# Se já vieram do pipeline 8/1/1, apenas reutilize e garanta as colunas corretas:
if all(name in globals() for name in ["X_train", "X_test", "y_train", "y_test"]):
    X_train = X_train.loc[:, X_feat.columns]
    X_test  = X_test.loc[:,  X_feat.columns]
    X_train_real, X_test_real = X_train, X_test
else:
    # Reconstrói 8/1/1 localmente (estratificado)
    # 1) separa 10% para teste
    idx_trval, idx_tst = train_test_split(
        X_feat.index, test_size=0.10, stratify=y_alinhado, random_state=42
    )
    # 2) dos 90% restantes, separa 1/9 (~10%) para validação => sobra ~80% treino
    y_trval = y_alinhado.loc[idx_trval]
    idx_tr, idx_val = train_test_split(
        idx_trval, test_size=(1/9), stratify=y_trval, random_state=42
    )

    X_train = X_feat.loc[idx_tr].copy()          # ~80%
    y_train = y_alinhado.loc[idx_tr].copy()
    X_val   = X_feat.loc[idx_val].copy()         # ~10% (não usado aqui, mas preservado)
    y_val   = y_alinhado.loc[idx_val].copy()
    X_test  = X_feat.loc[idx_tst].copy()         # ~10%
    y_test  = y_alinhado.loc[idx_tst].copy()

    X_train_real, X_test_real = X_train, X_test

# --------------------------------- nomes de classes ---------------------------------
unique_classes = sorted(pd.unique(y_alinhado.dropna()))
if len(unique_classes) == 2 and set(unique_classes) <= {0,1}:
    class_names = ['Risco Baixo','Risco Alto']
else:
    class_names = [f'Classe {c}' for c in unique_classes]

# --------------------------------- métricas: helpers ---------------------------------
METRIC_COLS = [
    'fidelity(%)','infidelity(%)','faithfulness(%)','simplicity(%)','consistency(%)',
    'robustez(%)','stability(%)','selectivity(%)','soundness(%)','directional soundness(%)',
    'sufficiency-1(%)','sufficiency-5(%)','sufficiency-10(%)','sufficiency-20(%)','completeness(%)'
]

_num_regex = re.compile(r"[-+]?\d*\.?\d+(?:[eE][-+]?\d+)?")

def _to_float_or_nan(x):
    if pd.isna(x): return np.nan
    if isinstance(x,(int,float,np.number)): return float(x)
    s = str(x)
    m = _num_regex.search(s)
    return float(m.group(0)) if m else np.nan

def get_metrics_instance_mean_median(df_metrics, inst_idx):
    row_inst = df_metrics.loc[df_metrics['instancia'] == inst_idx]
    row_mean = df_metrics.loc[df_metrics['instancia'].astype(str).str.upper().eq('MÉDIA')]
    row_medi = df_metrics.loc[df_metrics['instancia'].astype(str).str.upper().eq('MEDIANA')]

    if row_mean.empty or row_medi.empty:
        tmp = df_metrics.copy()
        for c in METRIC_COLS:
            if c in tmp.columns:
                tmp[c] = tmp[c].map(_to_float_or_nan)
        only_num = tmp.loc[pd.to_numeric(tmp['instancia'], errors='coerce').notna(), METRIC_COLS]
        mean_vals = only_num.mean(numeric_only=True)
        medi_vals = only_num.median(numeric_only=True)
    else:
        mean_vals = row_mean.iloc[0][METRIC_COLS].map(_to_float_or_nan)
        medi_vals = row_medi.iloc[0][METRIC_COLS].map(_to_float_or_nan)

    inst_vals = pd.Series({c: np.nan for c in METRIC_COLS})
    if not row_inst.empty:
        inst_vals = row_inst.iloc[0][METRIC_COLS].map(_to_float_or_nan)

    out = pd.DataFrame({
        "metric": METRIC_COLS,
        "inst":  [inst_vals.get(c, np.nan) for c in METRIC_COLS],
        "mean":  [mean_vals.get(c, np.nan) for c in METRIC_COLS],
        "median":[medi_vals.get(c, np.nan) for c in METRIC_COLS],
    })
    return out


# ====== Função para desenhar a árvore Anchor (reaproveitada do código Anchor) ======
def draw_decision_tree(ax, conditions, prediction, precisao=None, cobertura=None):
    n = len(conditions)
    h_spacing = 1.5
    v_spacing = -1.5
    x = 0.0
    y = 0.0

    for i, cond in enumerate(conditions):
        circ = plt.Circle((x, y), 0.3, ec='black', fc="lightblue")
        ax.add_patch(circ)
        ax.text(x, y, cond, fontsize=9, ha='center', va='center', wrap=True)
        if i < n - 1:
            x_left = x - h_spacing / (2 ** i)
            y_next = y + v_spacing
            ax.plot([x, x_left], [y, y_next], 'k-')
            x, y = x_left, y_next

    y_final = y + v_spacing
    ax.plot([x, x], [y, y_final], 'k-')

    circ_leaf = plt.Circle((x, y_final), 0.3, ec='black', fc="lightgreen")
    ax.add_patch(circ_leaf)
    ax.text(x, y_final, prediction.upper(), fontsize=10, ha='center', va='center')

    if (precisao is not None) and (cobertura is not None):
        ax.set_title(f"Prec: {precisao:.2f} | Cob: {cobertura:.2f}", fontsize=10)

    ax.set_xlim(-3, 3)
    ax.set_ylim(y_final - 1, 1.5)
    ax.axis("off")


from sklearn.inspection import permutation_importance

def plot_anchor_with_real_values_and_metrics(inst_id, df_metrics, n_permut=30):
    """
    AGORA: Usa Anchor + métricas, mantendo layout geral.

    inst_id: valor da coluna 'id' (chave externa)
    df_metrics: DataFrame com métricas por instância (coluna 'instancia' com id externo)
    n_permut: (IGNORADO) mantido apenas para compatibilidade
    """
    # --- checar se existe coluna id ---
    if 'id' not in X.columns:
        raise KeyError("Coluna 'id' não encontrada em X.")

    # --- localizar a linha correspondente ao inst_id ---
    mask = (X['id'] == inst_id)
    n_rows = mask.sum()

    if n_rows == 0:
        raise ValueError(f"inst_id={inst_id} não encontrado em X['id'].")
    if n_rows > 1:
        raise ValueError(f"inst_id={inst_id} aparece {n_rows} vezes em X['id'] (esperado 1).")

    # DataFrame 1xD com TODAS as colunas dessa instância
    x_row_full = X.loc[mask].copy()          # shape (1, D)
    idx_label  = x_row_full.index[0]

    # Somente features (remoção das colunas excluídas)
    x_row_feats = x_row_full.drop(columns=EXCLUDE_COLS, errors='ignore')
    feature_cols = list(x_row_feats.columns)

    # espaço de referência (todas as linhas, mesmas features)
    X_ref_feats = X.drop(columns=EXCLUDE_COLS, errors='ignore')

    # --- probabilidade original da classe predita (da instância fixa) ---
    probs0 = model.predict_proba(x_row_feats)[0]
    classes = list(model.classes_)
    pred_idx = int(np.argmax(probs0))
    p0 = float(probs0[pred_idx])

    # prob. de classe 0 (Risco Baixo) e 1 (Risco Alto), se existirem
    if 0 in classes:
        p_neg = float(probs0[classes.index(0)])
    else:
        p_neg = float(probs0.min())

    if 1 in classes:
        p_pos = float(probs0[classes.index(1)])
    else:
        p_pos = float(probs0.max())

    # <<< NOVO: resgatar a explicação Anchor já calculada (df_anchor global) >>>
    if 'df_anchor' not in globals():
        raise RuntimeError("df_anchor não encontrado. Execute antes o código de cálculo do Anchor.")

    row_anchor = df_anchor.loc[df_anchor['idx_instancia'] == inst_id]
    if row_anchor.empty:
        raise ValueError(f"Instância id={inst_id} não encontrada em df_anchor['idx_instancia'].")

    regra_anchor     = row_anchor.iloc[0]['regra_anchor']
    pred_anchor      = row_anchor.iloc[0]['predicao']
    precisao_anchor  = float(row_anchor.iloc[0]['precision_anchor'])
    cobertura_anchor = float(row_anchor.iloc[0]['cobertura_anchor'])

    if isinstance(regra_anchor, str) and regra_anchor.strip():
        conditions = regra_anchor.split(" AND ")
    else:
        conditions = ["<sem condição>"]

    # <<< NOVO: valores reais da instância para as features >>>
    row_real = x_row_feats.iloc[0]

    # Preparar variáveis para o plot (mantendo lógica original, mas sem PFI)
    TOP_K = 15
    feat_names = feature_cols
    labels_txt = feat_names[:TOP_K]
    feat_idx_order = list(range(len(labels_txt)))
    real_values = [row_real.get(f, np.nan) for f in labels_txt]
    m = len(labels_txt)

    # --- métricas por id externo ---
    mt = get_metrics_instance_mean_median(df_metrics, inst_id)

    # --- plot (mantive todo o layout original, só troquei o painel esquerdo para Anchor) ---
    fig = plt.figure(figsize=(12.5, 6.0), dpi=120)
    gs = fig.add_gridspec(1, 3, width_ratios=[3.0, 2.0, 2.2], wspace=0.24)

    fig.suptitle(
        f"Explicações – Instância id: {inst_id}",
        fontsize=22,
        fontweight="bold",
        x=0.02,
        ha="left",
        y=0.96
    )

    txt_p_neg = f"{p_neg:.2f}".replace(".", ",")
    txt_p_pos = f"{p_pos:.2f}".replace(".", ",")

    fig.text(0.60, 0.94, "Risco Baixo",
             fontsize=15, fontweight="bold",
             color="#1f77ff", ha="center", va="bottom")
    fig.text(0.60, 0.89, txt_p_neg,
             fontsize=20, fontweight="bold",
             color="#1f77ff", ha="center", va="top")

    fig.text(0.82, 0.94, "Risco Alto",
             fontsize=15, fontweight="bold",
             color="#ff7f0e", ha="center", va="bottom")
    fig.text(0.82, 0.89, txt_p_pos,
             fontsize=20, fontweight="bold",
             color="#ff7f0e", ha="center", va="top")

    # 1) PAINEL ESQUERDO: ÁRVORE ANCHOR (substituindo o plot de PFI)
    ax = fig.add_subplot(gs[0, 0])
    ax.set_title("Regra Anchor", fontsize=13, pad=10)
    draw_decision_tree(ax, conditions, pred_anchor, precisao_anchor, cobertura_anchor)

    # 2) Tabela de valores reais (coluna central) – mantida
    axr = fig.add_subplot(gs[0, 1])
    axr.axis("off")

    axr.text(0.00, 0.98, "Feature", fontsize=11, fontweight="bold",
             ha="left", va="top")
    axr.text(0.88, 0.98, "Valor (real)", fontsize=11, fontweight="bold",
             ha="right", va="top")

    def _fmt_val(v):
        try:
            return f"{float(v):.2f}".replace(".", ",")
        except Exception:
            return str(v)

    y0, dy = 0.92, 0.055
    for k, idx_f in enumerate(feat_idx_order):
        if y0 - k*dy < 0.05:
            break
        fname = feat_names[idx_f] if idx_f < len(feat_names) else labels_txt[k]
        fval = real_values[k]

        axr.text(0.00, y0 - k*dy, fname, ha="left", va="top", fontsize=9)
        axr.text(0.88, y0 - k*dy, _fmt_val(fval), ha="right", va="top", fontsize=9)

    # 3) Tabela de métricas (coluna direita) – mantida
    axm = fig.add_subplot(gs[0, 2])
    axm.axis("off")

    from matplotlib.lines import Line2D

    bbox_val = axr.get_position()
    bbox_met = axm.get_position()
    x_sep = (bbox_val.x1 + bbox_met.x0) / 2.0

    linha_sep = Line2D(
        [x_sep, x_sep],
        [0.10, 0.90],
        transform=fig.transFigure,
        color="#cccccc",
        linewidth=1.0,
        alpha=0.8
    )
    fig.add_artist(linha_sep)

    x_feat   = 0.00
    x_inst   = 1.00
    x_media  = 1.38
    x_mediana= 1.78

    axm.text(x_feat,    0.98, "Métrica",   fontsize=11, fontweight="bold", ha="left",  va="top")
    axm.text(x_inst,    0.98, "Valor",     fontsize=11, fontweight="bold", ha="right", va="top")
    axm.text(x_media,   0.98, "Média",     fontsize=11, fontweight="bold", ha="right", va="top")
    axm.text(x_mediana, 0.98, "Mediana",   fontsize=11, fontweight="bold", ha="right", va="top")

    def _fmt_metric(v):
        return "0,00" if pd.isna(v) else f"{float(v):.3f}".replace(".", ",")

    y0, dy = 0.92, 0.055
    for i, row in mt.iterrows():
        if y0 - i*dy < 0.05:
            break

        axm.text(x_feat,  y0 - i*dy, row['metric'],               ha="left",  va="top", fontsize=9)
        axm.text(x_inst,  y0 - i*dy, _fmt_metric(row['inst']),    ha="right", va="top", fontsize=9)
        axm.text(x_media, y0 - i*dy, _fmt_metric(row['mean']),    ha="right", va="top", fontsize=9)
        axm.text(x_mediana, y0 - i*dy, _fmt_metric(row['median']),ha="right", va="top", fontsize=9)

    plt.subplots_adjust(top=0.82)
    plt.show()



# ----------------- rodar para as suas instâncias -----------------
try:
    ENTRADAS_SELECIONADAS
except NameError:
    ENTRADAS_SELECIONADAS = [0]

ENTRADAS_SELECIONADAS = np.asarray(list(ENTRADAS_SELECIONADAS), dtype=int)

# IMPORTANTE: agora usa df_anchor_metrics em vez de df_permutation_metrics
for idx_pos in ENTRADAS_SELECIONADAS:
    plot_anchor_with_real_values_and_metrics(int(idx_pos), df_anchor_metrics, 30)
```

```python
# === Anchor — Dendrograma de Métricas ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform

# --------- Parâmetros ---------
USE_ABS_CORR = True                # True => distância = 1 - |corr| ; False => 1 - corr
DROP_COLS_EXTRA = ["class_consistency()"]  # se existir e não quiser incluir
SAVE_PATH = "anchor_dendrogram_metricas.png"

# --------- 1) Preparar matriz de correlação a partir do df_anchor_metrics ---------
df_anchor_metrics_filter = df_anchor_metrics.copy()

# remover linhas agregadas se existirem
if "instancia" in df_anchor_metrics_filter.columns:
    df_anchor_metrics_filter = df_anchor_metrics_filter[
        ~df_anchor_metrics_filter["instancia"].isin(["MÉDIA", "MEDIANA"])
    ]

# remover colunas extra indesejadas
df_anchor_metrics_filter = df_anchor_metrics_filter.drop(
    columns=[c for c in DROP_COLS_EXTRA if c in df_anchor_metrics_filter.columns],
    errors="ignore"
)

# remover colunas claramente constantes em "0%" ou "100%"
cols_const = []
for c in df_anchor_metrics_filter.columns:
    if c == "instancia":
        continue
    s = df_anchor_metrics_filter[c].astype(str).str.replace("%", "", regex=False).str.strip()
    s = s[s != ""]
    if len(s) > 0 and s.nunique() == 1 and s.iloc[0] in {"0", "0.0", "100", "100.0"}:
        cols_const.append(c)

df_anchor_metrics_filter = df_anchor_metrics_filter.drop(
    columns=cols_const + (["instancia"] if "instancia" in df_anchor_metrics_filter.columns else []),
    errors="ignore"
)

# converter "xx%" -> numérico
for c in df_anchor_metrics_filter.columns:
    df_anchor_metrics_filter[c] = pd.to_numeric(
        df_anchor_metrics_filter[c].astype(str).replace("%", "", regex=False),
        errors="coerce"
    )

# remover colunas sem variância (evita NaNs na correlação)
var0_cols = [c for c in df_anchor_metrics_filter.columns if df_anchor_metrics_filter[c].nunique(dropna=True) <= 1]
df_anchor_metrics_filter = df_anchor_metrics_filter.drop(columns=var0_cols, errors="ignore")

# checagem mínima
if df_anchor_metrics_filter.shape[1] < 2:
    raise ValueError("Após os filtros, restaram menos de 2 métricas numéricas para correlacionar.")

# correlação de Pearson
# --------- 2) Dendrograma com distância baseada em correlação ---------
C = corr_matrix.to_numpy()

# --- correção 1: limpar NaN (vira 0 => distância = 1) ---
C = np.nan_to_num(C, nan=0.0)

# --- correção 2: limitar correlações no intervalo [-1, 1] ---
C = np.clip(C, -1.0, 1.0)

# --- correção 3: usar correlação absoluta se pedido ---
C = np.abs(C) if USE_ABS_CORR else C

# --- garantir simetria perfeita (pode ter erros numéricos pequenos) ---
C = (C + C.T) / 2

# --- distância (1 - correlação) ---
D = 1.0 - C

# --- correção 4: distância não pode ser negativa ---
D = np.clip(D, 0.0, None)

# --- diagonal precisa ser zero ---
np.fill_diagonal(D, 0.0)

# --- condensed matrix ---
condensed_D = squareform(D, checks=False)

# linkage
Z = hierarchy.linkage(condensed_D, method='average')

# dendrograma
plt.figure(figsize=(12, 6))
dn = hierarchy.dendrogram(
    Z,
    labels=corr_matrix.columns.tolist(),
    leaf_rotation=90
)
plt.title(f"Anchor — Dendrograma das Métricas (distância = 1 - {'|corr|' if USE_ABS_CORR else 'corr'})")
plt.ylabel("Distância")
plt.tight_layout()
plt.savefig(f'{BASE_PATH}/{SAVE_PATH}', dpi=150)
plt.show()
print(f"✅ Dendrograma salvo em: {SAVE_PATH}")

# ordem das métricas segundo a árvore (útil para reordenar heatmaps)
order_anchor = [corr_matrix.columns[i] for i in dn["leaves"]]
print("Ordem sugerida das métricas:", order_anchor)
```

```python
# === Anchor — Venn adaptado a partir da correlação das métricas ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from scipy.spatial.distance import squareform

# -------------------------------------------------
# 1) Construir corr_matrix a partir de df_anchor_metrics
# -------------------------------------------------
def build_corr_from_anchor_metrics(df_anchor_metrics: pd.DataFrame) -> pd.DataFrame:
    df = df_anchor_metrics.copy()

    # remover linhas agregadas, se existirem
    if "instancia" in df.columns:
        df = df[~df["instancia"].isin(["MÉDIA", "MEDIANA"])].copy()

    # remover coluna de id, se presente
    df = df.drop(columns=["instancia"], errors="ignore")

    # converter "xx%" -> número
    for c in df.columns:
        df[c] = pd.to_numeric(
            df[c].astype(str).str.replace("%", "", regex=False).str.strip(),
            errors="coerce"
        )

    # remover colunas sem variação (constantes ou todas NaN)
    nun = df.nunique(dropna=True)
    keep_cols = nun.index[(nun > 1) & (df.notna().sum() > 1)]
    df = df[keep_cols].copy()

    if df.shape[1] < 2:
        raise ValueError("Após limpeza, há menos de 2 métricas numéricas para correlacionar.")

    # correlação de Pearson
    corr = df.corr(method="pearson")
    return corr

# constrói a matriz de correlação das métricas do Anchor
corr_matrix = build_corr_from_anchor_metrics(df_anchor_metrics)

# -------------------------------------------------
# 2) Sua função de Venn (inalterada na lógica, apenas título/nomes)
# -------------------------------------------------
def venn_from_corr_mds_labels(
    corr_matrix,
    radius=1.0,               # raio (mesmo para todos)
    epsilon_frac=0.06,        # ε aplicado SÓ quando r>0: d <- d + ε*R
    sign_mode="both",         # "both" | "positive" | "negative"
    use_abs=False,            # se True, ignora sinais (sign_mode fica sem efeito)
    random_state=42,
    n_init=8,
    max_iter=600,
    figsize=(8,8),
    face_alpha=0.30,
    edge_lw=1.5,
    label_fontsize=8,         # fonte 8
    label_weight='normal',    # sem negrito
    label_jitter=0.03,        # jitter vertical inicial (fração do raio)
    repel_iters=150,          # iterações de repulsão dos rótulos
    repel_step=0.02,          # passo da repulsão (fração do raio)
    inside_margin=0.12,       # margem p/ manter texto dentro do círculo
    title="Anchor — Venn adaptado (d = (1 − r)·R, ε>0 só para r>0, filtro de sinal)"
):
    """
    Distância alvo entre centros: d(r) = (1 − r)*R  (+ ε*R se r>0).
    - sign_mode:
        * "both": usa r como está;
        * "positive": se r<=0, força r_eff=-1 (distância máxima 2R);
        * "negative": se r>=0, força r_eff=-1 (distância máxima 2R).
    - use_abs=True ignora sinais e, portanto, sign_mode não tem efeito.
    """

    # ---------- 1) correlação -> distância alvo (com epsilon e filtro de sinal) ----------
    C_raw = corr_matrix.to_numpy(dtype=float)
    C_raw = np.clip(C_raw, -1.0, 1.0)
    n = C_raw.shape[0]

    # máscara de r>0 (para epsilon)
    mask_pos = (C_raw > 0)

    if use_abs:
        C_eff = np.abs(C_raw)
    else:
        # aplica filtro de sinal
        if sign_mode == "positive":
            # mantém só r>0; demais vão para -1 (distância máxima 2R)
            C_eff = np.where(C_raw > 0, C_raw, -1.0)
        elif sign_mode == "negative":
            # mantém só r<0; demais vão para -1
            C_eff = np.where(C_raw < 0, C_raw, -1.0)
        else:
            C_eff = C_raw

    R = float(radius)
    T = (1.0 - C_eff) * R  # d(r) base
    # aplica ε apenas para r>0 (com base no sinal original, como pedido)
    if epsilon_frac and epsilon_frac > 0:
        T = T + (epsilon_frac * R) * mask_pos.astype(float)

    # simetriza e zera diagonal
    T = 0.5 * (T + T.T)
    np.fill_diagonal(T, 0.0)

    # ---------- 2) embedding por MDS (SMACOF; fallback p/ MDS clássico) ----------
    try:
        from sklearn.manifold import MDS
        mds = MDS(
            n_components=2, metric=True, dissimilarity='precomputed',
            n_init=n_init, max_iter=max_iter, random_state=random_state
        )
        X = mds.fit_transform(T)
    except Exception:
        # MDS clássico (Torgerson)
        J = np.eye(n) - np.ones((n, n))/n
        B = -0.5 * J @ (T**2) @ J
        eigvals, eigvecs = np.linalg.eigh(B)
        idx = np.argsort(eigvals)[::-1]
        L = np.clip(eigvals[idx][:2], 0, None)
        V = eigvecs[:, idx][:, :2]
        X = V * np.sqrt(L + 1e-12)

    # reescala global para casar melhor T
    DX = np.sqrt(((X[:,None,:] - X[None,:,:])**2).sum(axis=2))
    iu = np.triu_indices_from(DX, k=1)
    num = np.sum(DX[iu] * T[iu])
    den = np.sum(DX[iu]**2) + 1e-12
    s = num/den if den > 0 else 1.0
    X *= s
    X -= X.mean(axis=0, keepdims=True)  # centraliza

    labels = corr_matrix.columns.tolist()

    # ---------- 3) paleta de cores (fixa para métricas comuns de Anchor; fallback tab20) ----------
    def _lc(s): return str(s).strip().lower()
    color_fixed = {
        "robustness":            "#9467bd",
        "stability":             "#17becf",
        "soundness":             "#f7b6d2",
        "directional soundness": "#ff9896",
        "selectivity":           "#c7c7c7",
        "fidelity":              "#2ca02c",
        "infidelity":            "#d62728",
        "coverage":              "#1f77b4",
        "completeness":          "#1f77b4"
    }
    cmap = plt.get_cmap("tab20")
    colors, k_auto = [], 0
    for name in labels:
        key = _lc(name)
        colors.append(color_fixed.get(key, cmap(k_auto % 20)))
        if key not in color_fixed: k_auto += 1

    # ---------- 4) desenha círculos ----------
    fig, ax = plt.subplots(figsize=figsize, dpi=150, facecolor="white")
    for (x, y), col in zip(X, colors):
        circ = Circle((x, y), R, facecolor=col, edgecolor=col,
                      alpha=face_alpha, lw=edge_lw)
        ax.add_patch(circ)

    # ---------- 5) rótulos centralizados + repulsão vertical ----------
    rng = np.random.default_rng(random_state)
    text_xy = X.copy()
    text_xy[:, 1] += (rng.uniform(-0.03, 0.03, size=len(labels))) * R  # jitter pequeno

    def clamp_inside(i):
        cx, cy = X[i]
        tx, ty = text_xy[i]
        vx, vy = tx - cx, ty - cy
        dist = np.hypot(vx, vy)
        maxd = R * (1.0 - inside_margin)
        if dist > maxd and dist > 1e-12:
            k = maxd / dist
            text_xy[i] = np.array([cx + k*vx, cy + k*vy])

    for _ in range(150):
        moved = False
        for i in range(len(labels)):
            for j in range(i+1, len(labels)):
                xi, yi = text_xy[i]
                xj, yj = text_xy[j]
                # caixa aproximada do texto
                w = 0.9 * R
                h = 0.35 * R
                if (abs(xi - xj) < w) and (abs(yi - yj) < h):
                    dy = 0.02 * R
                    text_xy[i, 1] += dy
                    text_xy[j, 1] -= dy
                    clamp_inside(i)
                    clamp_inside(j)
                    moved = True
        if not moved:
            break

    for (tx, ty), name in zip(text_xy, labels):
        ax.text(tx, ty, name, ha="center", va="center",
                fontsize=8, fontweight='normal',
                color="black", zorder=3)

    # limites e estética
    xmin = X[:,0].min() - R*1.3
    xmax = X[:,0].max() + R*1.3
    ymin = X[:,1].min() - R*1.3
    ymax = X[:,1].max() + R*1.3
    ax.set_aspect("equal")
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    ax.axis("off")
    ax.set_title(title, fontsize=10)
    plt.tight_layout()
    plt.show()

# -------------------------------------------------
# 3) Exemplos de uso (escolha um)
# -------------------------------------------------
# a) Usando ambos os sinais, com pequeno epsilon
# venn_from_corr_mds_labels(corr_matrix, radius=1.0, epsilon_frac=0.5, sign_mode="both", use_abs=False)

# b) Só correlações positivas (negativas viram distância máxima)
venn_from_corr_mds_labels(corr_matrix, radius=1.0, epsilon_frac=0.5, sign_mode="positive", use_abs=False)

# c) Só correlações negativas (positivas viram distância máxima)
# venn_from_corr_mds_labels(corr_matrix, radius=1.0, epsilon_frac=0.5, sign_mode="negative", use_abs=False)

# d) Ignorar o sinal (agrupa por força de correlação apenas)
# venn_from_corr_mds_labels(corr_matrix, radius=1.0, epsilon_frac=0.0, use_abs=True, title="Anchor — Venn (distância = 1 − |r|)")
```

```python
# === Anchor — Venn adaptado (subset de métricas) ===
import re
import unicodedata
import difflib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from scipy.spatial.distance import squareform  # opcional (não usado diretamente aqui)

# -------------------------------------------------
# 1) Corr de métricas do Anchor a partir de df_anchor_metrics
# -------------------------------------------------
def build_corr_from_anchor_metrics(df_anchor_metrics: pd.DataFrame) -> pd.DataFrame:
    df = df_anchor_metrics.copy()

    # remover linhas agregadas, se existirem
    if "instancia" in df.columns:
        df = df[~df["instancia"].isin(["MÉDIA", "MEDIANA"])].copy()

    # remover coluna de id, se presente
    df = df.drop(columns=["instancia"], errors="ignore")

    # converter "xx%" -> número, quando aparecer
    for c in df.columns:
        df[c] = pd.to_numeric(
            df[c].astype(str).str.replace("%", "", regex=False).str.strip(),
            errors="coerce"
        )

    # remover colunas sem variação (constantes) ou totalmente NaN
    nun = df.nunique(dropna=True)
    keep_cols = nun.index[(nun > 1) & (df.notna().sum() > 1)]
    df = df[keep_cols].copy()

    if df.shape[1] < 2:
        raise ValueError("Após limpeza, há menos de 2 métricas numéricas para correlacionar.")

    # correlação de Pearson
    corr = df.corr(method="pearson")
    return corr

# Ex.: construa a matriz de correlação (assumindo df_anchor_metrics já existe)
corr_matrix = build_corr_from_anchor_metrics(df_anchor_metrics)

# -------------------------------------------------
# 2) Normalização e resolução de nomes de métricas (subset)
# -------------------------------------------------
def _normalize_name(s: str) -> str:
    s = str(s)
    s = unicodedata.normalize('NFKD', s)
    s = ''.join(ch for ch in s if not unicodedata.combining(ch))
    s = s.lower().strip()
    s = re.sub(r'\(\s*%\s*\)', '', s)
    s = s.replace('%', '')
    s = re.sub(r'[\s_\-]+', '', s)
    s = re.sub(r'[^a-z0-9]', '', s)
    return s

def _resolve_metric_names(corr_columns, requested_names):
    """Mapeia nomes pedidos para colunas reais da corr_matrix via normalização + fuzzy."""
    norm_to_col = {}
    for col in corr_columns:
        norm_to_col.setdefault(_normalize_name(col), col)

    resolved, missing, suggestions = [], [], {}
    for name in requested_names:
        key = _normalize_name(name)
        if key in norm_to_col:
            resolved.append(norm_to_col[key])
        else:
            choices = list(norm_to_col.keys())
            match = difflib.get_close_matches(key, choices, n=1, cutoff=0.6)
            if match:
                resolved.append(norm_to_col[match[0]])
                suggestions[name] = norm_to_col[match[0]]
            else:
                missing.append(name)
    return resolved, missing, suggestions

# -------------------------------------------------
# 3) Venn adaptado para subset de métricas
# -------------------------------------------------
def venn_from_corr_subset(
    corr_matrix: pd.DataFrame,
    metrics=None,            # lista com os nomes desejados (em qualquer formato)
    radius=1.0,              # raio R (igual p/ todos)
    epsilon_pos_frac=0.06,   # ε (fração de R) só p/ pares com r>0
    random_state=42,
    n_init=8,
    max_iter=600,
    figsize=(8,8),
    face_alpha=0.30,
    edge_lw=1.5,
    label_fontsize=8,
    label_weight='normal',
    label_jitter=0.03,
    repel_iters=150,
    repel_step=0.02,
    inside_margin=0.12,
    title="Anchor — Venn adaptado (subset; d = (1 − r)·R; ε>0 só para r>0)"
):
    # ---------- 0) resolver nomes do subset ----------
    if metrics is not None:
        resolved, missing, sugg = _resolve_metric_names(corr_matrix.columns, metrics)
        if missing and not resolved:
            cols_list = ', '.join(map(str, corr_matrix.columns))
            raise ValueError(
                "Nenhuma métrica do subset encontrada nas colunas da corr_matrix.\n"
                f"Pedidas: {metrics}\n"
                f"Colunas disponíveis: {cols_list}"
            )
        if missing:
            print("[aviso] Métricas não encontradas e ignoradas:", missing)
            if sugg:
                print("[aviso] Sugestões aplicadas:", sugg)
        if not resolved:
            raise ValueError("Após normalização, não sobrou nenhuma métrica válida para plotar.")
        labels = resolved
        C_raw = corr_matrix.loc[labels, labels].to_numpy(dtype=float)
    else:
        labels = corr_matrix.columns.tolist()
        C_raw = corr_matrix.to_numpy(dtype=float)

    # ---------- 1) correlação → distância-alvo ----------
    C_raw = np.clip(C_raw, -1.0, 1.0)
    R = float(radius)
    T = (1.0 - C_raw) * R  # d(r) base
    if epsilon_pos_frac and epsilon_pos_frac > 0:
        mask_pos = (C_raw > 0).astype(float)  # ε só quando r>0
        T += (epsilon_pos_frac * R) * mask_pos

    np.fill_diagonal(T, 0.0)
    T = 0.5 * (T + T.T)

    # ---------- 2) embedding (MDS SMACOF; fallback p/ clássico) ----------
    try:
        from sklearn.manifold import MDS
        mds = MDS(n_components=2, metric=True, dissimilarity='precomputed',
                  n_init=n_init, max_iter=max_iter, random_state=random_state)
        X = mds.fit_transform(T)
    except Exception:
        n = T.shape[0]
        J = np.eye(n) - np.ones((n, n))/n
        B = -0.5 * J @ (T**2) @ J
        eigvals, eigvecs = np.linalg.eigh(B)
        idx = np.argsort(eigvals)[::-1]
        L = np.clip(eigvals[idx][:2], 0, None)
        V = eigvecs[:, idx][:, :2]
        X = V * np.sqrt(L + 1e-12)

    # reescala global para aproximar T
    DX = np.sqrt(((X[:,None,:]-X[None,:,:])**2).sum(axis=2))
    iu = np.triu_indices_from(DX, k=1)
    num = np.sum(DX[iu] * T[iu])
    den = np.sum(DX[iu]**2) + 1e-12
    s = num/den if den > 0 else 1.0
    X *= s
    X -= X.mean(axis=0, keepdims=True)

    # ---------- 3) cores (fixas para métricas comuns de Anchor; fallback tab20) ----------
    def _lc(s): return str(s).strip().lower()
    color_fixed = {
        "robustness":            "#9467bd",
        "stability":             "#17becf",
        "soundness":             "#f7b6d2",
        "directional soundness": "#ff9896",
        "selectivity":           "#c7c7c7",
        "fidelity":              "#2ca02c",
        "infidelity":            "#d62728",
        "coverage":              "#1f77b4",
        "completeness":          "#1f77b4"
    }
    cmap = plt.get_cmap("tab20")
    colors, k_auto = [], 0
    for name in labels:
        key = _lc(name)
        colors.append(color_fixed.get(key, cmap(k_auto % 20)))
        if key not in color_fixed: k_auto += 1

    # ---------- 4) círculos ----------
    fig, ax = plt.subplots(figsize=figsize, dpi=150, facecolor="white")
    for (x, y), col in zip(X, colors):
        circ = Circle((x, y), R, facecolor=col, edgecolor=col, alpha=face_alpha, lw=edge_lw)
        ax.add_patch(circ)

    # ---------- 5) rótulos centralizados + repulsão vertical ----------
    rng = np.random.default_rng(random_state)
    text_xy = X.copy()
    text_xy[:, 1] += (rng.uniform(-label_jitter, label_jitter, size=len(labels))) * R

    def clamp_inside(i):
        cx, cy = X[i]
        tx, ty = text_xy[i]
        vx, vy = tx - cx, ty - cy
        dist = np.hypot(vx, vy)
        maxd = R * (1.0 - inside_margin)
        if dist > maxd and dist > 1e-12:
            k = maxd / dist
            text_xy[i] = np.array([cx + k*vx, cy + k*vy])

    for _ in range(repel_iters):
        moved = False
        for i in range(len(labels)):
            for j in range(i+1, len(labels)):
                xi, yi = text_xy[i]; xj, yj = text_xy[j]
                w = 0.9 * R
                h = 0.35 * R
                if (abs(xi - xj) < w) and (abs(yi - yj) < h):
                    dy = repel_step * R
                    text_xy[i, 1] += dy
                    text_xy[j, 1] -= dy
                    clamp_inside(i); clamp_inside(j)
                    moved = True
        if not moved:
            break

    for (tx, ty), name in zip(text_xy, labels):
        ax.text(tx, ty, name, ha="center", va="center",
                fontsize=label_fontsize, fontweight=label_weight,
                color="black", zorder=3)

    # ---------- 6) estética ----------
    xmin = X[:,0].min() - R*1.3
    xmax = X[:,0].max() + R*1.3
    ymin = X[:,1].min() - R*1.3
    ymax = X[:,1].max() + R*1.3
    ax.set_aspect("equal")
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    ax.axis("off")
    ax.set_title(title, fontsize=10)
    plt.tight_layout()
    plt.show()

# ===== Exemplo de uso =====
subset = ["Robustness", "Fidelity", "Infidelity", "Coverage"]  # nomes podem variar; há fuzzy-match
venn_from_corr_subset(
    corr_matrix,
    metrics=subset,
    radius=1.0,
    epsilon_pos_frac=0.06,
    figsize=(8, 8),
    label_fontsize=8
)
```

```python
# === UpSet de Correlação entre Métricas ANCHOR ===
# Nomes das métricas à esquerda (padrão) + replicados na base (vertical)
# Requer: df_anchor_metrics e (opcional) BASE_PATH (str ou Path)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# upsetplot
try:
    from upsetplot import UpSet, from_memberships
except Exception:
    raise ImportError("Instale: pip install upsetplot")

# ----------------- CONFIG -----------------
TAU = 0.30        # limiar de |correlação| p/ considerar “alta correlação”
MIN_GROUP = 2     # grupos com pelo menos 2 métricas
SAFE_BASE = Path(BASE_PATH) if 'BASE_PATH' in globals() else Path.cwd()
OUTFILE = SAFE_BASE / "upset_anchor_corr.png"

# ----------------- PREPARO -----------------
# remove média/mediana
dfu = df_anchor_metrics[
    ~df_anchor_metrics["instancia"].astype(str).str.upper().isin(["MÉDIA","MEDIANA"])
].copy()

# seleciona métricas "(%)" e converte p/ float
metric_cols = [c for c in dfu.columns if c.endswith("(%)")]
if not metric_cols:
    raise ValueError("Não encontrei colunas de métricas no formato 'nome(%)'.")

M = (
    dfu[metric_cols]
      .replace('%', '', regex=True)
      .replace('N/A', np.nan)
      .apply(pd.to_numeric, errors='coerce')
)

# correlação coluna-coluna
corr = M.corr(method="pearson")

# memberships: para cada métrica i, conjunto das que têm |r| ≥ TAU com i
metrics = corr.columns.tolist()
memberships, labels_anchor = [], []
for i in metrics:
    group = [
        j for j in metrics
        if j != i and np.isfinite(corr.loc[i, j]) and abs(corr.loc[i, j]) >= TAU
    ]
    if len(group) >= MIN_GROUP:
        memberships.append(group)
        labels_anchor.append(i)

if not memberships:
    print(f"⚠️ Nenhum grupo de métricas com |r| ≥ {TAU:.2f}. Tente reduzir TAU.")
else:
    data = from_memberships(memberships)

    # ----------------- PLOT (horizontal: nomes à esquerda) -----------------
    fig = plt.figure(figsize=(22, 10), dpi=300)
    up = UpSet(
        data,
        show_counts=True,
        sort_by="cardinality",
        sort_categories_by="cardinality",
        element_size=30,
        orientation="horizontal",  # mantém nomes à esquerda
        subset_size="count"        # 👈 CORREÇÃO: lida com grupos não únicos
    )
    up.plot()

    # --------- Faixa auxiliar com nomes das métricas na base ----------
    # Encontrar, de forma robusta, o eixo que contém os rótulos Y das métricas
    fig = plt.gcf()
    metric_set = set(metrics)
    left_axis = None
    for ax in fig.axes:
        ytxt = [t.get_text() for t in ax.get_yticklabels()]
        match_count = sum(1 for t in ytxt if t in metric_set)
        if match_count >= max(3, len(metric_set) // 2):  # heurística robusta
            left_axis = ax
            break

    if left_axis is None:
        # fallback: usa o eixo com mais rótulos de métricas
        best_ax, best_score = None, -1
        for ax in fig.axes:
            score = sum(1 for t in ax.get_yticklabels() if t.get_text() in metric_set)
            if score > best_score:
                best_ax, best_score = ax, score
        left_axis = best_ax

    # Se ainda assim não achou, evita crash
    if left_axis is not None:
        pos = left_axis.get_position()  # bbox do eixo da matriz

        # >>> AJUSTES APENAS NOS LABELS DA PARTE INFERIOR <<<
        gutter_h = 0.10  # aumenta a altura da faixa inferior para não sobrepor
        bottom = max(0.02, pos.y0 - gutter_h - 0.02)
        ax_bottom = fig.add_axes([pos.x0, bottom, pos.width, gutter_h])
        ax_bottom.set_xlim(0, 1)
        ax_bottom.set_ylim(-0.2, 1.0)
        ax_bottom.axis("off")

        # ordem dos nomes conforme aparecem à esquerda
        ylabels = [lab.get_text() for lab in left_axis.get_yticklabels() if lab.get_text()]
        n = len(ylabels)
        if n > 0:
            xs = np.linspace(0.0, 1.0, n, endpoint=True)
            for x, name in zip(xs, ylabels):
                ax_bottom.text(
                    x, -0.12, name,
                    rotation=90,
                    ha="center",
                    va="top",
                    fontsize=10
                )

    plt.suptitle(f"UpSet anchor — grupos de métricas com |corr| ≥ {TAU:.2f}", y=0.98)
    plt.savefig(OUTFILE, dpi=600, bbox_inches="tight")
    plt.show()
    print(f"✅ UpSet (correlação) anchor salvo em: {OUTFILE}")
```
